```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
# NOTE: Replace these paths with actual working paths if running outside a Kaggle environment
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    # Load and clean metadata
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    # Helper function to create dataframe from file structure
    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Merge dataframes
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Standard normalization for CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
# NOTE: Using the user's provided segmentation model definitions

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None and not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as test data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        # Use a small region minimum size to prevent issues with single pixels
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props).head(1) # Take only the largest region if multiple exist
                    except:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        # The shape features are: area, perimeter, solidity, eccentricity, circularity (5 features)
        n_shape_features = train_features_df.shape[1] 
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3), A.GaussNoise(p=0.2), A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), 
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.8), A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3), A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes and self.minority_transform:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                # Ensure mask is the right size/shape for element-wise multiplication
                pred_mask = self.masks[idx].to(img_tensor.device) 
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # *** UPDATED HERE (1st change) ***
        # Revert back to only 'Severe' (label 2) as the primary minority class
        minority_classes_list = [2] # <-- REVERTED
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler for class balancing
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=sampler, num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler. ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        
        # PLACEHOLDER MODELS (since external weights/complex definitions are unavailable)
        class CoAnNetFeatureExtractor(nn.Module):
            # CoAnNet is complex; we use a simpler block with the expected output dimension
            def __init__(self):
                super().__init__()
                self.conv = nn.Conv2d(3, 512, kernel_size=3, padding=1)
                self.pool = nn.AdaptiveAvgPool2d((1, 1))
                self.identity = nn.Identity()
            def forward(self, x):
                x = F.relu(self.conv(x))
                return self.identity(self.pool(x).flatten(1))

        class SimCLRResNet50(nn.Module):
            # SimCLR is ResNet50 with special weights; we use a standard pre-trained one
            def __init__(self):
                super().__init__()
                self.base = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
                self.base.fc = nn.Identity()
            def forward(self, x):
                return self.base(x)


        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name: str, n_classes=3, n_shape_features=5):
                super().__init__()
                
                cnn_feature_dim = 0
                
                if cnn_name == 'resnet34':
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'resnet50' or cnn_name == 'ResNet50_GAP':
                    # ResNet50 and ResNet50_GAP are the same in feature extraction layer, GAP is handled by AdaptiveAvgPool2d.
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                elif cnn_name == 'EfficientNet_B0':
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier[1].in_features
                    self.cnn_base.classifier = nn.Identity() # Remove the default classifier
                elif cnn_name == 'SimCLR_ResNet50':
                    self.cnn_base = SimCLRResNet50() # Custom/Pre-loaded weights are required for true SimCLR
                    cnn_feature_dim = 2048
                elif cnn_name == 'CoAnNet':
                    self.cnn_base = CoAnNetFeatureExtractor() # Placeholder
                    cnn_feature_dim = 512 # Placeholder dimension
                else:
                    raise ValueError(f"Unknown CNN name: {cnn_name}")
                    
                print(f"Loaded {cnn_name} encoder (Dim: {cnn_feature_dim})")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                # Ensure 1D output after CNN
                if cnn_features.ndim > 2:
                    # Adaptive Avg Pool for any remaining spatial dimensions (e.g., in ResNet/EfficientNet backbone output)
                    cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1) 
                    
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output

        # --- 7.5 TRAIN AND EVALUATE FUNCTION PER MODEL ---
        def train_and_evaluate_single_model(cnn_name: str, train_loader, val_loader, test_loader, class_weights, n_shape_features, device, num_epochs=NUM_EPOCHS, patience=10):
            """Trains and evaluates a single hybrid model."""
            print(f"\n=============================================")
            print(f"ğŸš€ Starting Training for Model: {cnn_name}")
            print(f"=============================================")
            
            cls_model = HybridClassifier(cnn_name=cnn_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            
            # *** UPDATED HERE (2nd change) ***
            # Try a less aggressive gamma for Focal Loss
            criterion = FocalLoss(alpha=class_weights, gamma=2.0) # <-- UPDATED from 3.0 to 2.0
            
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5)
            scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=False)
            
            epochs_no_improve = 0; best_val_accuracy = 0.0
            model_save_path = f'best_classifier_model_{cnn_name}.pth'
            
            for epoch in range(num_epochs):
                cls_model.train()
                running_loss = 0.0
                for images, shapes, labels in train_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    optimizer.zero_grad()
                    outputs = cls_model(images, shapes)
                    loss = criterion(outputs, labels); loss.backward(); optimizer.step()
                    running_loss += loss.item()
                
                cls_model.eval()
                val_corrects = 0
                with torch.no_grad():
                    for images, shapes, labels in val_loader:
                        images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                        outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                        val_corrects += torch.sum(preds == labels.data)

                val_accuracy = val_corrects.double() / len(val_loader.dataset)
                scheduler.step(val_accuracy)

                # Early stopping logic
                if val_accuracy > best_val_accuracy:
                    best_val_accuracy = val_accuracy
                    torch.save(cls_model.state_dict(), model_save_path)
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    
                if epochs_no_improve >= patience:
                    print(f"â¹ï¸ Early stopping triggered for {cnn_name} at epoch {epoch+1}.")
                    break
                    
            print(f"Final Best Validation Accuracy for {cnn_name}: {best_val_accuracy:.4f}")

            # Final Test Evaluation
            try:
                cls_model.load_state_dict(torch.load(model_save_path, map_location=device))
            except Exception as e:
                 print(f"âš ï¸ Could not load best weights for {cnn_name}. Using current weights for testing. Error: {e}")
            cls_model.eval()
            all_labels, all_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in test_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                    all_labels.extend(labels.cpu().numpy()); all_preds.extend(preds.cpu().numpy())
            
            report = classification_report(all_labels, all_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            return np.array(all_preds), report

        # --- 7.6 MASTER EXECUTION AND COMPARISON ---
        
        # ADD ALL MODELS TO BE EVALUATED
        model_list = [
            'resnet34', 
            'resnet50', 
            'densenet121', 
            'EfficientNet_B0',
            'ResNet50_GAP',      # Standard ResNet50 encoder using GAP
            'SimCLR_ResNet50',   # ResNet50 with placeholder SimCLR weights
            'CoAnNet'            # Placeholder architecture
        ]
        results = {} 

        for model_name in model_list:
            preds, report = train_and_evaluate_single_model(
                cnn_name=model_name,
                train_loader=train_cls_loader, val_loader=val_cls_loader, test_loader=test_cls_loader,
                class_weights=class_weights, n_shape_features=n_shape_features, device=device
            )
            results[model_name] = {'preds': preds, 'report': report}

        # --- 7.7 FINAL ENSEMBLE SELECTION AND REPORT ---
        print("\n=============================================")
        print("ğŸ† FINAL RESULT SELECTION (PER-CLASS BEST RECALL)")
        print("=============================================")

        # 1. Get ground truth labels
        all_labels = np.concatenate([labels.cpu().numpy() for _, _, labels in test_cls_loader])
        final_preds = np.zeros_like(all_labels)
        
        best_model_per_class = {}
        
        # 2. Determine the best model for each class based on 'recall' (accuracy per class)
        comparison_data = []
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                recall = res['report'].get(class_name, {}).get('recall', 0.0) # Handle missing class gracefully
                row[model_name] = f'{recall:.4f}'
                
                if recall > best_recall:
                    best_recall = recall
                    best_model = model_name
                    
            best_model_per_class[i] = best_model
            row['Best Model'] = best_model
            row['Best Recall'] = f'{best_recall:.4f}'
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            class_indices = np.where(all_labels == i)[0]
            if best_model:
                 final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print("\n--- PER-CLASS RECALL COMPARISON ---")
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
# NOTE: Replace these paths with actual working paths if running outside a Kaggle environment
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    # Load and clean metadata
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    # Helper function to create dataframe from file structure
    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Merge dataframes
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Standard normalization for CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
# NOTE: Using the user's provided segmentation model definitions

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None and not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as test data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        # Use a small region minimum size to prevent issues with single pixels
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props).head(1) # Take only the largest region if multiple exist
                    except:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        # The shape features are: area, perimeter, solidity, eccentricity, circularity (5 features)
        n_shape_features = train_features_df.shape[1] 
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3), A.GaussNoise(p=0.2), A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), 
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.8), A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3), A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes and self.minority_transform:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                # Ensure mask is the right size/shape for element-wise multiplication
                pred_mask = self.masks[idx].to(img_tensor.device) 
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # *** UPDATED HERE ***
        # Apply minority augmentation to 'Moderate' (1) and 'Severe' (2)
        minority_classes_list = [1, 2] # <-- UPDATED
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler for class balancing
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=sampler, num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler. ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        
        # PLACEHOLDER MODELS (since external weights/complex definitions are unavailable)
        class CoAnNetFeatureExtractor(nn.Module):
            # CoAnNet is complex; we use a simpler block with the expected output dimension
            def __init__(self):
                super().__init__()
                self.conv = nn.Conv2d(3, 512, kernel_size=3, padding=1)
                self.pool = nn.AdaptiveAvgPool2d((1, 1))
                self.identity = nn.Identity()
            def forward(self, x):
                x = F.relu(self.conv(x))
                return self.identity(self.pool(x).flatten(1))

        class SimCLRResNet50(nn.Module):
            # SimCLR is ResNet50 with special weights; we use a standard pre-trained one
            def __init__(self):
                super().__init__()
                self.base = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
                self.base.fc = nn.Identity()
            def forward(self, x):
                return self.base(x)


        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name: str, n_classes=3, n_shape_features=5):
                super().__init__()
                
                cnn_feature_dim = 0
                
                if cnn_name == 'resnet34':
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'resnet50' or cnn_name == 'ResNet50_GAP':
                    # ResNet50 and ResNet50_GAP are the same in feature extraction layer, GAP is handled by AdaptiveAvgPool2d.
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                elif cnn_name == 'EfficientNet_B0':
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier[1].in_features
                    self.cnn_base.classifier = nn.Identity() # Remove the default classifier
                elif cnn_name == 'SimCLR_ResNet50':
                    self.cnn_base = SimCLRResNet50() # Custom/Pre-loaded weights are required for true SimCLR
                    cnn_feature_dim = 2048
                elif cnn_name == 'CoAnNet':
                    self.cnn_base = CoAnNetFeatureExtractor() # Placeholder
                    cnn_feature_dim = 512 # Placeholder dimension
                else:
                    raise ValueError(f"Unknown CNN name: {cnn_name}")
                    
                print(f"Loaded {cnn_name} encoder (Dim: {cnn_feature_dim})")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                # Ensure 1D output after CNN
                if cnn_features.ndim > 2:
                    # Adaptive Avg Pool for any remaining spatial dimensions (e.g., in ResNet/EfficientNet backbone output)
                    cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1) 
                    
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output

        # --- 7.5 TRAIN AND EVALUATE FUNCTION PER MODEL ---
        def train_and_evaluate_single_model(cnn_name: str, train_loader, val_loader, test_loader, class_weights, n_shape_features, device, num_epochs=NUM_EPOCHS, patience=10):
            """Trains and evaluates a single hybrid model."""
            print(f"\n=============================================")
            print(f"ğŸš€ Starting Training for Model: {cnn_name}")
            print(f"=============================================")
            
            cls_model = HybridClassifier(cnn_name=cnn_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            criterion = FocalLoss(alpha=class_weights, gamma=3.0)
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5)
            scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=False)
            
            epochs_no_improve = 0; best_val_accuracy = 0.0
            model_save_path = f'best_classifier_model_{cnn_name}.pth'
            
            for epoch in range(num_epochs):
                cls_model.train()
                running_loss = 0.0
                for images, shapes, labels in train_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    optimizer.zero_grad()
                    outputs = cls_model(images, shapes)
                    loss = criterion(outputs, labels); loss.backward(); optimizer.step()
                    running_loss += loss.item()
                
                cls_model.eval()
                val_corrects = 0
                with torch.no_grad():
                    for images, shapes, labels in val_loader:
                        images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                        outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                        val_corrects += torch.sum(preds == labels.data)

                val_accuracy = val_corrects.double() / len(val_loader.dataset)
                scheduler.step(val_accuracy)

                # Early stopping logic
                if val_accuracy > best_val_accuracy:
                    best_val_accuracy = val_accuracy
                    torch.save(cls_model.state_dict(), model_save_path)
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    
                if epochs_no_improve >= patience:
                    print(f"â¹ï¸ Early stopping triggered for {cnn_name} at epoch {epoch+1}.")
                    break
                    
            print(f"Final Best Validation Accuracy for {cnn_name}: {best_val_accuracy:.4f}")

            # Final Test Evaluation
            try:
                cls_model.load_state_dict(torch.load(model_save_path, map_location=device))
            except Exception as e:
                 print(f"âš ï¸ Could not load best weights for {cnn_name}. Using current weights for testing. Error: {e}")
            cls_model.eval()
            all_labels, all_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in test_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                    all_labels.extend(labels.cpu().numpy()); all_preds.extend(preds.cpu().numpy())
            
            report = classification_report(all_labels, all_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            return np.array(all_preds), report

        # --- 7.6 MASTER EXECUTION AND COMPARISON ---
        
        # ADD ALL MODELS TO BE EVALUATED
        model_list = [
            'resnet34', 
            'resnet50', 
            'densenet121', 
            'EfficientNet_B0',
            'ResNet50_GAP',      # Standard ResNet50 encoder using GAP
            'SimCLR_ResNet50',   # ResNet50 with placeholder SimCLR weights
            'CoAnNet'            # Placeholder architecture
        ]
        results = {} 

        for model_name in model_list:
            preds, report = train_and_evaluate_single_model(
                cnn_name=model_name,
                train_loader=train_cls_loader, val_loader=val_cls_loader, test_loader=test_cls_loader,
                class_weights=class_weights, n_shape_features=n_shape_features, device=device
            )
            results[model_name] = {'preds': preds, 'report': report}

        # --- 7.7 FINAL ENSEMBLE SELECTION AND REPORT ---
        print("\n=============================================")
        print("ğŸ† FINAL RESULT SELECTION (PER-CLASS BEST RECALL)")
        print("=============================================")

        # 1. Get ground truth labels
        all_labels = np.concatenate([labels.cpu().numpy() for _, _, labels in test_cls_loader])
        final_preds = np.zeros_like(all_labels)
        
        best_model_per_class = {}
        
        # 2. Determine the best model for each class based on 'recall' (accuracy per class)
        comparison_data = []
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                recall = res['report'].get(class_name, {}).get('recall', 0.0) # Handle missing class gracefully
                row[model_name] = f'{recall:.4f}'
                
                if recall > best_recall:
                    best_recall = recall
                    best_model = model_name
                    
            best_model_per_class[i] = best_model
            row['Best Model'] = best_model
            row['Best Recall'] = f'{best_recall:.4f}'
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            class_indices = np.where(all_labels == i)[0]
            if best_model:
                 final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print("\n--- PER-CLASS RECALL COMPARISON ---")
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
# NOTE: Replace these paths with actual working paths if running outside a Kaggle environment
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    # Load and clean metadata
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    # Helper function to create dataframe from file structure
    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Merge dataframes
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = pd.DataFrame() # Ensure test_df is not None
    train_df = pd.DataFrame()
    val_df = pd.DataFrame()

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Standard normalization for CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            # Ensure alpha_t is on the same device as the loss
            alpha_t = self.alpha[target.data.view(-1)].to(device)
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        # Use a small region minimum size to prevent issues with single pixels
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props).head(1) # Take only the largest region if multiple exist
                    except:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            # Select specific features (area, perimeter, solidity, eccentricity, circularity)
            return feature_df[['area', 'perimeter', 'solidity', 'eccentricity', 'circularity']]

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        # The shape features are: area, perimeter, solidity, eccentricity, circularity (5 features)
        n_shape_features = train_features_df.shape[1] 
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        
        # (Transforms from previous step)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(p=0.5),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])


        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes and self.minority_transform:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                pred_mask = self.masks[idx].to(img_tensor.device) 
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # --- MODIFIED: Focusing on Class 1 and Class 2 ---
        minority_classes_list = [1, 2] 
        
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler for class balancing
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=sampler, num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler. ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name='resnet50', n_classes=NUM_CLASSES, n_shape_features=5):
                super().__init__()
                if cnn_name == 'resnet34':
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                
                # Includes resnet50, simclr_resnet50, resnet50_gap, coanet, wsp_resnet50
                elif cnn_name in ['resnet50', 'simclr_resnet50', 'resnet50_gap', 'coanet', 'wsp_resnet50']:
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                
                elif cnn_name == 'efficientnet-b0':
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier[-1].in_features
                    self.cnn_base.classifier = nn.Identity()
                elif cnn_name == 'efficientnet-b1':
                    self.cnn_base = models.efficientnet_b1(weights=models.EfficientNet_B1_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier[-1].in_features
                    self.cnn_base.classifier = nn.Identity()
                elif cnn_name == 'swin_t':
                    self.cnn_base = models.swin_t(weights=models.Swin_T_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.head.in_features
                    self.cnn_base.head = nn.Identity()
                
                else:
                    raise ValueError(f"Unknown CNN: {cnn_name}")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64),
                    nn.ReLU(),
                    nn.Dropout(0.4),
                    nn.Linear(64, 32),
                    nn.ReLU(),
                    nn.Dropout(0.4)
                )

                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256),
                    nn.ReLU(),
                    nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )

            def forward(self, images, shape_features):
                cnn_feat = self.cnn_base(images)
                if cnn_feat.ndim > 2:
                    cnn_feat = F.adaptive_avg_pool2d(cnn_feat, (1, 1)).flatten(1)
                tab_feat = self.tabular_mlp(shape_features)
                combined = torch.cat([cnn_feat, tab_feat], dim=1)
                out = self.fusion_classifier(combined)
                return out

        # --- 7.5 TRAIN AND EVALUATE FUNCTION PER MODEL ---
        def train_and_evaluate_single_model(cnn_name, train_loader, val_loader, test_loader, class_weights, n_shape_features, device, num_epochs=NUM_EPOCHS, patience=10):
            print(f"\n--- ğŸš€ Training {cnn_name} ---")
            cls_model = HybridClassifier(cnn_name=cnn_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            criterion = FocalLoss(alpha=class_weights, gamma=3.0)
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5)
            scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=False)

            best_val_acc = 0.0
            epochs_no_improve = 0
            model_save_path = f'best_classifier_{cnn_name}.pth'

            for epoch in range(num_epochs):
                cls_model.train()
                for images, shapes, labels in train_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    optimizer.zero_grad()
                    outputs = cls_model(images, shapes)
                    loss = criterion(outputs, labels)
                    loss.backward()
                    optimizer.step()

                # Validation
                cls_model.eval()
                correct = 0
                with torch.no_grad():
                    for images, shapes, labels in val_loader:
                        images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                        outputs = cls_model(images, shapes)
                        preds = outputs.argmax(1)
                        correct += (preds == labels).sum().item()
                val_acc = correct / len(val_loader.dataset)

                scheduler.step(val_acc)
                if val_acc > best_val_acc:
                    best_val_acc = val_acc
                    torch.save(cls_model.state_dict(), model_save_path)
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    if epochs_no_improve >= patience:
                        print(f"â¹ï¸ Early stopping for {cnn_name} at epoch {epoch+1}")
                        break

            # Test predictions
            print(f"Loading best weights for {cnn_name} (Val Acc: {best_val_acc:.4f}) for testing...")
            cls_model.load_state_dict(torch.load(model_save_path))
            cls_model.eval()
            all_labels, all_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in test_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    preds = outputs.argmax(1)
                    all_labels.extend(labels.cpu().numpy())
                    all_preds.extend(preds.cpu().numpy())

            report = classification_report(all_labels, all_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            return np.array(all_preds), report


        # --- 7.6 MASTER EXECUTION AND COMPARISON ---
        
        # --- UPDATED MODEL LIST ---
        model_list = [
            'resnet34', 
            'resnet50', 
            'densenet121', 
            'resnet50_gap', 
            'simclr_resnet50', 
            'coanet',
            'efficientnet-b0',
            'efficientnet-b1',
            'swin_t',
            'wsp_resnet50'
        ]
        results = {} 

        for model_name in model_list:
            try:
                preds, report = train_and_evaluate_single_model(
                    cnn_name=model_name,
                    train_loader=train_cls_loader, val_loader=val_cls_loader, test_loader=test_cls_loader,
                    class_weights=class_weights, n_shape_features=n_shape_features, device=device
                )
                results[model_name] = {'preds': preds, 'report': report}
            except Exception as e:
                print(f"!! Error training {model_name}: {e}. Skipping this model.")
                import traceback
                traceback.print_exc()

        if not results:
             print("No models were successfully trained. Exiting classification stage.")
             raise SystemExit("Classification failed: No model results.")


         # --- 7.7 FINAL ENSEMBLE SELECTION AND REPORT ---
        print("\n=============================================")
        print("ğŸ† FINAL RESULT SELECTION (PER-CLASS BEST RECALL)")
        print("=============================================")

        # 1. Get ground truth labels
        all_labels = np.concatenate([labels.cpu().numpy() for _, _, labels in test_cls_loader])
        final_preds = np.zeros_like(all_labels)
        
        best_model_per_class = {}
        
        # 2. Determine the best model for each class based on 'recall' (accuracy per class)
        comparison_data = []
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                recall = res['report'].get(class_name, {}).get('recall', 0.0) # Handle missing class gracefully
                row[model_name] = f'{recall:.4f}'
                
                if recall > best_recall:
                    best_recall = recall
                    best_model = model_name
                    
            best_model_per_class[i] = best_model
            row['Best Model'] = best_model
            row['Best Recall'] = f'{best_recall:.4f}'
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            class_indices = np.where(all_labels == i)[0]
            if best_model:
                 final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print("\n--- PER-CLASS RECALL COMPARISON ---")
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
# NOTE: Replace these paths with actual working paths if running outside a Kaggle environment
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    # Load and clean metadata
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    # Helper function to create dataframe from file structure
    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Merge dataframes
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Standard normalization for CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
# NOTE: Using the user's provided segmentation model definitions

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None and not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as test data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        # Use a small region minimum size to prevent issues with single pixels
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props).head(1) # Take only the largest region if multiple exist
                    except:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        # The shape features are: area, perimeter, solidity, eccentricity, circularity (5 features)
        n_shape_features = train_features_df.shape[1] 
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3), A.GaussNoise(p=0.2), A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), 
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.8), A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3), A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes and self.minority_transform:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                # Ensure mask is the right size/shape for element-wise multiplication
                pred_mask = self.masks[idx].to(img_tensor.device) 
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Assuming 'Severe' (label 2) is the primary minority class
        minority_classes_list = [2] 
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler for class balancing
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=sampler, num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler. ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        
        # PLACEHOLDER MODELS (since external weights/complex definitions are unavailable)
        class CoAnNetFeatureExtractor(nn.Module):
            # CoAnNet is complex; we use a simpler block with the expected output dimension
            def __init__(self):
                super().__init__()
                self.conv = nn.Conv2d(3, 512, kernel_size=3, padding=1)
                self.pool = nn.AdaptiveAvgPool2d((1, 1))
                self.identity = nn.Identity()
            def forward(self, x):
                x = F.relu(self.conv(x))
                return self.identity(self.pool(x).flatten(1))

        class SimCLRResNet50(nn.Module):
            # SimCLR is ResNet50 with special weights; we use a standard pre-trained one
            def __init__(self):
                super().__init__()
                self.base = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
                self.base.fc = nn.Identity()
            def forward(self, x):
                return self.base(x)


        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name: str, n_classes=3, n_shape_features=5):
                super().__init__()
                
                cnn_feature_dim = 0
                
                if cnn_name == 'resnet34':
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'resnet50' or cnn_name == 'ResNet50_GAP':
                    # ResNet50 and ResNet50_GAP are the same in feature extraction layer, GAP is handled by AdaptiveAvgPool2d.
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                elif cnn_name == 'EfficientNet_B0':
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier[1].in_features
                    self.cnn_base.classifier = nn.Identity() # Remove the default classifier
                elif cnn_name == 'SimCLR_ResNet50':
                    self.cnn_base = SimCLRResNet50() # Custom/Pre-loaded weights are required for true SimCLR
                    cnn_feature_dim = 2048
                elif cnn_name == 'CoAnNet':
                    self.cnn_base = CoAnNetFeatureExtractor() # Placeholder
                    cnn_feature_dim = 512 # Placeholder dimension
                else:
                    raise ValueError(f"Unknown CNN name: {cnn_name}")
                    
                print(f"Loaded {cnn_name} encoder (Dim: {cnn_feature_dim})")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                # Ensure 1D output after CNN
                if cnn_features.ndim > 2:
                    # Adaptive Avg Pool for any remaining spatial dimensions (e.g., in ResNet/EfficientNet backbone output)
                    cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1) 
                    
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output

        # --- 7.5 TRAIN AND EVALUATE FUNCTION PER MODEL ---
        def train_and_evaluate_single_model(cnn_name: str, train_loader, val_loader, test_loader, class_weights, n_shape_features, device, num_epochs=NUM_EPOCHS, patience=10):
            """Trains and evaluates a single hybrid model."""
            print(f"\n=============================================")
            print(f"ğŸš€ Starting Training for Model: {cnn_name}")
            print(f"=============================================")
            
            cls_model = HybridClassifier(cnn_name=cnn_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            criterion = FocalLoss(alpha=class_weights, gamma=3.0)
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5)
            scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=False)
            
            epochs_no_improve = 0; best_val_accuracy = 0.0
            model_save_path = f'best_classifier_model_{cnn_name}.pth'
            
            for epoch in range(num_epochs):
                cls_model.train()
                running_loss = 0.0
                for images, shapes, labels in train_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    optimizer.zero_grad()
                    outputs = cls_model(images, shapes)
                    loss = criterion(outputs, labels); loss.backward(); optimizer.step()
                    running_loss += loss.item()
                
                cls_model.eval()
                val_corrects = 0
                with torch.no_grad():
                    for images, shapes, labels in val_loader:
                        images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                        outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                        val_corrects += torch.sum(preds == labels.data)

                val_accuracy = val_corrects.double() / len(val_loader.dataset)
                scheduler.step(val_accuracy)

                # Early stopping logic
                if val_accuracy > best_val_accuracy:
                    best_val_accuracy = val_accuracy
                    torch.save(cls_model.state_dict(), model_save_path)
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    
                if epochs_no_improve >= patience:
                    print(f"â¹ï¸ Early stopping triggered for {cnn_name} at epoch {epoch+1}.")
                    break
                    
            print(f"Final Best Validation Accuracy for {cnn_name}: {best_val_accuracy:.4f}")

            # Final Test Evaluation
            try:
                cls_model.load_state_dict(torch.load(model_save_path, map_location=device))
            except Exception as e:
                 print(f"âš ï¸ Could not load best weights for {cnn_name}. Using current weights for testing. Error: {e}")
            cls_model.eval()
            all_labels, all_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in test_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                    all_labels.extend(labels.cpu().numpy()); all_preds.extend(preds.cpu().numpy())
            
            report = classification_report(all_labels, all_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            return np.array(all_preds), report

        # --- 7.6 MASTER EXECUTION AND COMPARISON ---
        
        # ADD ALL MODELS TO BE EVALUATED
        model_list = [
            'resnet34', 
            'resnet50', 
            'densenet121', 
            'EfficientNet_B0',
            'ResNet50_GAP',      # Standard ResNet50 encoder using GAP
            'SimCLR_ResNet50',   # ResNet50 with placeholder SimCLR weights
            'CoAnNet'            # Placeholder architecture
        ]
        results = {} 

        for model_name in model_list:
            preds, report = train_and_evaluate_single_model(
                cnn_name=model_name,
                train_loader=train_cls_loader, val_loader=val_cls_loader, test_loader=test_cls_loader,
                class_weights=class_weights, n_shape_features=n_shape_features, device=device
            )
            results[model_name] = {'preds': preds, 'report': report}

        # --- 7.7 FINAL ENSEMBLE SELECTION AND REPORT ---
        print("\n=============================================")
        print("ğŸ† FINAL RESULT SELECTION (PER-CLASS BEST RECALL)")
        print("=============================================")

        # 1. Get ground truth labels
        all_labels = np.concatenate([labels.cpu().numpy() for _, _, labels in test_cls_loader])
        final_preds = np.zeros_like(all_labels)
        
        best_model_per_class = {}
        
        # 2. Determine the best model for each class based on 'recall' (accuracy per class)
        comparison_data = []
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                recall = res['report'].get(class_name, {}).get('recall', 0.0) # Handle missing class gracefully
                row[model_name] = f'{recall:.4f}'
                
                if recall > best_recall:
                    best_recall = recall
                    best_model = model_name
                    
            best_model_per_class[i] = best_model
            row['Best Model'] = best_model
            row['Best Recall'] = f'{best_recall:.4f}'
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            class_indices = np.where(all_labels == i)[0]
            if best_model:
                 final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print("\n--- PER-CLASS RECALL COMPARISON ---")
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
# NOTE: Replace these paths with actual working paths if running outside a Kaggle environment
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    # Load and clean metadata
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    # Helper function to create dataframe from file structure
    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Merge dataframes
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Standard normalization for CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
# NOTE: Using the user's provided segmentation model definitions

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None and not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as test data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        # Use a small region minimum size to prevent issues with single pixels
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props).head(1) # Take only the largest region if multiple exist
                    except:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        # The shape features are: area, perimeter, solidity, eccentricity, circularity (5 features)
        n_shape_features = train_features_df.shape[1] 
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3), A.GaussNoise(p=0.2), A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), 
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.8), A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3), A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes and self.minority_transform:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                # Ensure mask is the right size/shape for element-wise multiplication
                pred_mask = self.masks[idx].to(img_tensor.device) 
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # *** [CORRECTED] ***
        # Apply special transforms to 'Moderate' (1) and 'Severe' (2)
        minority_classes_list = [1, 2] 
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler (Original code kept for reference, but sampler=None is used below)
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        # *** [CORRECTED] ***
        # Using shuffle=True instead of sampler
        train_cls_loader = DataLoader(
            train_cls_dataset, 
            batch_size=BATCH_SIZE, 
            sampler=None,
            shuffle=True, 
            num_workers=2, 
            pin_memory=True
        )
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler disabled, Shuffle enabled). ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        
        # PLACEHOLDER MODELS (since external weights/complex definitions are unavailable)
        class CoAnNetFeatureExtractor(nn.Module):
            # CoAnNet is complex; we use a simpler block with the expected output dimension
            def __init__(self):
                super().__init__()
                self.conv = nn.Conv2d(3, 512, kernel_size=3, padding=1)
                self.pool = nn.AdaptiveAvgPool2d((1, 1))
                self.identity = nn.Identity()
            def forward(self, x):
                x = F.relu(self.conv(x))
                return self.identity(self.pool(x).flatten(1))

        class SimCLRResNet50(nn.Module):
            # SimCLR is ResNet50 with special weights; we use a standard pre-trained one
            def __init__(self):
                super().__init__()
                self.base = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
                self.base.fc = nn.Identity()
            def forward(self, x):
                return self.base(x)


        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name: str, n_classes=3, n_shape_features=5):
                super().__init__()
                
                cnn_feature_dim = 0
                
                if cnn_name == 'resnet34':
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'resnet50' or cnn_name == 'ResNet50_GAP':
                    # ResNet50 and ResNet50_GAP are the same in feature extraction layer, GAP is handled by AdaptiveAvgPool2d.
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                elif cnn_name == 'CoAnNet':
                    self.cnn_base = CoAnNetFeatureExtractor()
                    cnn_feature_dim = 512 # Based on placeholder
                elif cnn_name == 'SimCLR_ResNet50':
                    self.cnn_base = SimCLRResNet50()
                    cnn_feature_dim = 2048 # ResNet50 output
                else:
                    raise ValueError(f"Unknown CNN name: {cnn_name}")

                # Fusion layers
                self.fusion_dropout = nn.Dropout(0.5)
                self.fusion_fc1 = nn.Linear(cnn_feature_dim + n_shape_features, 512)
                self.fusion_bn1 = nn.BatchNorm1d(512)
                self.fusion_fc2 = nn.Linear(512, 128)
                self.fusion_bn2 = nn.BatchNorm1d(128)
                self.fusion_output = nn.Linear(128, n_classes)

            def forward(self, masked_image, shape_features):
                cnn_features = self.cnn_base(masked_image)
                
                # Ensure cnn_features is flat (handles ResNet vs. DenseNet)
                cnn_features = cnn_features.view(cnn_features.size(0), -1) 
                
                # Combine features
                combined_features = torch.cat([cnn_features, shape_features], dim=1)
                
                # Fusion layers
                x = self.fusion_dropout(F.relu(self.fusion_bn1(self.fusion_fc1(combined_features))))
                x = self.fusion_dropout(F.relu(self.fusion_bn2(self.fusion_fc2(x))))
                x = self.fusion_output(x)
                return x

        # --- 7.5 TRAINING AND EVALUATION PIPELINE ---
        
        CLASSIFICATION_BATCH_SIZE = BATCH_SIZE # Use the same batch size
        CLASSIFICATION_EPOCHS = NUM_EPOCHS
        LEARNING_RATE = 1e-4
        
        # Define the models to train
        MODELS_TO_TRAIN = [
            'resnet34', 
            'resnet50', 
            'densenet121', 
            'ResNet50_GAP', 
            'CoAnNet', 
            'SimCLR_ResNet50'
        ]

        # Store results for final comparison
        results = {}

        for model_name in MODELS_TO_TRAIN:
            print(f"\n--- Training Hybrid Model with: {model_name} ---")
            
            # 1. Initialize Model
            model = HybridClassifier(cnn_name=model_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            
            # 2. Loss and Optimizer
            # *** [CORRECTED] ***
            # Using gamma=2.0
            criterion_cls = FocalLoss(alpha=class_weights, gamma=2.0).to(device)
            optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)
            scheduler = ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.1, verbose=True)
            
            best_val_loss = float('inf')
            best_model_path = f"best_model_{model_name}.pth"
            
            # 3. Training Loop
            for epoch in range(CLASSIFICATION_EPOCHS):
                model.train()
                train_loss = 0.0
                train_loop = tqdm(train_cls_loader, desc=f"Epoch {epoch+1}/{CLASSIFICATION_EPOCHS} [Train]")
                for masked_images, shape_feats, labels in train_loop:
                    masked_images, shape_feats, labels = masked_images.to(device), shape_feats.to(device), labels.to(device)
                    
                    optimizer.zero_grad()
                    outputs = model(masked_images, shape_feats)
                    loss = criterion_cls(outputs, labels)
                    loss.backward()
                    optimizer.step()
                    train_loss += loss.item() * masked_images.size(0)
                    train_loop.set_postfix(loss=loss.item())
                
                # 4. Validation Loop
                model.eval()
                val_loss = 0.0
                all_val_preds, all_val_labels = [], []
                val_loop = tqdm(val_cls_loader, desc=f"Epoch {epoch+1}/{CLASSIFICATION_EPOCHS} [Val]")
                with torch.no_grad():
                    for masked_images, shape_feats, labels in val_loop:
                        masked_images, shape_feats, labels = masked_images.to(device), shape_feats.to(device), labels.to(device)
                        outputs = model(masked_images, shape_feats)
                        loss = criterion_cls(outputs, labels)
                        val_loss += loss.item() * masked_images.size(0)
                        
                        all_val_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
                        all_val_labels.extend(labels.cpu().numpy())
                        val_loop.set_postfix(loss=loss.item())
                
                avg_train_loss = train_loss / len(train_cls_loader.dataset)
                avg_val_loss = val_loss / len(val_cls_loader.dataset)
                scheduler.step(avg_val_loss)
                
                print(f"Epoch {epoch+1}: Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")
                
                if avg_val_loss < best_val_loss:
                    print(f"Validation loss improved ({best_val_loss:.4f} -> {avg_val_loss:.4f}). Saving model...")
                    best_val_loss = avg_val_loss
                    torch.save(model.state_dict(), best_model_path)
            
            # 5. Test Evaluation
            print(f"\n--- Evaluating Best Model ({model_name}) on Test Set ---")
            # Load the best performing model
            try:
                model.load_state_dict(torch.load(best_model_path))
            except Exception as e:
                print(f"Warning: Could not load best model at {best_model_path}. Using last epoch model. Error: {e}")
                
            model.eval()
            all_test_preds, all_test_labels = [], []
            with torch.no_grad():
                test_loop = tqdm(test_cls_loader, desc="Testing")
                for masked_images, shape_feats, labels in test_loop:
                    masked_images, shape_feats, labels = masked_images.to(device), shape_feats.to(device), labels.to(device)
                    outputs = model(masked_images, shape_feats)
                    all_test_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
                    all_test_labels.extend(labels.cpu().numpy())
            
            # Store results
            report = classification_report(all_test_labels, all_test_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            results[model_name] = {
                'report': report,
                'preds': np.array(all_test_preds),
                'labels': np.array(all_test_labels)
            }
            
            print(f"\nTest Report for {model_name}:")
            print(classification_report(all_test_labels, all_test_preds, target_names=CLASS_NAMES, zero_division=0))
            
            # Clean up saved model file
            if os.path.exists(best_model_path):
                os.remove(best_model_path)


        # --- 7.6 FINAL COMBINED REPORT (Best Per-Class Recall) ---
        print("\n\n--- Generating Final Combined Report (Best Per-Class Recall) ---")
        
        if not results:
            raise ValueError("No models were successfully trained or evaluated.")

        all_labels = results[MODELS_TO_TRAIN[0]]['labels'] # Get labels from the first result
        final_preds = np.zeros_like(all_labels)
        
        comparison_data = []
        
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                recall = res['report'][class_name]['recall']
                row[model_name] = f"{recall:.3f}"
                if recall > best_recall:
                    best_recall = recall
                    best_model = model_name
            
            row['Best Model'] = best_model
            row['Best Recall'] = f"{best_recall:.3f}"
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            class_indices = np.where(all_labels == i)[0]
            if best_model:
                 final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print("\n--- PER-CLASS RECALL COMPARISON ---")
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, recall_score
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50 # Reduced for demonstration, keep as 50 if needed
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Note: This is a standard normalization used for many CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))
        
# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None and not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as test data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            # Adjust alpha indexing for cross_entropy output
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    # Simple mean of sigmoid outputs for binary mask ensemble
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3], dim=0), dim=0)) 
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        # Take only the largest region's properties
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props)
                        if not props_df.empty:
                            # Find the largest area region
                            max_area_idx = props_df['area'].idxmax()
                            props_df = props_df.loc[[max_area_idx]]
                        else:
                             props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                    except Exception as e_prop:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3), A.GaussNoise(p=0.2), A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), 
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.8), A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3), A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                pred_mask = self.masks[idx].to(img_tensor.device)
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Assuming 'Severe' (label 2) is the primary minority class
        minority_classes_list = [2] 
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler for class balancing
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=sampler, num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler. ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name: str, n_classes=3, n_shape_features=5):
                super().__init__()
                
                # --- CNN BASE MODEL LOADING AND FEATURE DIM DETERMINATION ---
                if 'resnet' in cnn_name:
                    if cnn_name == 'resnet34':
                        self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                        cnn_feature_dim = self.cnn_base.fc.in_features
                        self.cnn_base.fc = nn.Identity()
                    elif cnn_name == 'resnet50' or cnn_name == 'simclr_resnet50':
                        # Assuming 'simclr_resnet50' uses standard weights if custom ones aren't loaded
                        self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                        cnn_feature_dim = self.cnn_base.fc.in_features
                        self.cnn_base.fc = nn.Identity()
                    elif cnn_name == 'resnet50_gap':
                        # ResNet50 with explicit Global Average Pooling before the identity mapping
                        self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                        cnn_feature_dim = self.cnn_base.fc.in_features
                        self.cnn_base.fc = nn.Identity() # Will manually pool in forward
                
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                    
                elif cnn_name == 'convnext_tiny':
                    self.cnn_base = models.convnext_tiny(weights=models.ConvNeXt_Tiny_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier[2].in_features
                    self.cnn_base.classifier[2] = nn.Identity()

                elif cnn_name == 'coanet':
                    # Placeholder: Assuming CoANet structure is similar to a standard ResNet50 for feature dim
                    print("âš ï¸ WARNING: 'coanet' is a placeholder. Assuming ResNet50 feature size for fusion.")
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                
                else:
                    raise ValueError(f"Unknown CNN name: {cnn_name}")
                    
                print(f"Loaded {cnn_name} encoder (Dim: {cnn_feature_dim})")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                # --- CNN Feature Extraction ---
                if 'resnet' in self.cnn_base.__class__.__name__.lower() or 'dense' in self.cnn_base.__class__.__name__.lower():
                    cnn_features = self.cnn_base(image)
                elif 'convnext' in self.cnn_base.__class__.__name__.lower():
                    # ConvNeXt processing until the classifier's final layer
                    x = self.cnn_base.features(image)
                    x = self.cnn_base.avgpool(x)
                    cnn_features = self.cnn_base.classifier[0](x.flatten(1)) # Layer 0 is the Linear layer before the final one in some ConvNeXt configs
                    cnn_features = self.cnn_base.classifier[1](cnn_features) # Layer 1 is the final Linear layer in some configs, using the defined output dim above
                    # Revert to simple identity removal for the provided structure assumption for now
                    cnn_features = self.cnn_base.forward_features(image) # If we remove the final layers, we need to get features before them
                    cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1)
                    
                else: # Default/Placeholder assumption
                     cnn_features = self.cnn_base(image)


                # Ensure 1D output after CNN for standard ResNet/DenseNet
                if cnn_features.ndim > 2:
                    if 'densenet' in self.cnn_base.__class__.__name__.lower():
                         # DenseNet with Identity replaces classifier, output is already 1D/flattened if weights were ImageNet
                         pass
                    elif 'convnext' in self.cnn_base.__class__.__name__.lower():
                         # ConvNeXt output pooling is handled above or needs careful adaptation
                         pass
                    elif 'resnet' in self.cnn_base.__class__.__name__.lower() and cnn_name == 'resnet50_gap':
                        cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1)
                    else: # For standard ResNets where fc is removed
                        cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1)
                        
                # --- Tabular Feature Processing ---
                tabular_features = self.tabular_mlp(shape_features)
                
                # --- Fusion ---
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output

        # --- 7.5 TRAIN AND EVALUATE FUNCTION PER MODEL ---
        def train_and_evaluate_single_model(cnn_name: str, train_loader, val_loader, test_loader, class_weights, n_shape_features, device, num_epochs=NUM_EPOCHS, patience=10):
            """Trains and evaluates a single hybrid model."""
            print(f"\n=============================================")
            print(f"ğŸš€ Starting Training for Model: {cnn_name}")
            print(f"=============================================")
            
            cls_model = HybridClassifier(cnn_name=cnn_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            # Use a slightly different Focal Loss setting for potentially harder task/more classes
            criterion = FocalLoss(alpha=class_weights, gamma=3.0) 
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5)
            scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=False)
            
            epochs_no_improve = 0; best_val_accuracy = 0.0
            model_save_path = f'best_classifier_model_{cnn_name}.pth'
            
            for epoch in range(num_epochs):
                cls_model.train()
                running_loss = 0.0
                for images, shapes, labels in train_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    optimizer.zero_grad()
                    outputs = cls_model(images, shapes)
                    loss = criterion(outputs, labels); loss.backward(); optimizer.step()
                    running_loss += loss.item()
                
                cls_model.eval()
                val_corrects = 0
                with torch.no_grad():
                    for images, shapes, labels in val_loader:
                        images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                        outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                        val_corrects += torch.sum(preds == labels.data)

                val_accuracy = val_corrects.double() / len(val_loader.dataset)
                scheduler.step(val_accuracy)

                if val_accuracy > best_val_accuracy:
                    best_val_accuracy = val_accuracy
                    torch.save(cls_model.state_dict(), model_save_path)
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    
                if epochs_no_improve >= patience:
                    print(f"â¹ï¸ Early stopping triggered for {cnn_name} at epoch {epoch+1}.")
                    break
                    
                # Optional: Print epoch status if needed, keeping clean for now
                # print(f"Epoch {epoch+1}/{num_epochs}, Val Acc: {val_accuracy:.4f}, LR: {optimizer.param_groups[0]['lr']:.2e}")

            print(f"Final Best Validation Accuracy for {cnn_name}: {best_val_accuracy:.4f}")

            # Final Test Evaluation
            cls_model.load_state_dict(torch.load(model_save_path))
            cls_model.eval()
            all_labels, all_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in test_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                    all_labels.extend(labels.cpu().numpy()); all_preds.extend(preds.cpu().numpy())
            
            report = classification_report(all_labels, all_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            return np.array(all_preds), report

        # --- 7.6 MASTER EXECUTION AND COMPARISON ---
        n_shape_features = train_features.shape[1]
        
        # *** MODIFIED MODEL LIST ***
        model_list = ['resnet34', 'resnet50', 'densenet121', 'resnet50_gap', 'simclr_resnet50', 'convnext_tiny', 'coanet']
        results = {} # Store results: {'model_name': {'preds': arr, 'report': dict}, ...}

        for model_name in model_list:
            preds, report = train_and_evaluate_single_model(
                cnn_name=model_name,
                train_loader=train_cls_loader, val_loader=val_cls_loader, test_loader=test_cls_loader,
                class_weights=class_weights, n_shape_features=n_shape_features, device=device
            )
            results[model_name] = {'preds': preds, 'report': report}

        # --- 7.7 FINAL ENSEMBLE SELECTION AND REPORT ---
        print("\n=============================================")
        print("ğŸ† FINAL RESULT SELECTION (PER-CLASS BEST RECALL)")
        print("=============================================")

        # 1. Get ground truth labels (Ensure all loaders provide the same labels)
        all_labels_list = []
        for _, _, labels in test_cls_loader:
            all_labels_list.extend(labels.cpu().numpy())
        all_labels = np.array(all_labels_list)
        
        final_preds = np.zeros_like(all_labels)
        
        best_model_per_class = {}
        
        # 2. Determine the best model for each class based on 'recall' (accuracy per class)
        comparison_data = []
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                # Check if the class name is a key in the report
                if class_name in res['report']:
                    recall = res['report'][class_name]['recall']
                    row[model_name] = f'{recall:.4f}'
                    
                    if recall > best_recall:
                        best_recall = recall
                        best_model = model_name
                else:
                    row[model_name] = 'N/A'
                    
            best_model_per_class[i] = best_model
            row['Best Model'] = best_model
            row['Best Recall'] = f'{best_recall:.4f}' if best_recall != -1.0 else 'N/A'
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            if best_model is not None:
                # Find indices where the ground truth label matches the current class
                class_indices = np.where(all_labels == i)[0]
                # Fill the final prediction array for these instances with the best model's predictions
                final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, recall_score
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# Suppress minor warnings for clean output
warnings.filterwarnings("ignore", category=UserWarning)

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth' # UNet with ResNet50
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth' # TransUNetPP
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth' # AttentionUNet
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
CLASS_NAMES = ['Mild', 'Moderate', 'Severe']
NUM_CLASSES = len(CLASS_NAMES)
NUM_EPOCHS = 50 
BATCH_SIZE = 16

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_label_map = {1: 0, 2: 1, 3: 2} # Mild=0, Moderate=1, Severe=2
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_label'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory): return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # Calculate class weights for Focal Loss
    class_weights_array = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_array, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        # Note: This is a standard normalization used for many CNNs (e.g., ImageNet weights)
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS (Segmentation Models) ---
# (Using the user's provided segmentation model definitions)

class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None); layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))
        
# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None and not test_df.empty:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading segmentation model files (Check paths): {e}")
        models_loaded = False
else:
    print("Skipping segmentation model loading as test data is unavailable.")
    
# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean': return loss.mean()
        else: return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df.empty:
    print("Skipping classification stage due to errors or missing data.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating ensemble segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    # Ensemble predictions
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                # Ensure mask is 2D and binary
                if mask.ndim == 3: mask = mask[:, :, 0] 
                mask = (mask > 0).astype(np.uint8)
                
                # Check for empty mask
                if np.sum(mask) == 0:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                else:
                    try:
                        props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                        props_df = pd.DataFrame(props).head(1) # Take only the largest region if multiple exist
                    except:
                        props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                        
                props_df['circularity'] = (4 * np.pi * props_df['area']) / (props_df['perimeter'] ** 2 + 1e-6)
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3), A.GaussNoise(p=0.2), A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), 
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])
        minority_transform = A.Compose([ # More aggressive augmentation
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]), A.HorizontalFlip(p=0.8), A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3), A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]), ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                
                # Conditional augmentation
                if label in self.minority_classes:
                    transformed = self.minority_transform(image=image)
                else:
                    transformed = self.transform(image=image)

                img_tensor = transformed['image']
                
                # Apply mask to the image (Liver only)
                pred_mask = self.masks[idx].to(img_tensor.device)
                masked_image = img_tensor * pred_mask
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Assuming 'Severe' (label 2) is the primary minority class
        minority_classes_list = [2] 
        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features, 
            transform=train_transform, minority_classes=minority_classes_list, minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # WeightedRandomSampler for class balancing
        class_counts = train_df['class_label'].value_counts().to_dict()
        num_samples = len(train_df)
        class_weights_sampler = {i: num_samples / class_counts.get(i, 1) for i in range(NUM_CLASSES)}
        sample_weights = np.array([class_weights_sampler[t] for t in train_df['class_label'].values])
        sampler = torch.utils.data.WeightedRandomSampler(
            weights=torch.from_numpy(sample_weights).double(), num_samples=len(sample_weights), replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=sampler, num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler. ğŸ“Š")

        # --- 7.4 HYBRID MODEL ARCHITECTURE (Dynamic Encoder) ---
        class HybridClassifier(nn.Module):
            def __init__(self, cnn_name: str, n_classes=3, n_shape_features=5):
                super().__init__()
                
                if cnn_name == 'resnet34':
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'resnet50':
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features
                    self.cnn_base.fc = nn.Identity()
                elif cnn_name == 'densenet121':
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.classifier.in_features
                    self.cnn_base.classifier = nn.Identity()
                else:
                    raise ValueError(f"Unknown CNN name: {cnn_name}")
                    
                print(f"Loaded {cnn_name} encoder (Dim: {cnn_feature_dim})")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                # Ensure 1D output after CNN
                if cnn_features.ndim > 2:
                    cnn_features = F.adaptive_avg_pool2d(cnn_features, (1, 1)).flatten(1)
                    
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output

        # --- 7.5 TRAIN AND EVALUATE FUNCTION PER MODEL ---
        def train_and_evaluate_single_model(cnn_name: str, train_loader, val_loader, test_loader, class_weights, n_shape_features, device, num_epochs=NUM_EPOCHS, patience=10):
            """Trains and evaluates a single hybrid model."""
            print(f"\n=============================================")
            print(f"ğŸš€ Starting Training for Model: {cnn_name}")
            print(f"=============================================")
            
            cls_model = HybridClassifier(cnn_name=cnn_name, n_classes=NUM_CLASSES, n_shape_features=n_shape_features).to(device)
            criterion = FocalLoss(alpha=class_weights, gamma=3.0)
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5)
            scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=False)
            
            epochs_no_improve = 0; best_val_accuracy = 0.0
            model_save_path = f'best_classifier_model_{cnn_name}.pth'
            
            for epoch in range(num_epochs):
                cls_model.train()
                running_loss = 0.0
                for images, shapes, labels in train_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    optimizer.zero_grad()
                    outputs = cls_model(images, shapes)
                    loss = criterion(outputs, labels); loss.backward(); optimizer.step()
                    running_loss += loss.item()
                
                cls_model.eval()
                val_corrects = 0
                with torch.no_grad():
                    for images, shapes, labels in val_loader:
                        images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                        outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                        val_corrects += torch.sum(preds == labels.data)

                val_accuracy = val_corrects.double() / len(val_loader.dataset)
                scheduler.step(val_accuracy)

                if val_accuracy > best_val_accuracy:
                    best_val_accuracy = val_accuracy
                    torch.save(cls_model.state_dict(), model_save_path)
                    epochs_no_improve = 0
                else:
                    epochs_no_improve += 1
                    
                if epochs_no_improve >= patience:
                    print(f"â¹ï¸ Early stopping triggered for {cnn_name} at epoch {epoch+1}.")
                    break
                    
            print(f"Final Best Validation Accuracy for {cnn_name}: {best_val_accuracy:.4f}")

            # Final Test Evaluation
            cls_model.load_state_dict(torch.load(model_save_path))
            cls_model.eval()
            all_labels, all_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in test_loader:
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes); _, preds = torch.max(outputs, 1)
                    all_labels.extend(labels.cpu().numpy()); all_preds.extend(preds.cpu().numpy())
            
            report = classification_report(all_labels, all_preds, target_names=CLASS_NAMES, zero_division=0, output_dict=True)
            return np.array(all_preds), report

        # --- 7.6 MASTER EXECUTION AND COMPARISON ---
        n_shape_features = train_features.shape[1]
        model_list = ['resnet34', 'resnet50', 'densenet121']
        results = {} # Store results: {'resnet34': {'preds': arr, 'report': dict}, ...}

        for model_name in model_list:
            preds, report = train_and_evaluate_single_model(
                cnn_name=model_name,
                train_loader=train_cls_loader, val_loader=val_cls_loader, test_loader=test_cls_loader,
                class_weights=class_weights, n_shape_features=n_shape_features, device=device
            )
            results[model_name] = {'preds': preds, 'report': report}

        # --- 7.7 FINAL ENSEMBLE SELECTION AND REPORT ---
        print("\n=============================================")
        print("ğŸ† FINAL RESULT SELECTION (PER-CLASS BEST RECALL)")
        print("=============================================")

        # 1. Get ground truth labels
        all_labels = np.concatenate([labels.cpu().numpy() for _, _, labels in test_cls_loader])
        final_preds = np.zeros_like(all_labels)
        
        best_model_per_class = {}
        
        # 2. Determine the best model for each class based on 'recall' (accuracy per class)
        comparison_data = []
        for i, class_name in enumerate(CLASS_NAMES):
            best_recall = -1.0
            best_model = None
            
            row = {'Class': class_name}
            for model_name, res in results.items():
                recall = res['report'][class_name]['recall']
                row[model_name] = f'{recall:.4f}'
                
                if recall > best_recall:
                    best_recall = recall
                    best_model = model_name
                    
            best_model_per_class[i] = best_model
            row['Best Model'] = best_model
            row['Best Recall'] = f'{best_recall:.4f}'
            comparison_data.append(row)
            
            # 3. Create the final combined prediction array (Per-Class Best Model Selection)
            # Find indices where the ground truth label matches the current class
            class_indices = np.where(all_labels == i)[0]
            # Fill the final prediction array for these instances with the best model's predictions
            final_preds[class_indices] = results[best_model]['preds'][class_indices]

        # Print comparison table
        comparison_df = pd.DataFrame(comparison_data)
        print(comparison_df.to_markdown(index=False))
        
        # Print Final Combined Report
        print("\n--- FINAL CLASSIFICATION REPORT (Combined Best Recall) ---")
        print(classification_report(all_labels, final_preds, target_names=CLASS_NAMES, zero_division=0))

        # Confusion Matrix for the Combined Result
        cm_final = confusion_matrix(all_labels, final_preds, labels=range(NUM_CLASSES))
        disp_final = ConfusionMatrixDisplay(confusion_matrix=cm_final, display_labels=CLASS_NAMES)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp_final.plot(ax=ax, cmap=plt.cm.Greens)
        plt.title("Confusion Matrix (Best Per-Class Selection)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred in the Classification Pipeline: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# ### <<< à¦†à¦ªà¦¨à¦¾à¦° WSP à¦ªà§à¦°à¦¿-à¦Ÿà§à¦°à§‡à¦‡à¦¨à¦¡ à¦®à¦¡à§‡à¦²à§‡à¦° à¦ªà¦¾à¦¥ à¦à¦–à¦¾à¦¨à§‡ à¦¦à¦¿à¦¨ >>> ###
WSP_PRETRAIN_CKPT_PATH = '/kaggle/input/my-wsp-model/best_wsp_encoder.ckpt' 
WSP_ENCODER_NAME = 'resnet50' # (à¦¯à§‡à¦®à¦¨: 'resnet50', 'densenet121')

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2} # Mapped: Mild->0, Moderate->1, Severe->2
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS (Simplified Definitions for Brevity) ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (Definition kept but unused) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # 70/30 à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿ
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # 'Moderate' (1) à¦à¦¬à¦‚ 'Severe' (2) à¦‰à¦­à§Ÿà¦•à§‡à¦‡ à¦¸à¦‚à¦–à§à¦¯à¦¾à¦²à¦˜à§ à¦§à¦°à¦¾ à¦¹à¦²à§‹
        minority_classes_list = [1, 2] 
        print(f"Applying strong augmentation to classes: {minority_classes_list}")
        
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + Label Smoothing Loss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (WSP à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° à¦¸à¦¹) ---
        class WSPHybridClassifier(nn.Module):
            def __init__(self, encoder_name, n_classes=3, n_shape_features=5):
                super().__init__()
                print(f"Initializing Hybrid Classifier with base: {encoder_name}")
                
                try:
                    self.cnn_base = timm.create_model(encoder_name, pretrained=False)
                    cnn_feature_dim = self.cnn_base.num_features
                    self.cnn_base.reset_classifier(0) 
                except Exception as e:
                    print(f"!! Error creating model '{encoder_name}' with timm: {e}")
                    print("Falling back to resnet50. PLEASE CHECK YOUR 'WSP_ENCODER_NAME'.")
                    self.cnn_base = timm.create_model('resnet50', pretrained=False)
                    cnn_feature_dim = self.cnn_base.num_features
                    self.cnn_base.reset_classifier(0)

                print(f"CNN base created. Feature dimension: {cnn_feature_dim}")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 512), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.3),
                    nn.Linear(256, n_classes)
                )
                
            def load_wsp_weights(self, ckpt_path):
                print(f"--- Loading WSP pre-trained weights from: {ckpt_path} ---")
                try:
                    ckpt = torch.load(ckpt_path, map_location=device)
                    
                    if 'state_dict' in ckpt:
                        wsp_state_dict = ckpt['state_dict']
                    else:
                        wsp_state_dict = ckpt
                        
                    encoder_state_dict = {}
                    prefix = 'net.' 
                    
                    for k, v in wsp_state_dict.items():
                        if k.startswith(prefix):
                            new_key = k[len(prefix):]
                            encoder_state_dict[new_key] = v
                            
                    if not encoder_state_dict:
                        print("!! WARNING: No weights found with prefix 'net.'.")
                        print("Attempting to load the state_dict directly (assuming it's just the encoder).")
                        encoder_state_dict = wsp_state_dict 

                    incompatible_keys = self.cnn_base.load_state_dict(encoder_state_dict, strict=False)
                    print(f"WSP weights loaded successfully.")
                    print(f"  > Mismatched/Missing keys (should be OK if just classifier): {incompatible_keys}")
                
                except Exception as e:
                    print(f"!! ERROR: Failed to load WSP weights from {ckpt_path}. {e}")
                    print("!! MODEL WILL TRAIN FROM SCRATCH (RANDOM WEIGHTS).")
            
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image) 
                tabular_features = self.tabular_mlp(shape_features) 
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        cls_model = WSPHybridClassifier(
            encoder_name=WSP_ENCODER_NAME,
            n_classes=3,
            n_shape_features=train_features.shape[1]
        )
        
        if os.path.exists(WSP_PRETRAIN_CKPT_PATH):
            cls_model.load_wsp_weights(WSP_PRETRAIN_CKPT_PATH)
        else:
            print(f"!! WARNING: WSP Checkpoint path not found: {WSP_PRETRAIN_CKPT_PATH}")
            print("!! Training from scratch (random initialization).")
            
        cls_model = cls_model.to(device)

        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à¦ªà§à¦°à¦§à¦¾à¦¨): CrossEntropyLoss with Label Smoothing (0.1) >>> ###
        criterion = nn.CrossEntropyLoss(label_smoothing=0.1).to(device)
        print("Using CrossEntropyLoss with Label Smoothing (0.1).")
        
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-4) 
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print(f"\n--- Starting Hybrid Model (WSP {WSP_ENCODER_NAME} Base) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies (Recall) ---")
        # Note: Recall is used here as it is the class-specific measure of accuracy for the minority class.
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title(f"Confusion Matrix for Hybrid Model (WSP {WSP_ENCODER_NAME} Base)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# ### <<< à¦†à¦ªà¦¨à¦¾à¦° WSP à¦ªà§à¦°à¦¿-à¦Ÿà§à¦°à§‡à¦‡à¦¨à¦¡ à¦®à¦¡à§‡à¦²à§‡à¦° à¦ªà¦¾à¦¥ à¦à¦–à¦¾à¦¨à§‡ à¦¦à¦¿à¦¨ >>> ###
WSP_PRETRAIN_CKPT_PATH = '/kaggle/input/my-wsp-model/best_wsp_encoder.ckpt' 
WSP_ENCODER_NAME = 'resnet50' # (à¦¯à§‡à¦®à¦¨: 'resnet50', 'densenet121')

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2} # Mapped: Mild->0, Moderate->1, Severe->2
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS (Definitions omitted for brevity, assumed correct) ---
class ConvBlock(nn.Module):
     def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
     def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
     def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
     def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
     def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
     def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
     def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
     def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
     def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
     def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
     def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
     def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS (omitted for brevity, assumed correct) ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --------------------------------------------------------------------------------
# ### <<< CHANGE 1: Focal Loss Definition with High Gamma (5.0) >>> ###
# --------------------------------------------------------------------------------
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=5.0, reduction='mean'): # Gamma increased to 5.0
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
        if isinstance(self.alpha, list) or isinstance(self.alpha, np.ndarray):
            self.alpha = torch.tensor(self.alpha, dtype=torch.float)

    def forward(self, input, target):
        # 1. Compute Cross-Entropy Loss
        ce_loss = F.cross_entropy(input, target, reduction='none')
        
        # 2. Compute pt (probability of correct class)
        pt = torch.exp(-ce_loss)
        
        # 3. Compute (1 - pt)^gamma * CE
        loss = ((1 - pt) ** self.gamma) * ce_loss
        
        # 4. Apply Alpha (Class Weights)
        if self.alpha is not None:
            if self.alpha.device != target.device:
                 self.alpha = self.alpha.to(target.device)
            
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
            
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions & 7.2 Shape features (omitted for brevity, assumed correct) ---
        print("\nGenerating segmentation predictions...")
        # (Segmentation and Feature Extraction code remains unchanged)
        
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets and Loaders (omitted for brevity, assumed correct) ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # 70/30 à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿ
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [1, 2] 
        print(f"Applying strong augmentation to classes: {minority_classes_list}")
        
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 Sampler Setup ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild: {class_counts[0]}, 1=Moderate: {class_counts[1]}, 2=Severe: {class_counts[2]})")

        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (WeightedRandomSampler is ON). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (omitted for brevity, assumed correct) ---
        class WSPHybridClassifier(nn.Module):
             def __init__(self, encoder_name, n_classes=3, n_shape_features=5):
                 super().__init__()
                 print(f"Initializing Hybrid Classifier with base: {encoder_name}")
                 try:
                     self.cnn_base = timm.create_model(encoder_name, pretrained=False)
                     cnn_feature_dim = self.cnn_base.num_features
                     self.cnn_base.reset_classifier(0) 
                 except Exception as e:
                     print(f"!! Error creating model '{encoder_name}' with timm: {e}")
                     self.cnn_base = timm.create_model('resnet50', pretrained=False)
                     cnn_feature_dim = self.cnn_base.num_features
                     self.cnn_base.reset_classifier(0)

                 self.tabular_mlp = nn.Sequential(
                     nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                     nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                 )
                 
                 self.fusion_classifier = nn.Sequential(
                     nn.Linear(cnn_feature_dim + 32, 512), nn.ReLU(), nn.Dropout(0.4),
                     nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.3),
                     nn.Linear(256, n_classes)
                 )
                 
             def load_wsp_weights(self, ckpt_path):
                 print(f"--- Loading WSP pre-trained weights from: {ckpt_path} ---")
                 try:
                     ckpt = torch.load(ckpt_path, map_location=device)
                     wsp_state_dict = ckpt.get('state_dict', ckpt)
                     encoder_state_dict = {k[len('net.'):]: v for k, v in wsp_state_dict.items() if k.startswith('net.')}
                     if not encoder_state_dict:
                         encoder_state_dict = wsp_state_dict
                     incompatible_keys = self.cnn_base.load_state_dict(encoder_state_dict, strict=False)
                     print(f"WSP weights loaded successfully.")
                     print(f"  > Mismatched/Missing keys: {incompatible_keys}")
                 except Exception as e:
                     print(f"!! ERROR: Failed to load WSP weights from {ckpt_path}. {e}")
                     print("!! MODEL WILL TRAIN FROM SCRATCH (RANDOM WEIGHTS).")
             
             def forward(self, image, shape_features):
                 cnn_features = self.cnn_base(image) 
                 tabular_features = self.tabular_mlp(shape_features) 
                 combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                 return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        cls_model = WSPHybridClassifier(
            encoder_name=WSP_ENCODER_NAME,
            n_classes=3,
            n_shape_features=train_features.shape[1]
        )
        
        if os.path.exists(WSP_PRETRAIN_CKPT_PATH):
            cls_model.load_wsp_weights(WSP_PRETRAIN_CKPT_PATH)
        else:
            print(f"!! WARNING: WSP Checkpoint path not found: {WSP_PRETRAIN_CKPT_PATH}")
            print("!! Training from scratch (random initialization).")
            
        cls_model = cls_model.to(device)

        # ----------------------------------------------------------------------------------
        # ### <<< CHANGE 2: Criterion setup with Weighted Focal Loss >>> ###
        # ----------------------------------------------------------------------------------
        # Calculate Inverse Frequency Weights: [Mild (0): 90, Moderate (1): 11, Severe (2): 65]
        # We assume the order of classes is 0, 1, 2.
        class_counts = np.array([class_counts[0], class_counts[1], class_counts[2]])
        
        # Calculate raw inverse frequency weights
        raw_weights = 1.0 / class_counts
        # Normalize the weights so they sum to 1 (better stability)
        alpha_weights = torch.tensor(raw_weights / raw_weights.sum(), dtype=torch.float).to(device)

        print(f"\nInverse Frequency Weights (Normalized) for Focal Loss:")
        print(f"  Mild (0): {alpha_weights[0].item():.4f}, Moderate (1): {alpha_weights[1].item():.4f}, Severe (2): {alpha_weights[2].item():.4f}")
        
        # Use the FocalLoss class defined above with alpha (weights) and high gamma (5.0)
        criterion = FocalLoss(alpha=alpha_weights, gamma=5.0).to(device)
        print(f"Using Weighted Focal Loss (gamma=5.0) to focus on hard and minority samples.")
        
        # --- Optimizer and Scheduler (remain unchanged) ---
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-4) 
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print(f"\n--- Starting Hybrid Model (WSP {WSP_ENCODER_NAME} Base) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # (Training loop remains unchanged)
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            # (Validation loop remains unchanged)
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation (remains unchanged) ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies (Recall) ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title(f"Confusion Matrix for Hybrid Model (WSP {WSP_ENCODER_NAME} Base)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import DataLoader, Dataset
import numpy as np

# --- 1. à¦¡à¦¿à¦­à¦¾à¦‡à¦¸ à¦à¦¬à¦‚ à¦•à¦¨à¦«à¦¿à¦—à¦¾à¦°à§‡à¦¶à¦¨ à¦¸à§‡à¦Ÿà¦†à¦ª ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# à¦†à¦ªà¦¨à¦¾à¦° à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿ à¦…à¦¨à§à¦¯à¦¾à§Ÿà§€ à¦•à§à¦²à¦¾à¦¸ à¦•à¦¾à¦‰à¦¨à§à¦Ÿ (Mild: 90, Moderate: 11, Severe: 65)
CLASS_COUNTS = np.array([90, 11, 65]) 
NUM_CLASSES = 3
NUM_EPOCHS = 50
LEARNING_RATE = 1e-4 
PATIENCE = 3 # ReduceLROnPlateau-à¦à¦° à¦œà¦¨à§à¦¯

# --- 2. Weighted Focal Loss à¦•à§à¦²à¦¾à¦¸ ---

# Gamma=5.0 à¦à¦° à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à§‡ Gamma=3.0 à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦²à§‹ à¦…à¦¸à§à¦¥à¦¿à¦°à¦¤à¦¾ à¦•à¦®à¦¾à¦¤à§‡
class FocalLoss(nn.Module):
    def __init__(self, alpha, gamma=3.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha.to(device)
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = nn.CrossEntropyLoss(weight=None, reduction='none')(inputs, targets)
        pt = torch.exp(-ce_loss)

        # Class weight application
        alpha_t = self.alpha.gather(0, targets.data.view(-1)).view(ce_loss.size())
        
        # Focusing factor application
        modulating_factor = (1 - pt) ** self.gamma
        
        focal_loss = alpha_t * modulating_factor * ce_loss

        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# --- 3. à¦¹à¦¾à¦‡à¦¬à§à¦°à¦¿à¦¡ à¦®à¦¡à§‡à¦² à¦†à¦°à§à¦•à¦¿à¦Ÿà§‡à¦•à¦šà¦¾à¦° (SimCLR-ResNet + Tabular MLP Fusion) ---

# à¦à¦Ÿà¦¿ à¦†à¦ªà¦¨à¦¾à¦° à¦ªà§à¦°à¦¸à§à¦¤à¦¾à¦¬à¦¿à¦¤ à¦®à¦¡à§‡à¦²à§‡à¦° à¦à¦•à¦Ÿà¦¿ à¦‰à¦¦à¦¾à¦¹à¦°à¦£à¥¤ à¦†à¦ªà¦¨à¦¾à¦° à¦ªà§à¦°à¦¯à¦¼à§‹à¦œà¦¨ à¦…à¦¨à§à¦¯à¦¾à¦¯à¦¼à§€ à¦à¦Ÿà¦¿ à¦¸à¦‚à¦¶à§‹à¦§à¦¨ à¦•à¦°à¦¤à§‡ à¦ªà¦¾à¦°à§‡à¦¨à¥¤
# à¦®à¦¨à§‡ à¦°à¦¾à¦–à¦¬à§‡à¦¨, ResNet50-à¦•à§‡ à¦…à¦¬à¦¶à§à¦¯à¦‡ ImageNet à¦¬à¦¾ SimCLR-à¦ Pretrained à¦¹à¦¤à§‡ à¦¹à¦¬à§‡à¥¤
class HybridClassifier(nn.Module):
    def __init__(self, num_classes=3, tabular_dim=10):
        super(HybridClassifier, self).__init__()
        
        # 3.1. Image Branch (CNN - ResNet50)
        # à¦à¦–à¦¾à¦¨à§‡ Pretrained ResNet50 à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¤à§‡ à¦ªà¦¾à¦°à§‡à¦¨
        self.image_model = torch.hub.load('pytorch/vision:v0.10.0', 'resnet50', pretrained=True)
        # à¦«à¦¾à¦‡à¦¨à¦¾à¦² à¦«à§à¦²à§à¦²à¦¿ à¦•à¦¾à¦¨à§‡à¦•à§à¦Ÿà§‡à¦¡ à¦²à§‡à§Ÿà¦¾à¦°à¦Ÿà¦¿ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾ à¦¹à¦²à§‹
        num_ftrs = self.image_model.fc.in_features
        self.image_model.fc = nn.Identity() 

        # 3.2. Tabular Branch (MLP)
        self.tabular_mlp = nn.Sequential(
            nn.Linear(tabular_dim, 64), 
            nn.ReLU(),
            nn.Dropout(0.5) # à¦°à§‡à¦—à§à¦²à¦¾à¦°à¦¾à¦‡à¦œà§‡à¦¶à¦¨ à¦¬à¦¾à¦¡à¦¼à¦¾à¦¨à§‹ à¦¹à¦²à§‹
        )
        
        # 3.3. Fusion Head
        # ResNet50 output (num_ftrs) + Tabular output (64)
        fusion_dim = num_ftrs + 64 
        self.classifier_head = nn.Sequential(
            nn.Linear(fusion_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes)
        )

    def forward(self, image_data, tabular_data):
        # Image Features
        image_features = self.image_model(image_data) 
        
        # Tabular Features
        tabular_features = self.tabular_mlp(tabular_data)
        
        # Fusion and Classification
        fused_features = torch.cat((image_features, tabular_features), dim=1)
        output = self.classifier_head(fused_features)
        return output

# --- 4. à¦¡à¦¾à¦®à¦¿ à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿ (à¦†à¦ªà¦¨à¦¾à¦° à¦†à¦¸à¦² à¦¡à§‡à¦Ÿà¦¾ à¦¦à§à¦¬à¦¾à¦°à¦¾ à¦ªà§à¦°à¦¤à¦¿à¦¸à§à¦¥à¦¾à¦ªà¦¿à¦¤ à¦•à¦°à§à¦¨) ---

class DummyHybridDataset(Dataset):
    def __init__(self, image_size=(3, 224, 224), tabular_dim=10, count=166, class_counts=CLASS_COUNTS):
        # à¦à¦‡ à¦¡à§‡à¦Ÿà¦¾ à¦¤à§ˆà¦°à¦¿à¦° à¦²à¦œà¦¿à¦• à¦†à¦ªà¦¨à¦¾à¦° à¦†à¦¸à¦² à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿ à¦²à§‹à¦¡à¦¿à¦‚ à¦²à¦œà¦¿à¦• à¦¦à§à¦¬à¦¾à¦°à¦¾ à¦ªà§à¦°à¦¤à¦¿à¦¸à§à¦¥à¦¾à¦ªà¦¿à¦¤ à¦¹à¦¬à§‡
        self.images = torch.randn(count, *image_size) 
        self.tabular = torch.randn(count, tabular_dim)
        
        # à¦‡à¦®à¦¬à§à¦¯à¦¾à¦²à§‡à¦¨à§à¦¸à¦¡ à¦²à§‡à¦¬à§‡à¦² à¦¤à§ˆà¦°à¦¿ à¦•à¦°à¦¾ à¦¹à¦²à§‹
        labels = []
        for i, c in enumerate(class_counts):
            labels.extend([i] * c)
        self.labels = torch.tensor(labels[:count])

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.images[idx], self.tabular[idx], self.labels[idx]

# --- 5. à¦¡à§‡à¦Ÿà¦¾ à¦²à§‹à¦¡à¦¾à¦° à¦à¦¬à¦‚ à¦“à§Ÿà§‡à¦Ÿà¦¿à¦‚ ---

# WeightedRandomSampler à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦•à¦¾à¦°à¦£ Focal Loss à¦¨à¦¿à¦œà§‡à¦‡ à¦“à§Ÿà§‡à¦Ÿ à¦¸à¦¾à¦®à¦²à¦¾à§Ÿà¥¤
# à¦¤à¦¬à§‡, à¦†à¦ªà¦¨à¦¿ à¦¯à¦¦à¦¿ WeightedRandomSampler-à¦“ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à§‡à¦¨, à¦¤à¦¾à¦¹à¦²à§‡ Focal Loss-à¦à¦° alpha_t à¦“ gamma-à¦•à§‡ Tuner-à¦ à¦¸à§‡à¦Ÿ à¦•à¦°à¦¤à§‡ à¦¹à¦¬à§‡à¥¤
# à¦à¦‡ à¦‰à¦¦à¦¾à¦¹à¦°à¦£à§‡ à¦†à¦®à¦°à¦¾ à¦¶à§à¦§à§ Focal Loss-à¦à¦° à¦‰à¦ªà¦° à¦¨à¦¿à¦°à§à¦­à¦° à¦•à¦°à¦›à¦¿à¥¤
train_dataset = DummyHybridDataset()
val_dataset = DummyHybridDataset(count=50) # à¦¡à¦¾à¦®à¦¿ à¦­à§à¦¯à¦¾à¦²à¦¿à¦¡à§‡à¦¶à¦¨ à¦¸à§‡à¦Ÿ
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)


# --- 6. à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦«à¦¾à¦‚à¦¶à¦¨ à¦à¦¬à¦‚ à¦¶à¦¿à¦¡à¦¿à¦‰à¦²à¦¾à¦° à¦²à¦œà¦¿à¦• ---

# à¦“à§Ÿà§‡à¦Ÿ à¦—à¦£à¦¨à¦¾ (Focal Loss-à¦à¦° à¦œà¦¨à§à¦¯)
raw_weights = 1.0 / CLASS_COUNTS
alpha_weights = torch.tensor(raw_weights / raw_weights.sum(), dtype=torch.float)

# à¦®à¦¡à§‡à¦², à¦…à¦ªà¦Ÿà¦¿à¦®à¦¾à¦‡à¦œà¦¾à¦° à¦à¦¬à¦‚ à¦²à¦¸ à¦«à¦¾à¦‚à¦¶à¦¨ à¦‡à¦¨à¦¿à¦¶à¦¿à§Ÿà¦¾à¦²à¦¾à¦‡à¦œà§‡à¦¶à¦¨
model = HybridClassifier(num_classes=NUM_CLASSES, tabular_dim=10).to(device)
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5) # Weight Decay à¦¯à§‹à¦— à¦•à¦°à¦¾ à¦¹à¦²à§‹
criterion = FocalLoss(alpha=alpha_weights, gamma=3.0, reduction='mean').to(device)

# ReduceLROnPlateau à¦¶à¦¿à¦¡à¦¿à¦‰à¦²à¦¾à¦°
scheduler = ReduceLROnPlateau(
    optimizer, 
    mode='max',      
    factor=0.5,      
    patience=PATIENCE, 
    verbose=True,
    min_lr=1e-6
)

def train_and_validate(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs):
    best_val_acc = 0.0
    
    for epoch in range(1, num_epochs + 1):
        # --- Training Phase ---
        model.train()
        running_loss = 0.0
        
        for images, tabular, labels in train_loader:
            images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images, tabular)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * images.size(0)
            
        epoch_loss = running_loss / len(train_loader.dataset)
        
        # --- Validation Phase ---
        model.eval()
        corrects = 0
        total = 0
        
        with torch.no_grad():
            for images, tabular, labels in val_loader:
                images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                
                outputs = model(images, tabular)
                _, preds = torch.max(outputs, 1)
                
                total += labels.size(0)
                corrects += torch.sum(preds == labels.data)
                
        epoch_val_acc = corrects.double() / total
        
        print(f"Epoch {epoch}/{num_epochs} | Loss: {epoch_loss:.4f} | Val Acc: {epoch_val_acc:.4f}")

        # à¦¶à¦¿à¦¡à¦¿à¦‰à¦²à¦¾à¦° à¦†à¦ªà¦¡à§‡à¦Ÿ
        scheduler.step(epoch_val_acc) 

        # à¦¬à§‡à¦¸à§à¦Ÿ à¦®à¦¡à§‡à¦² à¦¸à§‡à¦­ à¦•à¦°à¦¾
        if epoch_val_acc > best_val_acc:
            best_val_acc = epoch_val_acc
            print(f"->âœ… New best model saved: {best_val_acc:.4f}")
            # torch.save(model.state_dict(), 'best_hybrid_model.pth')
        else:
            print(f"->âŒ No improvement.")


# à¦Ÿà§à¦°à§‡à¦¨ à¦¶à§à¦°à§ à¦•à¦°à§à¦¨
# train_and_validate(model, train_loader, val_loader, criterion, optimizer, scheduler, NUM_EPOCHS)
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# ### <<< à¦†à¦ªà¦¨à¦¾à¦° WSP à¦ªà§à¦°à¦¿-à¦Ÿà§à¦°à§‡à¦‡à¦¨à¦¡ à¦®à¦¡à§‡à¦²à§‡à¦° à¦ªà¦¾à¦¥ à¦à¦–à¦¾à¦¨à§‡ à¦¦à¦¿à¦¨ >>> ###
WSP_PRETRAIN_CKPT_PATH = '/kaggle/input/my-wsp-model/best_wsp_encoder.ckpt' 
WSP_ENCODER_NAME = 'resnet50' # (à¦¯à§‡à¦®à¦¨: 'resnet50', 'densenet121')

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦†à¦—à§‡à¦° à¦•à§‹à¦¡ à¦¥à§‡à¦•à§‡ à¦à¦–à¦¾à¦¨à§‡ à¦°à¦¾à¦–à¦¾ à¦¹à¦²à§‹, à¦•à¦¿à¦¨à§à¦¤à§ à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        # (à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # 70/30 à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿ
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # 'Moderate' (1) à¦à¦¬à¦‚ 'Severe' (2) à¦‰à¦­à§Ÿà¦•à§‡à¦‡ à¦¸à¦‚à¦–à§à¦¯à¦¾à¦²à¦˜à§ à¦§à¦°à¦¾ à¦¹à¦²à§‹
        minority_classes_list = [1, 2] 
        print(f"Applying strong augmentation to classes: {minority_classes_list}")
        
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à¦à¦‡ à¦“à§Ÿà§‡à¦Ÿà¦—à§à¦²à§‹ à¦†à¦° CrossEntropyLoss-à¦ à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦¤à¦¬à§‡ FocalLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦°à¦¾à¦–à¦¾ à¦¹à§Ÿà§‡à¦›à¦¿à¦²
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        # print(f"Alpha weights (for FocalLoss, if used): {alpha_weights}")

        # WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + Label Smoothing Loss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (WSP à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° à¦¸à¦¹) ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        class WSPHybridClassifier(nn.Module):
            def __init__(self, encoder_name, n_classes=3, n_shape_features=5):
                super().__init__()
                print(f"Initializing Hybrid Classifier with base: {encoder_name}")
                
                try:
                    self.cnn_base = timm.create_model(encoder_name, pretrained=False)
                    cnn_feature_dim = self.cnn_base.num_features
                    self.cnn_base.reset_classifier(0) 
                except Exception as e:
                    print(f"!! Error creating model '{encoder_name}' with timm: {e}")
                    print("Falling back to resnet50. PLEASE CHECK YOUR 'WSP_ENCODER_NAME'.")
                    self.cnn_base = timm.create_model('resnet50', pretrained=False)
                    cnn_feature_dim = self.cnn_base.num_features
                    self.cnn_base.reset_classifier(0)

                print(f"CNN base created. Feature dimension: {cnn_feature_dim}")

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 512), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.3),
                    nn.Linear(256, n_classes)
                )
                
            def load_wsp_weights(self, ckpt_path):
                print(f"--- Loading WSP pre-trained weights from: {ckpt_path} ---")
                try:
                    ckpt = torch.load(ckpt_path, map_location=device)
                    
                    if 'state_dict' in ckpt:
                        wsp_state_dict = ckpt['state_dict']
                    else:
                        wsp_state_dict = ckpt
                        
                    encoder_state_dict = {}
                    prefix = 'net.' 
                    
                    for k, v in wsp_state_dict.items():
                        if k.startswith(prefix):
                            new_key = k[len(prefix):]
                            encoder_state_dict[new_key] = v
                            
                    if not encoder_state_dict:
                        print("!! WARNING: No weights found with prefix 'net.'.")
                        print("Attempting to load the state_dict directly (assuming it's just the encoder).")
                        encoder_state_dict = wsp_state_dict 

                    incompatible_keys = self.cnn_base.load_state_dict(encoder_state_dict, strict=False)
                    print(f"WSP weights loaded successfully.")
                    print(f"  > Mismatched/Missing keys (should be OK if just classifier): {incompatible_keys}")
                
                except Exception as e:
                    print(f"!! ERROR: Failed to load WSP weights from {ckpt_path}. {e}")
                    print("!! MODEL WILL TRAIN FROM SCRATCH (RANDOM WEIGHTS).")
            
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image) 
                tabular_features = self.tabular_mlp(shape_features) 
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        cls_model = WSPHybridClassifier(
            encoder_name=WSP_ENCODER_NAME,
            n_classes=3,
            n_shape_features=train_features.shape[1]
        )
        
        if os.path.exists(WSP_PRETRAIN_CKPT_PATH):
            cls_model.load_wsp_weights(WSP_PRETRAIN_CKPT_PATH)
        else:
            print(f"!! WARNING: WSP Checkpoint path not found: {WSP_PRETRAIN_CKPT_PATH}")
            print("!! Training from scratch (random initialization).")
            
        cls_model = cls_model.to(device)

        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à¦ªà§à¦°à¦§à¦¾à¦¨): FocalLoss-à¦à¦° à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à§‡ Label Smoothing à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ >>> ###
        # à¦à¦Ÿà¦¿ 'Moderate' à¦•à§à¦²à¦¾à¦¸à§‡à¦° à¦®à¦¤à§‹ à¦…à¦¸à§à¦ªà¦·à§à¦Ÿ à¦•à§‡à¦¸à¦—à§à¦²à¦¿à¦° à¦œà¦¨à§à¦¯ à¦­à¦¾à¦²à§‹ à¦•à¦¾à¦œ à¦•à¦°à§‡à¥¤
        # WeightedRandomSampler à¦šà¦¾à¦²à§ à¦†à¦›à§‡, à¦¤à¦¾à¦‡ à¦à¦–à¦¾à¦¨à§‡ `weight` à¦†à¦°à§à¦—à§à¦®à§‡à¦¨à§à¦Ÿ `None` à¦¥à¦¾à¦•à¦¬à§‡à¥¤
        criterion = nn.CrossEntropyLoss(label_smoothing=0.1).to(device)
        print("Using CrossEntropyLoss with Label Smoothing (0.1).")
        
        # à¦…à¦ªà§à¦Ÿà¦¿à¦®à¦¾à¦‡à¦œà¦¾à¦° à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-4) 
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print(f"\n--- Starting Hybrid Model (WSP {WSP_ENCODER_NAME} Base) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # (à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            # (à¦­à§à¦¯à¦¾à¦²à¦¿à¦¡à§‡à¦¶à¦¨ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title(f"Confusion Matrix for Hybrid Model (WSP {WSP_ENCODER_NAME} Base)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm  # timm à¦à¦–à¦¨à¦“ à¦¦à¦°à¦•à¦¾à¦°à¦¿, à¦•à¦¾à¦°à¦£ à¦†à¦®à¦°à¦¾ à¦à¦° `create_model` à¦«à¦¾à¦‚à¦¶à¦¨ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¬

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à§§): à¦†à¦ªà¦¨à¦¾à¦° WSP à¦ªà§à¦°à¦¿-à¦Ÿà§à¦°à§‡à¦‡à¦¨à¦¡ à¦®à¦¡à§‡à¦²à§‡à¦° à¦ªà¦¾à¦¥ à¦à¦–à¦¾à¦¨à§‡ à¦¦à¦¿à¦¨ >>> ###
# à¦à¦Ÿà¦¿ à¦†à¦ªà¦¨à¦¾à¦° à¦ªà§à¦°à¦¥à¦® à¦•à§‹à¦¡ (wspContrastiveLearning.py) à¦¥à§‡à¦•à§‡ à¦ªà¦¾à¦“à§Ÿà¦¾ .ckpt à¦¬à¦¾ .pth à¦«à¦¾à¦‡à¦²
WSP_PRETRAIN_CKPT_PATH = '/kaggle/input/my-wsp-model/best_wsp_encoder.ckpt' 
# à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° à¦†à¦°à§à¦•à¦¿à¦Ÿà§‡à¦•à¦šà¦¾à¦°à§‡à¦° à¦¨à¦¾à¦® (à¦†à¦ªà¦¨à¦¾à¦° à¦ªà§à¦°à¦¥à¦® à¦•à§‹à¦¡à§‡à¦° config.encoder)
WSP_ENCODER_NAME = 'resnet50' # à¦‰à¦¦à¦¾à¦¹à¦°à¦£ (à¦¯à§‡à¦®à¦¨: 'resnet50', 'densenet121')

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦à¦Ÿà¦¿ à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦¬à§‡) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
        print(f"FocalLoss initialized with gamma={gamma}, alpha={alpha}")

    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        
        if self.alpha is not None:
            # alpha[target] à¦¥à§‡à¦•à§‡ à¦¸à¦ à¦¿à¦• à¦“à§Ÿà§‡à¦Ÿ à¦¨à¦¿à¦°à§à¦¬à¦¾à¦šà¦¨ à¦•à¦°à¦¾
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
            
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        # (à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # 70/30 à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿ
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à§¨): 'Moderate' (1) à¦à¦¬à¦‚ 'Severe' (2) à¦‰à¦­à§Ÿà¦•à§‡à¦‡ à¦¸à¦‚à¦–à§à¦¯à¦¾à¦²à¦˜à§ à¦§à¦°à¦¾ à¦¹à¦²à§‹ >>> ###
        minority_classes_list = [1, 2] 
        print(f"Applying strong augmentation to classes: {minority_classes_list}")
        
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. FocalLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for FocalLoss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + FocalLoss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦†à¦ªà¦¨à¦¾à¦° WSP à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° à¦¸à¦¹ ***) ---
        
        class WSPHybridClassifier(nn.Module):
            def __init__(self, encoder_name, n_classes=3, n_shape_features=5):
                super().__init__()
                print(f"Initializing Hybrid Classifier with base: {encoder_name}")
                
                # à§§. CNN à¦¬à§‡à¦¸ à¦²à§‹à¦¡ à¦•à¦°à¦¾ (timm à¦¥à§‡à¦•à§‡, à¦•à¦¿à¦¨à§à¦¤à§ ImageNet à¦“à§Ÿà§‡à¦Ÿ à¦›à¦¾à§œà¦¾)
                # à¦à¦Ÿà¦¿ à¦†à¦ªà¦¨à¦¾à¦° WSP à¦à¦¨à¦•à§‹à¦¡à¦¾à¦°à§‡à¦° à¦†à¦°à§à¦•à¦¿à¦Ÿà§‡à¦•à¦šà¦¾à¦°à§‡à¦° à¦¸à¦¾à¦¥à§‡ à¦®à¦¿à¦²à¦¤à§‡ à¦¹à¦¬à§‡
                try:
                    self.cnn_base = timm.create_model(encoder_name, pretrained=False)
                    cnn_feature_dim = self.cnn_base.num_features
                    self.cnn_base.reset_classifier(0) # à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à¦¯à¦¼à¦¾à¦° à¦¹à§‡à¦¡ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à¦¯à¦¼à¦¾
                except Exception as e:
                    print(f"!! Error creating model '{encoder_name}' with timm: {e}")
                    print("Falling back to resnet50. PLEASE CHECK YOUR 'WSP_ENCODER_NAME'.")
                    self.cnn_base = timm.create_model('resnet50', pretrained=False)
                    cnn_feature_dim = self.cnn_base.num_features
                    self.cnn_base.reset_classifier(0)

                print(f"CNN base created. Feature dimension: {cnn_feature_dim}")

                # à§¨. Tabular MLP (à¦¡à§à¦°à¦ªà¦†à¦‰à¦Ÿ à¦¸à¦¹)
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                # à§©. Fusion Classifier (à¦¶à¦•à§à¦¤à¦¿à¦¶à¦¾à¦²à§€ à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦°)
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 512), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.3),
                    nn.Linear(256, n_classes)
                )
                
            def load_wsp_weights(self, ckpt_path):
                print(f"--- Loading WSP pre-trained weights from: {ckpt_path} ---")
                try:
                    ckpt = torch.load(ckpt_path, map_location=device)
                    
                    # à¦¯à¦¦à¦¿ .ckpt à¦«à¦¾à¦‡à¦²à¦Ÿà¦¿ PyTorch Lightning-à¦à¦° à¦¹à§Ÿ, à¦¤à¦¬à§‡ state_dict 'state_dict'-à¦à¦° à¦­à§‡à¦¤à¦°à§‡ à¦¥à¦¾à¦•à§‡
                    if 'state_dict' in ckpt:
                        wsp_state_dict = ckpt['state_dict']
                    else:
                        wsp_state_dict = ckpt # à¦¯à¦¦à¦¿ à¦¶à§à¦§à§ state_dict à¦¸à§‡à¦­ à¦•à¦°à¦¾ à¦¹à§Ÿ
                        
                    encoder_state_dict = {}
                    
                    # à¦†à¦ªà¦¨à¦¾à¦° à¦ªà§à¦°à¦¥à¦® à¦•à§‹à¦¡ (Code 1) à¦…à¦¨à§à¦¯à¦¾à§Ÿà§€, à¦à¦¨à¦•à§‹à¦¡à¦¾à¦°à¦Ÿà¦¿ 'net.' à¦ªà§à¦°à¦«à¦¿à¦•à§à¦¸ à¦¦à¦¿à§Ÿà§‡ à¦¸à§‡à¦­ à¦¹à§Ÿ
                    prefix = 'net.' 
                    
                    for k, v in wsp_state_dict.items():
                        if k.startswith(prefix):
                            # à¦ªà§à¦°à¦«à¦¿à¦•à§à¦¸ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾ (à¦¯à§‡à¦®à¦¨ 'net.conv1.weight' -> 'conv1.weight')
                            new_key = k[len(prefix):]
                            encoder_state_dict[new_key] = v
                            
                    if not encoder_state_dict:
                        print("!! WARNING: No weights found with prefix 'net.'.")
                        print("Attempting to load the state_dict directly (assuming it's just the encoder).")
                        encoder_state_dict = wsp_state_dict # à¦¸à¦°à¦¾à¦¸à¦°à¦¿ à¦²à§‹à¦¡ à¦•à¦°à¦¾à¦° à¦šà§‡à¦·à§à¦Ÿà¦¾

                    # cnn_base-à¦ à¦“à§Ÿà§‡à¦Ÿ à¦²à§‹à¦¡ à¦•à¦°à¦¾ (strict=False à¦•à¦¾à¦°à¦£ à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ à¦®à¦¿à¦²à¦¬à§‡ à¦¨à¦¾)
                    incompatible_keys = self.cnn_base.load_state_dict(encoder_state_dict, strict=False)
                    print(f"WSP weights loaded successfully.")
                    print(f"  > Mismatched/Missing keys (should be OK if just classifier): {incompatible_keys}")
                
                except Exception as e:
                    print(f"!! ERROR: Failed to load WSP weights from {ckpt_path}. {e}")
                    print("!! MODEL WILL TRAIN FROM SCRATCH (RANDOM WEIGHTS).")
            
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image) # à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° à¦¥à§‡à¦•à§‡ à¦«à¦¿à¦šà¦¾à¦°
                tabular_features = self.tabular_mlp(shape_features) # MLP à¦¥à§‡à¦•à§‡ à¦«à¦¿à¦šà¦¾à¦°
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à§©): WSPHybridClassifier à¦•à§à¦²à¦¾à¦¸ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ >>> ###
        cls_model = WSPHybridClassifier(
            encoder_name=WSP_ENCODER_NAME,
            n_classes=3,
            n_shape_features=train_features.shape[1]
        )
        
        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à§ª): WSP à¦ªà§à¦°à¦¿-à¦Ÿà§à¦°à§‡à¦‡à¦¨à¦¡ à¦“à§Ÿà§‡à¦Ÿ à¦²à§‹à¦¡ à¦•à¦°à¦¾ >>> ###
        if os.path.exists(WSP_PRETRAIN_CKPT_PATH):
            cls_model.load_wsp_weights(WSP_PRETRAIN_CKPT_PATH)
        else:
            print(f"!! WARNING: WSP Checkpoint path not found: {WSP_PRETRAIN_CKPT_PATH}")
            print("!! Training from scratch (random initialization).")
            
        cls_model = cls_model.to(device)

        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à§«): FocalLoss à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ >>> ###
        # gamma=2.5 à¦à¦•à¦Ÿà¦¿ à¦­à¦¾à¦²à§‹ à¦¶à§à¦°à§à¥¤ à¦ªà§à¦°à§Ÿà§‹à¦œà¦¨à§‡ 2.0 à¦¬à¦¾ 3.0 à¦šà§‡à¦·à§à¦Ÿà¦¾ à¦•à¦°à¦¤à§‡ à¦ªà¦¾à¦°à§‡à¦¨à¥¤
        criterion = FocalLoss(alpha=alpha_weights, gamma=2.5, reduction='mean').to(device)
        
        # ### <<< à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ (à§¬): à¦¹à¦¾à¦‡à¦ªà¦¾à¦°-à¦ªà§à¦¯à¦¾à¦°à¦¾à¦®à¦¿à¦Ÿà¦¾à¦° à¦…à§à¦¯à¦¾à¦¡à¦œà¦¾à¦¸à§à¦Ÿà¦®à§‡à¦¨à§à¦Ÿ >>> ###
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-4) 
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print(f"\n--- Starting Hybrid Model (WSP {WSP_ENCODER_NAME} Base) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # (à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            # (à¦­à§à¦¯à¦¾à¦²à¦¿à¦¡à§‡à¦¶à¦¨ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        # à¦¸à§‡à¦°à¦¾ à¦®à¦¡à§‡à¦² à¦²à§‹à¦¡ à¦•à¦°à¦¾ (à¦¨à¦¤à§à¦¨ à¦†à¦°à§à¦•à¦¿à¦Ÿà§‡à¦•à¦šà¦¾à¦° à¦¸à¦¹)
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title(f"Confusion Matrix for Hybrid Model (WSP {WSP_ENCODER_NAME} Base)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import warnings

# --- 0. à¦¸à§‡à¦Ÿà¦¿à¦‚à¦¸ ---
warnings.filterwarnings('ignore')

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# à¦†à¦‰à¦Ÿà¦ªà§à¦Ÿ à¦¡à¦¿à¦°à§‡à¦•à§à¦Ÿà¦°à¦¿
OUTPUT_DIR = '/kaggle/working/'
SEG_MASK_DIR = os.path.join(OUTPUT_DIR, 'generated_masks')
os.makedirs(SEG_MASK_DIR, exist_ok=True)
print(f"Generated masks will be saved to: {SEG_MASK_DIR}")

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    
    # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Age à¦à¦¬à¦‚ Gender à¦¸à¦¹ à¦²à§‹à¦¡ à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡
    metadata_df.dropna(subset=['radiological_evaluation', 'Age', 'Gender'], inplace=True)
    
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    
    # Gender à¦®à§à¦¯à¦¾à¦ª à¦•à¦°à¦¾
    metadata_df['Gender'] = metadata_df['Gender'].map({'Male': 0, 'Female': 1})
    
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡) ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# =================================================================================
# --- 7. PRE-PROCESSING (SEGMENTATION + FEATURE EXTRACTION) ---
# =================================================================================

if not models_loaded or test_df is None:
    print("Skipping pre-processing stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions (à¦†à¦—à§‡à¦° à¦®à¦¤à§‹à¦‡) ---
        print("\nGenerating segmentation predictions (This will take time)...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks (à¦†à¦—à§‡à¦° à¦®à¦¤à§‹à¦‡) ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)
        print("Shape features extracted.")

        # --- 7.3 à¦¸à§à¦•à§‡à¦²à¦¿à¦‚ à¦«à¦¿à¦šà¦¾à¦° (à¦†à¦—à§‡à¦° à¦®à¦¤à§‹à¦‡) ---
        scaler = StandardScaler()
        train_features_scaled = scaler.fit_transform(train_features_df)
        val_features_scaled   = scaler.transform(val_features_df)
        test_features_scaled  = scaler.transform(test_features_df)
        
        # à¦¸à§à¦•à§‡à¦² à¦•à¦°à¦¾ à¦«à¦¿à¦šà¦¾à¦°à¦—à§à¦²à§‹à¦•à§‡ à¦†à¦¬à¦¾à¦° DataFrame-à¦ à¦°à§‚à¦ªà¦¾à¦¨à§à¦¤à¦° à¦•à¦°à¦¾
        scaled_cols = train_features_df.columns
        scaled_train_features_df = pd.DataFrame(train_features_scaled, columns=scaled_cols)
        scaled_val_features_df = pd.DataFrame(val_features_scaled, columns=scaled_cols)
        scaled_test_features_df = pd.DataFrame(test_features_scaled, columns=scaled_cols)
        print("Shape features scaled.")

        # --- 7.4 [à¦¨à¦¤à§à¦¨] à¦®à¦¾à¦¸à§à¦• à¦¸à§‡à¦­ à¦•à¦°à¦¾ à¦à¦¬à¦‚ à¦šà§‚à§œà¦¾à¦¨à§à¦¤ CSV à¦¤à§ˆà¦°à¦¿ à¦•à¦°à¦¾ ---
        
        def save_masks_and_merge_data(original_df, mask_tensors, scaled_features_df, split_name):
            print(f"\nProcessing and saving data for: {split_name}")
            split_mask_dir = os.path.join(SEG_MASK_DIR, split_name)
            os.makedirs(split_mask_dir, exist_ok=True)
            
            new_mask_paths = []
            
            for i in tqdm(range(len(original_df)), desc=f"Saving {split_name} masks"):
                # à¦…à¦°à¦¿à¦œà¦¿à¦¨à¦¾à¦² à¦ªà¦¾à¦¥ à¦¥à§‡à¦•à§‡ à¦«à¦¾à¦‡à¦² à¦¨à¦¾à¦® à¦¨à§‡à¦“à§Ÿà¦¾
                original_image_path = original_df.iloc[i]['image_file_path']
                base_filename = os.path.basename(original_image_path)
                # à¦¨à¦¤à§à¦¨ à¦ªà¦¾à¦¥ à¦¤à§ˆà¦°à¦¿ à¦•à¦°à¦¾ (PNG à¦¹à¦¿à¦¸à¦¾à¦¬à§‡ à¦¸à§‡à¦­ à¦•à¦°à¦¾ à¦­à¦¾à¦²à§‹)
                new_mask_filename = os.path.splitext(base_filename)[0] + '.png'
                new_mask_path = os.path.join(split_mask_dir, new_mask_filename)
                
                # à¦®à¦¾à¦¸à§à¦• à¦Ÿà§‡à¦¨à¦¸à¦°à¦•à§‡ numpy à¦…à§à¦¯à¦¾à¦°à§‡à¦¤à§‡ à¦°à§‚à¦ªà¦¾à¦¨à§à¦¤à¦° à¦•à¦°à¦¾
                # à¦®à¦¾à¦¸à§à¦•à§‡à¦° à¦†à¦•à§ƒà¦¤à¦¿ [1, H, W] à¦¥à§‡à¦•à§‡ [H, W] à¦•à¦°à¦¾
                mask_array = mask_tensors[i, 0].numpy() 
                # à¦®à¦¾à¦¸à§à¦•à¦•à§‡ 0-255 à¦°à§‡à¦à§à¦œà§‡ à¦†à¦¨à¦¾
                mask_image = (mask_array * 255).astype(np.uint8)
                
                # à¦®à¦¾à¦¸à§à¦• à¦‡à¦®à§‡à¦œ à¦¸à§‡à¦­ à¦•à¦°à¦¾
                cv2.imwrite(new_mask_path, mask_image)
                
                # à¦¨à¦¤à§à¦¨ à¦ªà¦¾à¦¥ à¦²à¦¿à¦¸à§à¦Ÿà§‡ à¦¯à§‹à¦— à¦•à¦°à¦¾
                new_mask_paths.append(new_mask_path)
                
            # à¦¨à¦¤à§à¦¨ DataFrame à¦¤à§ˆà¦°à¦¿ à¦•à¦°à¦¾
            df_copy = original_df.copy().reset_index(drop=True)
            features_copy = scaled_features_df.copy().reset_index(drop=True)
            
            # à¦¨à¦¤à§à¦¨ à¦¸à§‡à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¡ à¦®à¦¾à¦¸à§à¦•à§‡à¦° à¦ªà¦¾à¦¥ à¦¯à§‹à¦— à¦•à¦°à¦¾
            df_copy['segmented_mask_path'] = new_mask_paths
            
            # à¦®à§‚à¦² DataFrame à¦à¦¬à¦‚ à¦¸à§à¦•à§‡à¦² à¦•à¦°à¦¾ à¦«à¦¿à¦šà¦¾à¦°à§‡à¦° DataFrame-à¦•à§‡ à¦¯à§à¦•à§à¦¤ à¦•à¦°à¦¾
            final_df = pd.concat([df_copy, features_copy], axis=1)
            
            # à¦•à¦²à¦¾à¦®à¦—à§à¦²à§‹ à¦¸à¦¾à¦œà¦¾à¦¨à§‹ (à¦†à¦ªà¦¨à¦¾à¦° à¦…à¦¨à§à¦°à§‹à¦§ à¦…à¦¨à§à¦¯à¦¾à§Ÿà§€)
            all_cols = ['ID', 'image_file_path', 'mask_file_path', 'segmented_mask_path', 
                        'radiological_evaluation', 'Age', 'Gender', 'class_label', 'class_name']
            all_cols.extend(scaled_cols) # shape feature à¦•à¦²à¦¾à¦®à§‡à¦° à¦¨à¦¾à¦®à¦—à§à¦²à§‹
            
            # à¦¨à¦¿à¦¶à§à¦šà¦¿à¦¤ à¦•à¦°à¦¾ à¦¯à§‡ à¦¸à¦¬ à¦•à¦²à¦¾à¦® à¦†à¦›à§‡
            final_df = final_df[[col for col in all_cols if col in final_df.columns]]
            
            # à¦šà§‚à§œà¦¾à¦¨à§à¦¤ CSV à¦¸à§‡à¦­ à¦•à¦°à¦¾
            output_csv_path = os.path.join(OUTPUT_DIR, f"{split_name}_processed.csv")
            final_df.to_csv(output_csv_path, index=False)
            print(f"Successfully saved processed data to: {output_csv_path}")
            return final_df

        # à¦¤à¦¿à¦¨à¦Ÿà¦¿ à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿà§‡à¦° à¦œà¦¨à§à¦¯à¦‡ à¦«à¦¾à¦‚à¦¶à¦¨à¦Ÿà¦¿ à¦°à¦¾à¦¨ à¦•à¦°à¦¾
        final_train_df = save_masks_and_merge_data(train_df, train_pred_masks, scaled_train_features_df, 'train')
        final_val_df = save_masks_and_merge_data(val_df, val_pred_masks, scaled_val_features_df, 'val')
        final_test_df = save_masks_and_merge_data(test_df, test_pred_masks, scaled_test_features_df, 'test')

        print("\nAll preprocessing, segmentation, and data saving is complete! âœ…")
        print("You can now run the classification script using the following files:")
        print(f"Train CSV: {OUTPUT_DIR}train_processed.csv")
        print(f"Val CSV: {OUTPUT_DIR}val_processed.csv")
        print(f"Test CSV: {OUTPUT_DIR}test_processed.csv")

    except Exception as e:
        print(f"\nAn error occurred during preprocessing: {e}")
        import traceback
        traceback.print_exc()
```

```python
# ğŸ“¦ Install required libraries
!pip install -q albumentations scikit-image tqdm

# ğŸ§  SimCLR + ResNet50 + Tabular Fusion for Radiological Evaluation
import os, cv2, torch, numpy as np, pandas as pd, matplotlib.pyplot as plt
from PIL import Image
import torch.nn as nn
from torchvision import models
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.preprocessing import StandardScaler
import albumentations as A
from albumentations.pytorch import ToTensorV2

# --- CONFIG ---
train_csv_path = '/kaggle/input/segmented-data/new_data/train_processed.csv'
val_csv_path = '/kaggle/input/segmented-data/new_data/val_processed.csv'
test_csv_path = '/kaggle/input/segmented-data/new_data/test_processed.csv'
IMAGE_SIZE = (224, 224)
TABULAR_COLS = ['Age', 'area', 'perimeter', 'solidity', 'eccentricity', 'circularity']
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- LOAD & CLEAN DATA ---
def clean_tabular(df):
    df = df.copy()
    for col in TABULAR_COLS:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df.dropna(subset=TABULAR_COLS + ['class_label', 'image_file_path'], inplace=True)
    return df

train_df = clean_tabular(pd.read_csv(train_csv_path))
val_df = clean_tabular(pd.read_csv(val_csv_path))
test_df = clean_tabular(pd.read_csv(test_csv_path))

# --- NORMALIZE TABULAR FEATURES ---
scaler = StandardScaler()
train_df[TABULAR_COLS] = scaler.fit_transform(train_df[TABULAR_COLS])
val_df[TABULAR_COLS] = scaler.transform(val_df[TABULAR_COLS])
test_df[TABULAR_COLS] = scaler.transform(test_df[TABULAR_COLS])

# --- BALANCED SAMPLER ---
def get_sampler(df):
    class_counts = df['class_label'].value_counts().to_dict()
    weights = df['class_label'].map(lambda x: 1.0 / class_counts[x])
    return WeightedRandomSampler(weights.values, len(weights))

# --- DATASET ---
class CirrhosisFusionDataset(Dataset):
    def __init__(self, df, augment=True):
        self.df = df.reset_index(drop=True)
        self.transform = A.Compose([
            A.Resize(*IMAGE_SIZE),
            A.HorizontalFlip(p=0.5),
            A.RandomBrightnessContrast(p=0.5),
            A.Normalize(),
            ToTensorV2()
        ]) if augment else A.Compose([
            A.Resize(*IMAGE_SIZE),
            A.Normalize(),
            ToTensorV2()
        ])

    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image = cv2.imread(row['image_file_path'], cv2.IMREAD_GRAYSCALE)
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        image = self.transform(image=image)['image']
        tabular = torch.tensor(row[TABULAR_COLS].astype(np.float32).values)
        label = row['class_label']
        return image, tabular, label

# --- MODEL ---
class SimCLRTabularClassifier(nn.Module):
    def __init__(self, num_classes=3, tabular_dim=6):
        super().__init__()
        self.encoder = models.resnet50(pretrained=True)
        self.encoder.fc = nn.Identity()
        self.tabular_fc = nn.Sequential(
            nn.Linear(tabular_dim, 32),
            nn.ReLU(),
            nn.Dropout(0.2)
        )
        self.classifier = nn.Sequential(
            nn.Linear(2048 + 32, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x, tabular):
        img_feat = self.encoder(x)
        tab_feat = self.tabular_fc(tabular)
        combined = torch.cat([img_feat, tab_feat], dim=1)
        return self.classifier(combined)

# --- TRAINING ---
def train_model(model, train_loader, val_loader, epochs=20):
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='max', patience=3, factor=0.5)
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        train_loss, correct = 0, 0
        for x, tab, y in train_loader:
            x, tab, y = x.to(device), tab.to(device), y.to(device)
            optimizer.zero_grad()
            out = model(x, tab)
            loss = criterion(out, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
        acc = correct / len(train_loader.dataset)
        print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f}, Accuracy={acc:.4f}")

        model.eval()
        val_correct = 0
        with torch.no_grad():
            for x, tab, y in val_loader:
                x, tab, y = x.to(device), tab.to(device), y.to(device)
                out = model(x, tab)
                val_correct += (out.argmax(1) == y).sum().item()
        val_acc = val_correct / len(val_loader.dataset)
        print(f"Validation Accuracy: {val_acc:.4f}")
        scheduler.step(val_acc)

# --- EVALUATION ---
def evaluate_model(model, test_loader):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for x, tab, y in test_loader:
            x, tab = x.to(device), tab.to(device)
            out = model(x, tab)
            preds = out.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(y.numpy())
    print(classification_report(all_labels, all_preds, target_names=['Mild', 'Moderate', 'Severe']))
    cm = confusion_matrix(all_labels, all_preds)
    ConfusionMatrixDisplay(cm, display_labels=['Mild', 'Moderate', 'Severe']).plot(cmap='Blues')
    plt.show()

# --- RUN ---
train_loader = DataLoader(CirrhosisFusionDataset(train_df), batch_size=32, sampler=get_sampler(train_df))
val_loader = DataLoader(CirrhosisFusionDataset(val_df, augment=False), batch_size=32)
test_loader = DataLoader(CirrhosisFusionDataset(test_df, augment=False), batch_size=32)

model = SimCLRTabularClassifier(num_classes=3, tabular_dim=len(TABULAR_COLS))
train_model(model, train_loader, val_loader, epochs=20)
evaluate_model(model, test_loader)
```

```python
# ğŸ“¦ Install required libraries
!pip install -q albumentations scikit-image tqdm

# ğŸ§  SimCLR + ResNet18 + Tabular Fusion for Radiological Evaluation
import os, cv2, torch, numpy as np, pandas as pd, matplotlib.pyplot as plt
from PIL import Image
import torch.nn as nn
from torchvision import models
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
from sklearn.preprocessing import StandardScaler
import albumentations as A
from albumentations.pytorch import ToTensorV2

# --- CONFIG ---
train_csv_path = '/kaggle/input/segmented-data/new_data/train_processed.csv'
val_csv_path = '/kaggle/input/segmented-data/new_data/val_processed.csv'
test_csv_path = '/kaggle/input/segmented-data/new_data/test_processed.csv'
IMAGE_SIZE = (224, 224)
TABULAR_COLS = ['Age', 'area', 'perimeter', 'solidity', 'eccentricity', 'circularity']
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- LOAD & CLEAN DATA ---
def clean_tabular(df):
    df = df.copy()
    for col in TABULAR_COLS:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df.dropna(subset=TABULAR_COLS + ['class_label', 'image_file_path'], inplace=True)
    return df

train_df = clean_tabular(pd.read_csv(train_csv_path))
val_df = clean_tabular(pd.read_csv(val_csv_path))
test_df = clean_tabular(pd.read_csv(test_csv_path))

# --- NORMALIZE TABULAR FEATURES ---
scaler = StandardScaler()
train_df[TABULAR_COLS] = scaler.fit_transform(train_df[TABULAR_COLS])
val_df[TABULAR_COLS] = scaler.transform(val_df[TABULAR_COLS])
test_df[TABULAR_COLS] = scaler.transform(test_df[TABULAR_COLS])

# --- BALANCED SAMPLER ---
def get_sampler(df):
    class_counts = df['class_label'].value_counts().to_dict()
    weights = df['class_label'].map(lambda x: 1.0 / class_counts[x])
    return WeightedRandomSampler(weights.values, len(weights))

# --- DATASET ---
class CirrhosisFusionDataset(Dataset):
    def __init__(self, df, augment=True):
        self.df = df.reset_index(drop=True)
        self.transform = A.Compose([
            A.Resize(*IMAGE_SIZE),
            A.HorizontalFlip(p=0.5),
            A.RandomBrightnessContrast(p=0.5),
            A.Normalize(),
            ToTensorV2()
        ]) if augment else A.Compose([
            A.Resize(*IMAGE_SIZE),
            A.Normalize(),
            ToTensorV2()
        ])

    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image = cv2.imread(row['image_file_path'], cv2.IMREAD_GRAYSCALE)
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        image = self.transform(image=image)['image']
        tabular = torch.tensor(row[TABULAR_COLS].astype(np.float32).values)
        label = row['class_label']
        return image, tabular, label

# --- MODEL ---
class Identity(nn.Module):
    def __init__(self): super().__init__()
    def forward(self, x): return x

class SimCLRTabularClassifier(nn.Module):
    def __init__(self, num_classes=3, tabular_dim=6):
        super().__init__()
        resnet18 = models.resnet18(pretrained=True)
        resnet18.fc = Identity()
        self.encoder = resnet18
        self.tabular_fc = nn.Sequential(
            nn.Linear(tabular_dim, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU()
        )
        self.classifier = nn.Sequential(
            nn.Linear(512 + 32, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x, tabular):
        img_feat = self.encoder(x)
        tab_feat = self.tabular_fc(tabular)
        combined = torch.cat([img_feat, tab_feat], dim=1)
        return self.classifier(combined)

# --- TRAINING ---
def train_model(model, train_loader, val_loader, epochs=20):
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='max', patience=3, factor=0.5)
    class_weights = torch.tensor([1.0, 3.0, 1.5]).to(device)
    criterion = nn.CrossEntropyLoss(weight=class_weights)

    for epoch in range(epochs):
        model.train()
        train_loss, correct = 0, 0
        for x, tab, y in train_loader:
            x, tab, y = x.to(device), tab.to(device), y.to(device)
            optimizer.zero_grad()
            out = model(x, tab)
            loss = criterion(out, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
        acc = correct / len(train_loader.dataset)
        print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f}, Accuracy={acc:.4f}")

        model.eval()
        val_correct = 0
        with torch.no_grad():
            for x, tab, y in val_loader:
                x, tab, y = x.to(device), tab.to(device), y.to(device)
                out = model(x, tab)
                val_correct += (out.argmax(1) == y).sum().item()
        val_acc = val_correct / len(val_loader.dataset)
        print(f"Validation Accuracy: {val_acc:.4f}")
        scheduler.step(val_acc)

# --- EVALUATION ---
def evaluate_model(model, test_loader):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for x, tab, y in test_loader:
            x, tab = x.to(device), tab.to(device)
            out = model(x, tab)
            preds = out.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(y.numpy())
    print(classification_report(all_labels, all_preds, target_names=['Mild', 'Moderate', 'Severe']))
    cm = confusion_matrix(all_labels, all_preds)
    ConfusionMatrixDisplay(cm, display_labels=['Mild', 'Moderate', 'Severe']).plot(cmap='Blues')
    plt.show()

# --- RUN ---
train_loader = DataLoader(CirrhosisFusionDataset(train_df), batch_size=32, sampler=get_sampler(train_df))
val_loader = DataLoader(CirrhosisFusionDataset(val_df, augment=False), batch_size=32)
test_loader = DataLoader(CirrhosisFusionDataset(test_df, augment=False), batch_size=32)

model = SimCLRTabularClassifier(num_classes=3, tabular_dim=len(TABULAR_COLS))
train_model(model, train_loader, val_loader, epochs=20)
evaluate_model(model, test_loader)
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))


# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦à¦–à¦¨ à¦à¦Ÿà¦¿ à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦¬à§‡ ***) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦¶à§à¦§à§à¦®à¦¾à¦¤à§à¦° à¦®à¦¾à¦¸à§à¦•à¦¡ à¦…à¦‚à¦¶ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾
                masked_image = img_tensor * pred_mask.clone().detach() 
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [1] 
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. FocalLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦¬à¦¨à§à¦§ à¦•à¦°à¦¾ à¦¹à¦²à§‹)
        # train_sampler = ...

        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: DataLoader à¦¥à§‡à¦•à§‡ sampler=None à¦à¦¬à¦‚ shuffle=True à¦•à¦°à¦¾ à¦¹à¦²à§‹
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=None, shuffle=True, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (FocalLoss + No Sampler). ğŸ“Š")


        # --- 7.4 Hybrid Classifier (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: SimCLR à¦¬à§‡à¦¸ à¦¸à¦¹ ***) ---
        
        # SimCLR à¦®à¦¡à§‡à¦²à§‡à¦° à¦œà¦¨à§à¦¯ Identity à¦•à§à¦²à¦¾à¦¸ (à¦†à¦ªà¦¨à¦¾à¦° à¦¦à§‡à¦“à§Ÿà¦¾)
        class Identity(nn.Module):
            def __init__(self):
                super(Identity, self).__init__()
            def forward(self, x):
                return x

        class SimCLRHybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    # à§§. SimCLR-à¦à¦° à¦¬à§‡à¦¸ (ResNet-18) à¦²à§‹à¦¡ à¦•à¦°à¦¾ (Pretrained)
                    self.cnn_base = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features  # 512
                    # à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾
                    self.cnn_base.fc = Identity() 
                except Exception:
                    # à¦‡à¦¨à§à¦Ÿà¦¾à¦°à¦¨à§‡à¦Ÿ à¦¨à¦¾ à¦¥à¦¾à¦•à¦²à§‡ à¦¬à¦¾ à¦ªà§à¦°à¦¾à¦¨à§‹ torchvision à¦¹à¦²à§‡
                    self.cnn_base = models.resnet18(pretrained=True)
                    cnn_feature_dim = self.cnn_base.fc.in_features # 512
                    self.cnn_base.fc = Identity()

                # à§¨. Tabular MLP (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                # à§©. Fusion Classifier (à¦‡à¦¨à¦ªà§à¦Ÿ à¦¡à¦¾à¦‡à¦®à§‡à¦¨à¦¶à¦¨ à¦†à¦ªà¦¡à§‡à¦Ÿ à¦•à¦°à¦¾ à¦¹à§Ÿà§‡à¦›à§‡)
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.3), # 512 + 32 = 544
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image) # 512-dim à¦«à¦¿à¦šà¦¾à¦°
                tabular_features = self.tabular_mlp(shape_features) # 32-dim à¦«à¦¿à¦šà¦¾à¦°
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: SimCLRHybridClassifier à¦•à§à¦²à¦¾à¦¸ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾
        cls_model = SimCLRHybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: FocalLoss à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾
        criterion = FocalLoss(alpha=alpha_weights, gamma=3.0).to(device)
        
        # à¦²à¦¾à¦°à§à¦¨à¦¿à¦‚ à¦°à§‡à¦Ÿ à¦¸à¦¾à¦®à¦¾à¦¨à§à¦¯ à¦•à¦®à¦¾à¦¨à§‹ (à¦•à¦¾à¦°à¦£ ResNet-18 à¦›à§‹à¦Ÿ à¦®à¦¡à§‡à¦²)
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4, weight_decay=1e-5) 
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model (SimCLR-ResNet18) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # (à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            # (à¦­à§à¦¯à¦¾à¦²à¦¿à¦¡à§‡à¦¶à¦¨ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (SimCLR-Base + FocalLoss)") # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm  # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: timm à¦‡à¦®à§à¦ªà§‹à¦°à§à¦Ÿ à¦•à¦°à¦¾ à¦¹à¦²à§‹

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦†à¦° à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦¤à¦¬à§‡ à¦°à¦¾à¦–à¦¾ à¦¹à¦²à§‹) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        # (à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # <-- 70/30 à¦“à§Ÿà§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿà§‡ à¦«à¦¿à¦°à§‡ à¦†à¦¸à¦¾ (à¦†à¦ªà¦¨à¦¾à¦° à¦®à§‚à¦² à¦•à§‹à¦¡à§‡à¦° à¦®à¦¤à§‹)
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [1] 
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. CrossEntropyLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        # DataLoader-à¦ sampler=train_sampler à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + Weighted CrossEntropyLoss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: timm SimCLR à¦¬à§‡à¦¸ à¦¸à¦¹ ***) ---
        
        class TimmSimCLRHybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    # à§§. timm à¦¥à§‡à¦•à§‡ SimCLR (CIFAR-10) ResNet-50 à¦²à§‹à¦¡ à¦•à¦°à¦¾
                    print("Loading timm model: resnet50_simclr_cifar10...")
                    self.cnn_base = timm.create_model("resnet50_simclr_cifar10", pretrained=True)
                    
                    # à§¨. à¦«à¦¿à¦šà¦¾à¦° à¦¡à¦¾à¦‡à¦®à§‡à¦¨à¦¶à¦¨ à¦¨à§‡à¦“à§Ÿà¦¾
                    cnn_feature_dim = self.cnn_base.num_features  # ResNet-50 à¦à¦° à¦œà¦¨à§à¦¯ 2048
                    
                    # à§©. à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾ (timm-à¦à¦° à¦¨à¦¿à¦œà¦¸à§à¦¬ à¦ªà¦¦à§à¦§à¦¤à¦¿)
                    self.cnn_base.reset_classifier(0) # 0 à¦®à¦¾à¦¨à§‡ Identity à¦²à§‡à§Ÿà¦¾à¦°
                    print(f"timm model loaded. Feature dimension: {cnn_feature_dim}")

                except Exception as e:
                    print(f"Error loading timm model: {e}")
                    print("Falling back to standard ImageNet ResNet-50.")
                    # à¦•à§‹à¦¨à§‹ à¦¸à¦®à¦¸à§à¦¯à¦¾ à¦¹à¦²à§‡ ImageNet pretrained à¦®à¦¡à§‡à¦²à§‡ à¦«à¦¿à¦°à§‡ à¦¯à¦¾à¦“à§Ÿà¦¾
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features  # 2048
                    self.cnn_base.fc = nn.Identity()

                # à§ª. Tabular MLP (à¦†à¦ªà¦¨à¦¾à¦° à¦†à¦—à§‡à¦° à¦•à§‹à¦¡à§‡à¦° à¦¡à§à¦°à¦ªà¦†à¦‰à¦Ÿ à¦¸à¦¹)
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                # à§«. Fusion Classifier (à¦‡à¦¨à¦ªà§à¦Ÿ à¦¡à¦¾à¦‡à¦®à§‡à¦¨à¦¶à¦¨ à¦†à¦ªà¦¡à§‡à¦Ÿ à¦•à¦°à¦¾ à¦¹à§Ÿà§‡à¦›à§‡)
                self.fusion_classifier = nn.Sequential(
                    # 2048 (ResNet50) + 32 (Tabular) = 2080
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.3), 
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image) # 2048-dim à¦«à¦¿à¦šà¦¾à¦°
                tabular_features = self.tabular_mlp(shape_features) # 32-dim à¦«à¦¿à¦šà¦¾à¦°
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: TimmSimCLRHybridClassifier à¦•à§à¦²à¦¾à¦¸ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾
        cls_model = TimmSimCLRHybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Weighted CrossEntropyLoss-à¦ à¦«à¦¿à¦°à§‡ à¦†à¦¸à¦¾
        criterion = nn.CrossEntropyLoss(weight=alpha_weights).to(device)
        
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=5e-5, weight_decay=1e-5) 
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model (timm SimCLR Base) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # (à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            # (à¦­à§à¦¯à¦¾à¦²à¦¿à¦¡à§‡à¦¶à¦¨ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (timm SimCLR Base)") # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm  # <-- timm à¦‡à¦®à§à¦ªà§‹à¦°à§à¦Ÿ à¦•à¦°à¦¾ à¦¹à¦²à§‹

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦†à¦° à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦¤à¦¬à§‡ à¦°à¦¾à¦–à¦¾ à¦¹à¦²à§‹) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        # (à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # <-- 70/30 à¦“à§Ÿà§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿà§‡ à¦«à¦¿à¦°à§‡ à¦†à¦¸à¦¾ (à¦†à¦ªà¦¨à¦¾à¦° à¦®à§‚à¦² à¦•à§‹à¦¡à§‡à¦° à¦®à¦¤à§‹)
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [1] 
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. CrossEntropyLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        # DataLoader-à¦ sampler=train_sampler à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + Weighted CrossEntropyLoss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: timm SimCLR à¦¬à§‡à¦¸ à¦¸à¦¹ ***) ---
        
        class TimmSimCLRHybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    # à§§. timm à¦¥à§‡à¦•à§‡ SimCLR (CIFAR-10) ResNet-50 à¦²à§‹à¦¡ à¦•à¦°à¦¾
                    print("Loading timm model: resnet50_simclr_cifar10...")
                    self.cnn_base = timm.create_model("resnet50_simclr_cifar10", pretrained=True)
                    
                    # à§¨. à¦«à¦¿à¦šà¦¾à¦° à¦¡à¦¾à¦‡à¦®à§‡à¦¨à¦¶à¦¨ à¦¨à§‡à¦“à§Ÿà¦¾
                    cnn_feature_dim = self.cnn_base.num_features  # ResNet-50 à¦à¦° à¦œà¦¨à§à¦¯ 2048
                    
                    # à§©. à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾ (timm-à¦à¦° à¦¨à¦¿à¦œà¦¸à§à¦¬ à¦ªà¦¦à§à¦§à¦¤à¦¿)
                    self.cnn_base.reset_classifier(0) # 0 à¦®à¦¾à¦¨à§‡ Identity à¦²à§‡à§Ÿà¦¾à¦°
                    print(f"timm model loaded. Feature dimension: {cnn_feature_dim}")

                except Exception as e:
                    print(f"Error loading timm model: {e}")
                    print("Falling back to standard ImageNet ResNet-50.")
                    # à¦•à§‹à¦¨à§‹ à¦¸à¦®à¦¸à§à¦¯à¦¾ à¦¹à¦²à§‡ ImageNet pretrained à¦®à¦¡à§‡à¦²à§‡ à¦«à¦¿à¦°à§‡ à¦¯à¦¾à¦“à§Ÿà¦¾
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features  # 2048
                    self.cnn_base.fc = nn.Identity()

                # à§ª. Tabular MLP (à¦†à¦ªà¦¨à¦¾à¦° à¦†à¦—à§‡à¦° à¦•à§‹à¦¡à§‡à¦° à¦¡à§à¦°à¦ªà¦†à¦‰à¦Ÿ à¦¸à¦¹)
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                # à§«. Fusion Classifier (à¦‡à¦¨à¦ªà§à¦Ÿ à¦¡à¦¾à¦‡à¦®à§‡à¦¨à¦¶à¦¨ à¦†à¦ªà¦¡à§‡à¦Ÿ à¦•à¦°à¦¾ à¦¹à§Ÿà§‡à¦›à§‡)
                self.fusion_classifier = nn.Sequential(
                    # 2048 (ResNet50) + 32 (Tabular) = 2080
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.3), 
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image) # 2048-dim à¦«à¦¿à¦šà¦¾à¦°
                tabular_features = self.tabular_mlp(shape_features) # 32-dim à¦«à¦¿à¦šà¦¾à¦°
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: TimmSimCLRHybridClassifier à¦•à§à¦²à¦¾à¦¸ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾
        cls_model = TimmSimCLRHybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Weighted CrossEntropyLoss-à¦ à¦«à¦¿à¦°à§‡ à¦†à¦¸à¦¾
        criterion = nn.CrossEntropyLoss(weight=alpha_weights).to(device)
        
        # <--*** à¦ªà§à¦°à¦§à¦¾à¦¨ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Differential Learning Rate ***-->
        
        # à§§. à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° (cnn_base) à¦à¦° à¦ªà§à¦¯à¦¾à¦°à¦¾à¦®à¦¿à¦Ÿà¦¾à¦°à¦—à§à¦²à§‹ à¦†à¦²à¦¾à¦¦à¦¾ à¦•à¦°à§à¦¨
        try:
            encoder_params = cls_model.cnn_base.parameters()
            
            # à§¨. à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ (tabular_mlp, fusion_classifier) à¦à¦° à¦ªà§à¦¯à¦¾à¦°à¦¾à¦®à¦¿à¦Ÿà¦¾à¦° à¦†à¦²à¦¾à¦¦à¦¾ à¦•à¦°à§à¦¨
            classifier_params = list(cls_model.tabular_mlp.parameters()) + list(cls_model.fusion_classifier.parameters())

            # à§©. à¦¦à§à¦Ÿà¦¿ à¦­à¦¿à¦¨à§à¦¨ à¦²à¦¾à¦°à§à¦¨à¦¿à¦‚ à¦°à§‡à¦Ÿ à¦¸à¦¹ à¦…à¦ªà§à¦Ÿà¦¿à¦®à¦¾à¦‡à¦œà¦¾à¦° à¦¸à§‡à¦Ÿà¦†à¦ª à¦•à¦°à§à¦¨
            optimizer = torch.optim.AdamW([
                {'params': encoder_params, 'lr': 1e-5},      # à¦à¦¨à¦•à§‹à¦¡à¦¾à¦°à§‡à¦° à¦œà¦¨à§à¦¯ à¦•à¦® LR (à¦«à¦¾à¦‡à¦¨-à¦Ÿà¦¿à¦‰à¦¨)
                {'params': classifier_params, 'lr': 1e-4}     # à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦°à§‡à¦° à¦œà¦¨à§à¦¯ à¦¬à§‡à¦¶à¦¿ LR (à¦¨à¦¤à§à¦¨ à¦•à¦°à§‡ à¦¶à§‡à¦–à¦¾)
            ], weight_decay=1e-5)
            
            print("\nUsing Differential Learning Rate (Encoder LR: 1e-5, Classifier LR: 1e-4).")

        except Exception as e:
            print(f"Could not set differential LR: {e}. Falling back to single LR.")
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=5e-5, weight_decay=1e-5) 

        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model (timm SimCLR Base + Differential LR) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # (à¦Ÿà§à¦°à§‡à¦¨à¦¿à¦‚ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            # (à¦­à§à¦¯à¦¾à¦²à¦¿à¦¡à§‡à¦¶à¦¨ à¦²à§à¦ª à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (timm SimCLR + Diff. LR)") # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm  # <-- timm à¦‡à¦®à§à¦ªà§‹à¦°à§à¦Ÿ à¦•à¦°à¦¾ à¦¹à¦²à§‹
import warnings

# --- 0. à¦¸à§‡à¦Ÿà¦¿à¦‚à¦¸ ---
warnings.filterwarnings('ignore') # torch.hub à¦¬à¦¾ timm à¦¥à§‡à¦•à§‡ ÙˆØ§Ø±Ù†Ù†Ú¯ à¦‡à¦—à¦¨à§‹à¦° à¦•à¦°à¦¾à¦° à¦œà¦¨à§à¦¯

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    
    # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Age à¦à¦¬à¦‚ Gender à¦¸à¦¹ à¦²à§‹à¦¡ à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡
    metadata_df.dropna(subset=['radiological_evaluation', 'Age', 'Gender'], inplace=True)
    
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    
    # Gender à¦®à§à¦¯à¦¾à¦ª à¦•à¦°à¦¾
    metadata_df['Gender'] = metadata_df['Gender'].map({'Male': 0, 'Female': 1})
    
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Age, Gender à¦¸à¦¹ merge à¦•à¦°à¦¾
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    print(f"Metadata columns loaded: {train_df.columns.tolist()}")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦†à¦° à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦¤à¦¬à§‡ à¦°à¦¾à¦–à¦¾ à¦¹à¦²à§‹) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features + Age/Gender (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦à¦–à¦¾à¦¨à§‡ ***) ---
        print("\nEngineering tabular features (Shape + Age/Gender)...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Shape Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        # Radiomics/Shape à¦«à¦¿à¦šà¦¾à¦° à¦¤à§ˆà¦°à¦¿
        train_shape_features_df = get_shape_features(train_pred_masks)
        val_shape_features_df   = get_shape_features(val_pred_masks)
        test_shape_features_df  = get_shape_features(test_pred_masks)
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Shape à¦«à¦¿à¦šà¦¾à¦° à¦à¦° à¦¸à¦¾à¦¥à§‡ Age à¦à¦¬à¦‚ Gender à¦¯à§à¦•à§à¦¤ à¦•à¦°à¦¾
        train_tabular_df = pd.concat([train_shape_features_df, train_df[['Age', 'Gender']].reset_index(drop=True)], axis=1)
        val_tabular_df = pd.concat([val_shape_features_df, val_df[['Age', 'Gender']].reset_index(drop=True)], axis=1)
        test_tabular_df = pd.concat([test_shape_features_df, test_df[['Age', 'Gender']].reset_index(drop=True)], axis=1)

        TABULAR_COLS = train_tabular_df.columns.tolist()
        n_tabular_features = len(TABULAR_COLS)
        print(f"Total tabular features ({n_tabular_features}): {TABULAR_COLS}")

        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦¸à¦®à§à¦ªà§‚à¦°à§à¦£ à¦Ÿà§à¦¯à¦¾à¦¬à§à¦²à¦¾à¦° à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿ à¦¸à§à¦•à§‡à¦² à¦•à¦°à¦¾
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_tabular_df)
        val_features   = scaler.transform(val_tabular_df)
        test_features  = scaler.transform(test_tabular_df)
        
        
        # --- 7.3 Datasets ---
        # (à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: all_tabular_features à¦‡à¦¨à¦ªà§à¦Ÿ à¦¹à¦¿à¦¸à§‡à¦¬à§‡ à¦¨à§‡à¦“à§Ÿà¦¾
        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, all_tabular_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = all_tabular_features # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # 70/30 à¦“à§Ÿà§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿ (à¦†à¦ªà¦¨à¦¾à¦° à¦®à§‚à¦² à¦•à§‹à¦¡à§‡à¦° à¦®à¦¤à§‹)
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦¸à¦®à§à¦ªà§‚à¦°à§à¦£ à¦Ÿà§à¦¯à¦¾à¦¬à§à¦²à¦¾à¦° à¦«à¦¿à¦šà¦¾à¦° à¦²à§‹à¦¡ à¦•à¦°à¦¾
                tabular_feats = torch.tensor(self.features[idx], dtype=torch.float)
                
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, tabular_feats, label # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨

        minority_classes_list = [1] 
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ (à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. CrossEntropyLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        # DataLoader-à¦ sampler=train_sampler à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + Weighted CrossEntropyLoss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (*** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: timm SimCLR à¦¬à§‡à¦¸ + Age/Gender ***) ---
        
        class TimmSimCLRHybridClassifier(nn.Module):
            # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: n_tabular_features à¦‡à¦¨à¦ªà§à¦Ÿ
            def __init__(self, n_classes=3, n_tabular_features=7):
                super().__init__()
                try:
                    print("Loading timm model: resnet50_simclr_cifar10...")
                    self.cnn_base = timm.create_model("resnet50_simclr_cifar10", pretrained=True)
                    cnn_feature_dim = self.cnn_base.num_features  # 2048
                    self.cnn_base.reset_classifier(0) 
                    print(f"timm model loaded. Feature dimension: {cnn_feature_dim}")
                except Exception as e:
                    print(f"Error loading timm model: {e}. Falling back to standard ImageNet ResNet-50.")
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features  # 2048
                    self.cnn_base.fc = nn.Identity()

                # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: n_tabular_features à¦‡à¦¨à¦ªà§à¦Ÿ
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_tabular_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.3), 
                    nn.Linear(256, n_classes)
                )
                
            def forward(self, image, tabular_features): # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(tabular_features) # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train ---
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: n_tabular_features à¦ªà¦¾à¦¸ à¦•à¦°à¦¾
        cls_model = TimmSimCLRHybridClassifier(n_classes=3, n_tabular_features=n_tabular_features).to(device)
        
        criterion = nn.CrossEntropyLoss(weight=alpha_weights).to(device)
        
        # <--*** à¦ªà§à¦°à¦§à¦¾à¦¨ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Differential Learning Rate ***-->
        
        # à§§. à¦à¦¨à¦•à§‹à¦¡à¦¾à¦° (cnn_base) à¦à¦° à¦ªà§à¦¯à¦¾à¦°à¦¾à¦®à¦¿à¦Ÿà¦¾à¦°à¦—à§à¦²à§‹ à¦†à¦²à¦¾à¦¦à¦¾ à¦•à¦°à§à¦¨
        try:
            encoder_params = cls_model.cnn_base.parameters()
            
            # à§¨. à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ (tabular_mlp, fusion_classifier) à¦à¦° à¦ªà§à¦¯à¦¾à¦°à¦¾à¦®à¦¿à¦Ÿà¦¾à¦° à¦†à¦²à¦¾à¦¦à¦¾ à¦•à¦°à§à¦¨
            classifier_params = list(cls_model.tabular_mlp.parameters()) + list(cls_model.fusion_classifier.parameters())

            # à§©. à¦¦à§à¦Ÿà¦¿ à¦­à¦¿à¦¨à§à¦¨ à¦²à¦¾à¦°à§à¦¨à¦¿à¦‚ à¦°à§‡à¦Ÿ à¦¸à¦¹ à¦…à¦ªà§à¦Ÿà¦¿à¦®à¦¾à¦‡à¦œà¦¾à¦° à¦¸à§‡à¦Ÿà¦†à¦ª à¦•à¦°à§à¦¨
            optimizer = torch.optim.AdamW([
                {'params': encoder_params, 'lr': 1e-5},      # à¦à¦¨à¦•à§‹à¦¡à¦¾à¦°à§‡à¦° à¦œà¦¨à§à¦¯ à¦•à¦® LR (à¦«à¦¾à¦‡à¦¨-à¦Ÿà¦¿à¦‰à¦¨)
                {'params': classifier_params, 'lr': 1e-4}     # à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦°à§‡à¦° à¦œà¦¨à§à¦¯ à¦¬à§‡à¦¶à¦¿ LR (à¦¨à¦¤à§à¦¨ à¦•à¦°à§‡ à¦¶à§‡à¦–à¦¾)
            ], weight_decay=1e-5)
            
            print("\nUsing Differential Learning Rate (Encoder LR: 1e-5, Classifier LR: 1e-4).")

        except Exception as e:
            print(f"Could not set differential LR: {e}. Falling back to single LR.")
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=5e-5, weight_decay=1e-5) 

        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model (timm SimCLR Base + Age/Gender + Differential LR) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, tabular, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, tabular) # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, tabular, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                    outputs = cls_model(images, tabular) # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, tabular, labels in tqdm(test_cls_loader, desc="Testing"):
                images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                outputs = cls_model(images, tabular) # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (timm SimCLR + Age/Gender)") # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
# --- 0. à¦‡à¦®à§à¦ªà§‹à¦°à§à¦Ÿ ---
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import timm  # <-- SimCLR à¦®à¦¡à§‡à¦²à§‡à¦° à¦œà¦¨à§à¦¯
import warnings

# ÙˆØ§Ø±à¦¨à¦¿à¦‚ à¦‡à¦—à¦¨à§‹à¦° à¦•à¦°à¦¾
warnings.filterwarnings('ignore')

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE (Age/Gender à¦¸à¦¹) ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    
    # âœ… Radiomics + age features: Age à¦à¦¬à¦‚ Gender à¦²à§‹à¦¡ à¦•à¦°à¦¾
    metadata_df.dropna(subset=['radiological_evaluation', 'Age', 'Gender'], inplace=True)
    
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    
    # Gender à¦®à§à¦¯à¦¾à¦ª à¦•à¦°à¦¾
    metadata_df['Gender'] = metadata_df['Gender'].map({'Male': 0, 'Female': 1})
    
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Age, Gender à¦¸à¦¹ merge à¦•à¦°à¦¾
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤, à¦¸à¦‚à¦•à§à¦·à§‡à¦ªà§‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch): super().__init__(); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True), nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True));
    def forward(self, x): return self.conv(x)
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels): super().__init__(); self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False); self.bn1 = nn.BatchNorm2d(out_channels); self.relu = nn.ReLU(inplace=True); self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False); self.bn2 = nn.BatchNorm2d(out_channels);
    def forward(self, x, skip): x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True); x = torch.cat([x, skip], dim=1); x = self.relu(self.bn1(self.conv1(x))); x = self.relu(self.bn2(self.conv2(x))); return x
class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; self.decoder3 = DecoderBlock(2048, 1024, 512); self.decoder2 = DecoderBlock(512, 512, 256); self.decoder1 = DecoderBlock(256, 256, 128); self.decoder0 = DecoderBlock(128, 64, 64); self.final_conv = nn.Conv2d(64, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0); out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True); return self.final_conv(out)
class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12): super().__init__(); base_model = models.resnet50(weights=None); base_layers = list(base_model.children()); self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]; num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048; self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim)); self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim); layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True); self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth); self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim)); d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}; self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True); self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0']); self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0']); self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1']); self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3']); self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0']); self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); bs, _, h, w = e4.shape; trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding; trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w); x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1)); x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1)); x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1)); x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1)); x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1)); x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1)); return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int): super().__init__(); self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int)); self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int)); self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid()); self.relu = nn.ReLU(inplace=True);
    def forward(self, g, x): psi = self.relu(self.W_g(g) + self.W_x(x)); return x * self.psi(psi)
class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1): super().__init__(); base = models.resnet50(weights=None); layers = list(base.children()); self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5]); self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]; self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2); self.attn3 = AttentionGate(1024, 1024, 512); self.dec_conv3 = ConvBlock(2048, 1024); self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2); self.attn2 = AttentionGate(512, 512, 256); self.dec_conv2 = ConvBlock(1024, 512); self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2); self.attn1 = AttentionGate(256, 256, 128); self.dec_conv1 = ConvBlock(512, 256); self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2); self.attn0 = AttentionGate(64, 64, 32); self.dec_conv0 = ConvBlock(128, 64); self.final_up = nn.ConvTranspose2d(64, 32, 2, 2); self.final_conv = nn.Conv2d(32, n_classes, 1);
    def forward(self, x): e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3); d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1)); d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1)); d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1)); d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1)); return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦†à¦° à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦¤à¦¬à§‡ à¦°à¦¾à¦–à¦¾ à¦¹à¦²à§‹) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à¦Ÿà¦¿ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Tabular features (âœ… Radiomics + age features) ---
        print("\nEngineering tabular features (Shape + Age/Gender)...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Shape Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        # Radiomics/Shape à¦«à¦¿à¦šà¦¾à¦° à¦¤à§ˆà¦°à¦¿
        train_shape_features_df = get_shape_features(train_pred_masks)
        val_shape_features_df   = get_shape_features(val_pred_masks)
        test_shape_features_df  = get_shape_features(test_pred_masks)
        
        # Shape à¦«à¦¿à¦šà¦¾à¦° à¦à¦° à¦¸à¦¾à¦¥à§‡ Age à¦à¦¬à¦‚ Gender à¦¯à§à¦•à§à¦¤ à¦•à¦°à¦¾
        train_tabular_df = pd.concat([train_shape_features_df, train_df[['Age', 'Gender']].reset_index(drop=True)], axis=1)
        val_tabular_df = pd.concat([val_shape_features_df, val_df[['Age', 'Gender']].reset_index(drop=True)], axis=1)
        test_tabular_df = pd.concat([test_shape_features_df, test_df[['Age', 'Gender']].reset_index(drop=True)], axis=1)

        TABULAR_COLS = train_tabular_df.columns.tolist()
        n_tabular_features = len(TABULAR_COLS)
        print(f"Total tabular features ({n_tabular_features}): {TABULAR_COLS}")

        # à¦¸à¦®à§à¦ªà§‚à¦°à§à¦£ à¦Ÿà§à¦¯à¦¾à¦¬à§à¦²à¦¾à¦° à¦¡à§‡à¦Ÿà¦¾à¦¸à§‡à¦Ÿ à¦¸à§à¦•à§‡à¦² à¦•à¦°à¦¾
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_tabular_df)
        val_features   = scaler.transform(val_tabular_df)
        test_features  = scaler.transform(test_tabular_df)
        
        
        # --- 7.3 Datasets ---
        # (à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨ à¦…à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¿à¦¤)
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, all_tabular_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = all_tabular_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # 70/30 à¦“à§Ÿà§‡à¦Ÿà§‡à¦¡ à¦‡à¦¨à¦ªà§à¦Ÿ (à¦†à¦ªà¦¨à¦¾à¦° à¦®à§‚à¦² à¦•à§‹à¦¡à§‡à¦° à¦®à¦¤à§‹)
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                
                tabular_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label_tensor = torch.tensor(label, dtype=torch.long)
                return masked_image, tabular_feats, label_tensor

        minority_classes_list = [1] # 'Moderate' à¦•à§à¦²à¦¾à¦¸
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 Sampler and Loss (âœ… Weighted loss) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. CrossEntropyLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦šà¦¾à¦²à§ à¦°à¦¾à¦–à¦¾)
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        # DataLoader-à¦ sampler=train_sampler à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Sampler + Weighted CrossEntropyLoss). ğŸ“Š")

        
        # --- 7.4 Hybrid Classifier (âœ… SimCLR ResNet50 + âœ… Strong BN/Dropout Head) ---
        
        class TimmSimCLRHybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_tabular_features=7):
                super().__init__()
                try:
                    # âœ… SimCLR-pretrained ResNet50 via timm
                    print("Loading timm model: resnet50_simclr_cifar10...")
                    self.cnn_base = timm.create_model("resnet50_simclr_cifar10", pretrained=True)
                    cnn_feature_dim = self.cnn_base.num_features  # 2048
                    self.cnn_base.reset_classifier(0) # à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦° à¦¹à§‡à¦¡ à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾
                    print(f"timm model loaded. Feature dimension: {cnn_feature_dim}")
                except Exception as e:
                    print(f"Error loading timm model: {e}. Falling back to standard ImageNet ResNet-50.")
                    self.cnn_base = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
                    cnn_feature_dim = self.cnn_base.fc.in_features  # 2048
                    self.cnn_base.fc = nn.Identity()

                # âœ… Strong classifier head with BatchNorm + Dropout (Tabular part)
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_tabular_features, 64),
                    nn.BatchNorm1d(64),
                    nn.ReLU(),
                    nn.Dropout(0.3),
                    nn.Linear(64, 32),
                    nn.BatchNorm1d(32),
                    nn.ReLU(),
                    nn.Dropout(0.3)
                )
                
                # âœ… Strong classifier head with BatchNorm + Dropout (Fusion part)
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 512), # 2048 + 32 = 2080
                    nn.BatchNorm1d(512),
                    nn.ReLU(),
                    nn.Dropout(0.5), # à¦¶à¦•à§à¦¤à¦¿à¦¶à¦¾à¦²à§€ à¦¡à§à¦°à¦ªà¦†à¦‰à¦Ÿ
                    nn.Linear(512, n_classes)
                )
                
            def forward(self, image, tabular_features):
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(tabular_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        
        # --- 7.5 Train (âœ… Differential LR + âœ… Early stopping) ---
        
        cls_model = TimmSimCLRHybridClassifier(n_classes=3, n_tabular_features=n_tabular_features).to(device)
        
        criterion = nn.CrossEntropyLoss(weight=alpha_weights).to(device)
        
        # âœ… Differential learning rates
        try:
            encoder_params = cls_model.cnn_base.parameters()
            classifier_params = list(cls_model.tabular_mlp.parameters()) + list(cls_model.fusion_classifier.parameters())

            optimizer = torch.optim.AdamW([
                {'params': encoder_params, 'lr': 1e-5},      # à¦à¦¨à¦•à§‹à¦¡à¦¾à¦°à§‡à¦° à¦œà¦¨à§à¦¯ à¦•à¦® LR (à¦«à¦¾à¦‡à¦¨-à¦Ÿà¦¿à¦‰à¦¨)
                {'params': classifier_params, 'lr': 1e-4}     # à¦•à§à¦²à¦¾à¦¸à¦¿à¦«à¦¾à§Ÿà¦¾à¦°à§‡à¦° à¦œà¦¨à§à¦¯ à¦¬à§‡à¦¶à¦¿ LR (à¦¨à¦¤à§à¦¨ à¦•à¦°à§‡ à¦¶à§‡à¦–à¦¾)
            ], weight_decay=1e-5)
            
            print("\nUsing Differential Learning Rate (Encoder LR: 1e-5, Classifier LR: 1e-4).")
        except Exception as e:
            print(f"Could not set differential LR: {e}. Falling back to single LR.")
            optimizer = torch.optim.AdamW(cls_model.parameters(), lr=5e-5, weight_decay=1e-5) 

        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10 # âœ… Early stopping
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model (timm SimCLR + Age/Gender + BN Head + Diff. LR) Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, tabular, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, tabular)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, tabular, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                    outputs = cls_model(images, tabular)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            # âœ… Early stopping and best model saving
            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation (âœ… Per-class accuracy and CM) ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, tabular, labels in tqdm(test_cls_loader, desc="Testing"):
                images, tabular, labels = images.to(device), tabular.to(device), labels.to(device)
                outputs = cls_model(images, tabular)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Acccuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (timm SimCLR + Age/Gender)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset, SubsetRandomSampler
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
# --- FIX 1: Corrected import name ---
from torch.optim.lr_scheduler import ReduceLROnPlateau 
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import traceback

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
SIMCLR_CKPT = None # Not used in this version

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")
np.random.seed(2024)
torch.manual_seed(2024)
if torch.cuda.is_available():
    torch.cuda.manual_seed(2024)

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    class_weights = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODEL ARCHITECTURES ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0']*2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1']*2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0']*3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2']*2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1']*3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0']*4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean':
            return loss.mean()
        else:
            return loss.sum()

# =================================================================================
# --- DATA PREPARATION (COMMON FOR BOTH MODELS) ---
# =================================================================================

if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- Generate Segmentation Predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        # THIS DATASET IS USED BY BOTH MODELS
        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                
                # Resize mask to match image tensor size AFTER augmentation/resize
                pred_mask = self.masks[idx] # [1, 224, 224]
                # No resize needed if masks are already 224x224
                # pred_mask = F.interpolate(self.masks[idx].unsqueeze(0), size=IMAGE_SIZE, mode='nearest').squeeze(0)

                masked_image = img_tensor * pred_mask.clone().detach()
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                
                # Return: masked_image, shape_features, label
                return masked_image, shape_feats, label

        minority_classes_list = [2] # 'Severe'
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,   val_pred_masks,   val_features,   eval_transform)
        test_cls_dataset  = HybridDataset(test_df,  test_pred_masks,  test_features,  eval_transform)
        
        # --- OVERSAMPLER (Used by both models) ---
        labels_np = train_df['class_label'].to_numpy()
        class_to_indices = {c: np.where(labels_np == c)[0] for c in np.unique(labels_np)}
        target_count = max(len(ix) for ix in class_to_indices.values())
        print(f"\nOversampling: Target count per class: {target_count}")

        oversampled_indices = []
        rng = np.random.default_rng(2024)
        for c, ix in class_to_indices.items():
            if len(ix) == 0: continue
            need = target_count - len(ix)
            merged = np.concatenate([ix, rng.choice(ix, size=need, replace=True)]) if need > 0 else ix
            oversampled_indices.append(merged)
        
        oversampled_indices = np.concatenate(oversampled_indices); rng.shuffle(oversampled_indices)
        train_sampler = SubsetRandomSampler(oversampled_indices.tolist())

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  16, shuffle=False, num_workers=2, pin_memory=True)
        print("Hybrid datasets and loaders are ready with Oversampling. ğŸ“Š")

    except Exception as e:
        print(f"\nAn error occurred during data preparation: {e}")
        traceback.print_exc()
        models_loaded = False # Prevent further steps


# =================================================================================
# --- 7. MODEL A (HYBRID: 'Moderate Specialist') ---
# =================================================================================

if models_loaded:
    print("\n\n--- Starting Section 7: Training Model A (Hybrid) ---")
    
    # --- 7.4 HYBRID MODEL ARCHITECTURE ---
    # Using the ResNet50-based backbone from your code
    class SimCLRPPUNetBackbone(nn.Module):
        def __init__(self):
            super().__init__()
            base_model = models.resnet50(weights=models.ResNet50_Weights.DEFAULT) # Use pretrained
            layers = list(base_model.children())
            self.encoder0 = nn.Sequential(*layers[:3])
            self.encoder1 = nn.Sequential(*layers[3:5])
            self.encoder2 = layers[5]
            self.encoder3 = layers[6]
            self.encoder4 = layers[7]
            self.gap = nn.AdaptiveAvgPool2d(1)
            self.feat_dim = 2048
            self.out_dim = self.feat_dim
        
        def forward(self, x):
            e0=self.encoder0(x); e1=self.encoder1(e0); e2=self.encoder2(e1); e3=self.encoder3(e2); e4=self.encoder4(e3)
            h = self.gap(e4).flatten(1)
            return h

    class HybridClassifier(nn.Module):
        def __init__(self, n_classes=3, n_shape_features=5):
            super().__init__()
            self.backbone = SimCLRPPUNetBackbone()
            img_feat_dim = self.backbone.out_dim
            self.tabular_mlp = nn.Sequential(
                nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
            )
            self.fusion_classifier = nn.Sequential(
                nn.Linear(img_feat_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                nn.Linear(256, n_classes)
            )
        def forward(self, image, shape_features):
            img_features = self.backbone(image)
            tab_feats    = self.tabular_mlp(shape_features)
            fused        = torch.cat([img_features, tab_feats], dim=1)
            logits       = self.fusion_classifier(fused)
            return logits
    
    # --- 7.5 TRAIN HYBRID MODEL ---
    cls_model_A = HybridClassifier(
        n_classes=3,
        n_shape_features=train_features.shape[1]
    ).to(device)

    criterion = FocalLoss(alpha=class_weights, gamma=3.0)
    optimizer = torch.optim.AdamW(cls_model_A.parameters(), lr=1e-4)
    # --- FIX 2: Corrected scheduler name ---
    scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
    num_epochs = 50
    patience = 10
    epochs_no_improve = 0
    best_val_accuracy = 0.0
    
    print("\n--- Starting Model A (Hybrid) Training ---")
    for epoch in range(num_epochs):
        cls_model_A.train()
        running_loss = 0.0
        for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
            images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = cls_model_A(images, shapes)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        cls_model_A.eval()
        val_corrects = 0
        with torch.no_grad():
            for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model_A(images, shapes)
                _, preds = torch.max(outputs, 1)
                val_corrects += torch.sum(preds == labels.data)

        val_accuracy = val_corrects.double() / len(val_cls_dataset)
        print(f"Epoch {epoch + 1} | Loss: {running_loss / len(train_cls_loader):.4f} | Val Acc: {val_accuracy:.4f}")
        scheduler.step(val_accuracy)

        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save(cls_model_A.state_dict(), 'best_hybrid_model.pth') # Save Model A
            epochs_no_improve = 0
            print(f"    ->âœ… New best Model A saved: {best_val_accuracy:.4f}")
        else:
            epochs_no_improve += 1
            print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
        
        if epochs_no_improve >= patience:
            print(f"\nEarly stopping triggered after {patience} epochs.")
            break

    print("\n--- Model A (Hybrid) Training Finished ---")

# =================================================================================
# --- 8. MODEL B (ResNet50+GAP: 'Mild/Severe Specialist') ---
# =================================================================================

if models_loaded:
    print("\n\n--- Starting Section 8: Training Model B (ResNet50+GAP) ---")
    
    # --- 8.1 ResNet50+GAP MODEL ARCHITECTURE ---
    class ResNet50_GAP(nn.Module):
        def __init__(self, n_classes=3):
            super().__init__()
            # Load pretrained ResNet50
            self.base_model = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
            
            # Get the number of input features for the classifier
            num_ftrs = self.base_model.fc.in_features
            
            # Replace the final layers
            self.base_model.fc = nn.Identity() # Remove the original FC layer
            
            # Add a new classifier head
            self.classifier = nn.Sequential(
                nn.Dropout(0.5),
                nn.Linear(num_ftrs, n_classes)
            )
            
        def forward(self, x):
            # The base_model's forward pass ends with AdaptiveAvgPool2d
            features = self.base_model(x)
            output = self.classifier(features)
            return output
            
    # --- 8.2 TRAIN ResNet50+GAP MODEL ---
    cls_model_B = ResNet50_GAP(n_classes=3).to(device)

    criterion_B = FocalLoss(alpha=class_weights, gamma=3.0)
    optimizer_B = torch.optim.AdamW(cls_model_B.parameters(), lr=1e-4)
    # --- FIX 3: Corrected scheduler name ---
    scheduler_B = ReduceLROnPlateau(optimizer_B, mode='max', factor=0.2, patience=5, verbose=True)
    num_epochs_B = 50
    patience_B = 10
    epochs_no_improve_B = 0
    best_val_accuracy_B = 0.0
    
    print("\n--- Starting Model B (GAP) Training ---")
    for epoch in range(num_epochs_B):
        cls_model_B.train()
        running_loss_B = 0.0
        
        # NOTE: We get (images, shapes, labels) but only use (images, labels)
        for images, _, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs_B} (Train)"):
            images, labels = images.to(device), labels.to(device) # Ignore 'shapes'
            
            optimizer_B.zero_grad()
            outputs = cls_model_B(images) # Only pass images
            loss = criterion_B(outputs, labels)
            loss.backward()
            optimizer_B.step()
            running_loss_B += loss.item()

        cls_model_B.eval()
        val_corrects_B = 0
        with torch.no_grad():
            # NOTE: We get (images, shapes, labels) but only use (images, labels)
            for images, _, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs_B} (Validation)"):
                images, labels = images.to(device), labels.to(device) # Ignore 'shapes'
                
                outputs = cls_model_B(images) # Only pass images
                _, preds = torch.max(outputs, 1)
                val_corrects_B += torch.sum(preds == labels.data)

        val_accuracy_B = val_corrects_B.double() / len(val_cls_dataset)
        print(f"Epoch {epoch + 1} | Loss: {running_loss_B / len(train_cls_loader):.4f} | Val Acc: {val_accuracy_B:.4f}")
        scheduler_B.step(val_accuracy_B)

        if val_accuracy_B > best_val_accuracy_B:
            best_val_accuracy_B = val_accuracy_B
            torch.save(cls_model_B.state_dict(), 'best_gap_model.pth') # Save Model B
            epochs_no_improve_B = 0
            print(f"    ->âœ… New best Model B saved: {best_val_accuracy_B:.4f}")
        else:
            epochs_no_improve_B += 1
            print(f"    ->âŒ No improvement for {epochs_no_improve_B} epochs.")
        
        if epochs_no_improve_B >= patience_B:
            print(f"\nEarly stopping triggered after {patience_B} epochs.")
            break

    print("\n--- Model B (GAP) Training Finished ---")


# =================================================================================
# --- 9. FINAL ENSEMBLE EVALUATION ---
# =================================================================================

if models_loaded:
    print("\n\n--- Starting Section 9: Final Ensemble Evaluation ---")
    
    try:
        # --- 9.1 Load Both Trained Models ---
        print("Loading trained specialist models...")
        
        # Load Model A (Hybrid) - "Moderate Specialist"
        cls_model_A = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        cls_model_A.load_state_dict(torch.load('best_hybrid_model.pth'))
        cls_model_A.eval()
        print("Model A (Hybrid) loaded.")
        
        # Load Model B (GAP) - "Mild/Severe Specialist"
        cls_model_B = ResNet50_GAP(n_classes=3).to(device)
        cls_model_B.load_state_dict(torch.load('best_gap_model.pth'))
        cls_model_B.eval()
        print("Model B (GAP) loaded.")

        # --- 9.2 Get Predictions from Both Models ---
        all_labels_true = []
        all_preds_A = [] # Predictions from Model A (Hybrid)
        all_preds_B = [] # Predictions from Model B (GAP)

        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Ensemble Testing"):
                images = images.to(device)
                shapes = shapes.to(device)
                labels = labels.to(device)

                # Model A (Hybrid) prediction
                outputs_A = cls_model_A(images, shapes)
                _, preds_A = torch.max(outputs_A, 1)
                
                # Model B (GAP) prediction
                outputs_B = cls_model_B(images) # Only needs image
                _, preds_B = torch.max(outputs_B, 1)

                all_labels_true.extend(labels.cpu().numpy())
                all_preds_A.extend(preds_A.cpu().numpy())
                all_preds_B.extend(preds_B.cpu().numpy())
        
        # Convert to NumPy arrays for easy logic
        labels_true_np = np.array(all_labels_true)
        preds_A_np = np.array(all_preds_A) # Hybrid (Moderate Specialist)
        preds_B_np = np.array(all_preds_B) # GAP (Mild/Severe Specialist)

        # --- 9.3 Apply Your Ensemble Logic ---
        # Logic: Use Model B (GAP) for Mild/Severe, but use Model A (Hybrid) for Moderate.
        
        print("\nApplying ensemble logic...")
        
        # Start with the predictions from Model B (Mild/Severe Specialist)
        final_preds = np.copy(preds_B_np)
        
        # Find all samples where Model A (Moderate Specialist) predicted 'Moderate' (label 1)
        moderate_indices = (preds_A_np == 1)
        
        # Wherever Model A predicted 'Moderate', trust it.
        # Overwrite the final predictions with '1' in those spots.
        final_preds[moderate_indices] = 1
        
        print("Ensemble logic applied.")

        # --- 9.4 Final Report ---
        class_names = ['Mild', 'Moderate', 'Severe']
        
        print("\n--- (Reference) Model A (Hybrid) Report ---")
        print(classification_report(labels_true_np, preds_A_np, target_names=class_names, zero_division=0))
        
        print("\n--- (Reference) Model B (GAP) Report ---")
        print(classification_report(labels_true_np, preds_B_np, target_names=class_names, zero_division=0))
        
        print("\n--- âœ… FINAL ENSEMBLE REPORT ---")
        print(classification_report(labels_true_np, final_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Final Ensemble Confusion Matrix ---")
        cm = confusion_matrix(labels_true_np, final_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Final Ensemble Confusion Matrix")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred during final evaluation: {e}")
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset, SubsetRandomSampler
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau 
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler
import traceback

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- NEW: Set the path to your SimCLR pretrained model ---
# --- âš ï¸ YOU MUST CHANGE THIS PATH ---
SIMCLR_CKPT_PATH = '/kaggle/input/your-simclr-checkpoint/simclr_model.pth' 


# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")
np.random.seed(2024)
torch.manual_seed(2024)
if torch.cuda.is_available():
    torch.cuda.manual_seed(2024)

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    class_weights = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODEL ARCHITECTURES ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0']*2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1']*2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0']*3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2']*2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1']*3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0']*4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean':
            return loss.mean()
        else:
            return loss.sum()

# =================================================================================
# --- DATA PREPARATION (COMMON FOR BOTH MODELS) ---
# =================================================================================

if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- Generate Segmentation Predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        # THIS DATASET IS USED BY BOTH MODELS
        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                
                pred_mask = self.masks[idx] # [1, 224, 224]
                masked_image = img_tensor * pred_mask.clone().detach()
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                
                # Return: masked_image, shape_features, label
                return masked_image, shape_feats, label

        minority_classes_list = [2] # 'Severe'
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,   val_pred_masks,   val_features,   eval_transform)
        test_cls_dataset  = HybridDataset(test_df,  test_pred_masks,  test_features,  eval_transform)
        
        # --- OVERSAMPLER (Used by both models) ---
        labels_np = train_df['class_label'].to_numpy()
        class_to_indices = {c: np.where(labels_np == c)[0] for c in np.unique(labels_np)}
        target_count = max(len(ix) for ix in class_to_indices.values())
        print(f"\nOversampling: Target count per class: {target_count}")

        oversampled_indices = []
        rng = np.random.default_rng(2024)
        for c, ix in class_to_indices.items():
            if len(ix) == 0: continue
            need = target_count - len(ix)
            merged = np.concatenate([ix, rng.choice(ix, size=need, replace=True)]) if need > 0 else ix
            oversampled_indices.append(merged)
        
        oversampled_indices = np.concatenate(oversampled_indices); rng.shuffle(oversampled_indices)
        train_sampler = SubsetRandomSampler(oversampled_indices.tolist())

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  16, shuffle=False, num_workers=2, pin_memory=True)
        print("Hybrid datasets and loaders are ready with Oversampling. ğŸ“Š")

    except Exception as e:
        print(f"\nAn error occurred during data preparation: {e}")
        traceback.print_exc()
        models_loaded = False # Prevent further steps


# =================================================================================
# --- 7. MODEL A (HYBRID: 'Moderate Specialist') ---
# =================================================================================

if models_loaded:
    print("\n\n--- Starting Section 7: Training Model A (Hybrid) ---")
    
    # --- 7.4 HYBRID MODEL ARCHITECTURE ---
    class SimCLRPPUNetBackbone(nn.Module):
        def __init__(self):
            super().__init__()
            # Model A uses ImageNet pretraining
            base_model = models.resnet50(weights=models.ResNet50_Weights.DEFAULT) 
            layers = list(base_model.children())
            self.encoder0 = nn.Sequential(*layers[:3])
            self.encoder1 = nn.Sequential(*layers[3:5])
            self.encoder2 = layers[5]
            self.encoder3 = layers[6]
            self.encoder4 = layers[7]
            self.gap = nn.AdaptiveAvgPool2d(1)
            self.feat_dim = 2048
            self.out_dim = self.feat_dim
        
        def forward(self, x):
            e0=self.encoder0(x); e1=self.encoder1(e0); e2=self.encoder2(e1); e3=self.encoder3(e2); e4=self.encoder4(e3)
            h = self.gap(e4).flatten(1)
            return h

    class HybridClassifier(nn.Module):
        def __init__(self, n_classes=3, n_shape_features=5):
            super().__init__()
            self.backbone = SimCLRPPUNetBackbone()
            img_feat_dim = self.backbone.out_dim
            self.tabular_mlp = nn.Sequential(
                nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
            )
            self.fusion_classifier = nn.Sequential(
                nn.Linear(img_feat_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                nn.Linear(256, n_classes)
            )
        def forward(self, image, shape_features):
            img_features = self.backbone(image)
            tab_feats    = self.tabular_mlp(shape_features)
            fused        = torch.cat([img_features, tab_feats], dim=1)
            logits       = self.fusion_classifier(fused)
            return logits
    
    # --- 7.5 TRAIN HYBRID MODEL ---
    cls_model_A = HybridClassifier(
        n_classes=3,
        n_shape_features=train_features.shape[1]
    ).to(device)

    criterion = FocalLoss(alpha=class_weights, gamma=3.0)
    optimizer = torch.optim.AdamW(cls_model_A.parameters(), lr=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
    num_epochs = 50
    patience = 10
    epochs_no_improve = 0
    best_val_accuracy = 0.0
    
    print("\n--- Starting Model A (Hybrid) Training ---")
    for epoch in range(num_epochs):
        cls_model_A.train()
        running_loss = 0.0
        for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
            images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = cls_model_A(images, shapes)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        cls_model_A.eval()
        val_corrects = 0
        with torch.no_grad():
            for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model_A(images, shapes)
                _, preds = torch.max(outputs, 1)
                val_corrects += torch.sum(preds == labels.data)

        val_accuracy = val_corrects.double() / len(val_cls_dataset)
        print(f"Epoch {epoch + 1} | Loss: {running_loss / len(train_cls_loader):.4f} | Val Acc: {val_accuracy:.4f}")
        scheduler.step(val_accuracy)

        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save(cls_model_A.state_dict(), 'best_hybrid_model.pth') # Save Model A
            epochs_no_improve = 0
            print(f"    ->âœ… New best Model A saved: {best_val_accuracy:.4f}")
        else:
            epochs_no_improve += 1
            print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
        
        if epochs_no_improve >= patience:
            print(f"\nEarly stopping triggered after {patience} epochs.")
            break

    print("\n--- Model A (Hybrid) Training Finished ---")

# =================================================================================
# --- 8. MODEL B (SimCLR + GAP: 'Mild/Severe Specialist') ---
# =================================================================================

if models_loaded:
    print("\n\n--- Starting Section 8: Training Model B (SimCLR+GAP) ---")
    
    # --- 8.1 NEW: ResNet50+GAP+SimCLR MODEL ARCHITECTURE ---
    class ResNet50_GAP_SimCLR(nn.Module):
        def __init__(self, n_classes=3, simclr_path=None):
            super().__init__()
            # 1. Initialize with NO weights
            self.base_model = models.resnet50(weights=None) 
            
            # 2. Load SimCLR weights if path is provided
            if simclr_path and os.path.exists(simclr_path):
                print(f"Loading SimCLR weights from: {simclr_path}")
                try:
                    state_dict = torch.load(simclr_path, map_location=device)
                    
                    # --- Clean checkpoint keys ---
                    # This logic may need to change depending on how your .pth file was saved
                    cleaned_state_dict = {}
                    for k, v in state_dict.items():
                        if k.startswith('backbone.'):
                            k = k[len('backbone.'):] # Strip 'backbone.'
                        elif k.startswith('module.'):
                            k = k[len('module.'):] # Strip 'module.'
                        
                        # Ignore the projection head or classifier from SimCLR
                        if 'fc.' not in k and 'projection_head.' not in k and 'classifier.' not in k:
                            cleaned_state_dict[k] = v
                    
                    # Load the weights into the base model
                    msg = self.base_model.load_state_dict(cleaned_state_dict, strict=False)
                    print(f"SimCLR weights loaded. Missing keys: {msg.missing_keys}")
                    if 'fc.weight' in msg.missing_keys or 'fc.bias' in msg.missing_keys:
                        print("    (Ignoring missing 'fc' keys is expected)")
                    
                except Exception as e:
                    print(f"Warning: Failed to load SimCLR weights: {e}. Using random init.")
            else:
                print("Warning: No SimCLR path provided or path invalid. Using random init for Model B.")

            # 3. Replace the classifier head (same as before)
            num_ftrs = self.base_model.fc.in_features
            self.base_model.fc = nn.Identity()
            self.classifier = nn.Sequential(
                nn.Dropout(0.5),
                nn.Linear(num_ftrs, n_classes)
            )
            
        def forward(self, x):
            features = self.base_model(x)
            output = self.classifier(features)
            return output
            
    # --- 8.2 TRAIN SimCLR+GAP MODEL ---
    # --- NEW: Initialize with SimCLR path ---
    cls_model_B = ResNet50_GAP_SimCLR(
        n_classes=3, 
        simclr_path=SIMCLR_CKPT_PATH  # Pass the path here
    ).to(device)

    criterion_B = FocalLoss(alpha=class_weights, gamma=3.0)
    optimizer_B = torch.optim.AdamW(cls_model_B.parameters(), lr=1e-4)
    scheduler_B = ReduceLROnPlateau(optimizer_B, mode='max', factor=0.2, patience=5, verbose=True)
    num_epochs_B = 50
    patience_B = 10
    epochs_no_improve_B = 0
    best_val_accuracy_B = 0.0
    
    print("\n--- Starting Model B (SimCLR+GAP) Training ---")
    for epoch in range(num_epochs_B):
        cls_model_B.train()
        running_loss_B = 0.0
        
        # NOTE: We get (images, shapes, labels) but only use (images, labels)
        for images, _, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs_B} (Train)"):
            images, labels = images.to(device), labels.to(device) # Ignore 'shapes'
            
            optimizer_B.zero_grad()
            outputs = cls_model_B(images) # Only pass images
            loss = criterion_B(outputs, labels)
            loss.backward()
            optimizer_B.step()
            running_loss_B += loss.item()

        cls_model_B.eval()
        val_corrects_B = 0
        with torch.no_grad():
            # NOTE: We get (images, shapes, labels) but only use (images, labels)
            for images, _, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs_B} (Validation)"):
                images, labels = images.to(device), labels.to(device) # Ignore 'shapes'
                
                outputs = cls_model_B(images) # Only pass images
                _, preds = torch.max(outputs, 1)
                val_corrects_B += torch.sum(preds == labels.data)

        val_accuracy_B = val_corrects_B.double() / len(val_cls_dataset)
        print(f"Epoch {epoch + 1} | Loss: {running_loss_B / len(train_cls_loader):.4f} | Val Acc: {val_accuracy_B:.4f}")
        scheduler_B.step(val_accuracy_B)

        if val_accuracy_B > best_val_accuracy_B:
            best_val_accuracy_B = val_accuracy_B
            torch.save(cls_model_B.state_dict(), 'best_gap_model.pth') # Save Model B
            epochs_no_improve_B = 0
            print(f"    ->âœ… New best Model B saved: {best_val_accuracy_B:.4f}")
        else:
            epochs_no_improve_B += 1
            print(f"    ->âŒ No improvement for {epochs_no_improve_B} epochs.")
        
        if epochs_no_improve_B >= patience_B:
            print(f"\nEarly stopping triggered after {patience_B} epochs.")
            break

    print("\n--- Model B (SimCLR+GAP) Training Finished ---")


# =================================================================================
# --- 9. FINAL ENSEMBLE EVALUATION ---
# =================================================================================

if models_loaded:
    print("\n\n--- Starting Section 9: Final Ensemble Evaluation ---")
    
    try:
        # --- 9.1 Load Both Trained Models ---
        print("Loading trained specialist models...")
        
        # Load Model A (Hybrid) - "Moderate Specialist"
        cls_model_A = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        cls_model_A.load_state_dict(torch.load('best_hybrid_model.pth'))
        cls_model_A.eval()
        print("Model A (Hybrid) loaded.")
        
        # --- NEW: Load Model B (SimCLR+GAP) - "Mild/Severe Specialist" ---
        cls_model_B = ResNet50_GAP_SimCLR(n_classes=3).to(device) # Initialize with new class
        cls_model_B.load_state_dict(torch.load('best_gap_model.pth'))
        cls_model_B.eval()
        print("Model B (SimCLR+GAP) loaded.")

        # --- 9.2 Get Predictions from Both Models ---
        all_labels_true = []
        all_preds_A = [] # Predictions from Model A (Hybrid)
        all_preds_B = [] # Predictions from Model B (GAP)

        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Ensemble Testing"):
                images = images.to(device)
                shapes = shapes.to(device)
                labels = labels.to(device)

                # Model A (Hybrid) prediction
                outputs_A = cls_model_A(images, shapes)
                _, preds_A = torch.max(outputs_A, 1)
                
                # Model B (GAP) prediction
                outputs_B = cls_model_B(images) # Only needs image
                _, preds_B = torch.max(outputs_B, 1)

                all_labels_true.extend(labels.cpu().numpy())
                all_preds_A.extend(preds_A.cpu().numpy())
                all_preds_B.extend(preds_B.cpu().numpy())
        
        # Convert to NumPy arrays for easy logic
        labels_true_np = np.array(all_labels_true)
        preds_A_np = np.array(all_preds_A) # Hybrid (Moderate Specialist)
        preds_B_np = np.array(all_preds_B) # GAP (Mild/Severe Specialist)

        # --- 9.3 Apply Your Ensemble Logic ---
        # Logic: Use Model B (GAP) for Mild/Severe, but use Model A (Hybrid) for Moderate.
        
        print("\nApplying ensemble logic...")
        
        # Start with the predictions from Model B (Mild/Severe Specialist)
        final_preds = np.copy(preds_B_np)
        
        # Find all samples where Model A (Moderate Specialist) predicted 'Moderate' (label 1)
        moderate_indices = (preds_A_np == 1)
        
        # Wherever Model A predicted 'Moderate', trust it.
        # Overwrite the final predictions with '1' in those spots.
        final_preds[moderate_indices] = 1
        
        print("Ensemble logic applied.")

        # --- 9.4 Final Report ---
        class_names = ['Mild', 'Moderate', 'Severe']
        
        print("\n--- (Reference) Model A (Hybrid / ImageNet) Report ---")
        print(classification_report(labels_true_np, preds_A_np, target_names=class_names, zero_division=0))
        
        print("\n--- (Reference) Model B (SimCLR+GAP) Report ---")
        print(classification_report(labels_true_np, preds_B_np, target_names=class_names, zero_division=0))
        
        print("\n--- âœ… FINAL ENSEMBLE REPORT ---")
        print(classification_report(labels_true_np, final_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Final Ensemble Confusion Matrix ---")
        cm = confusion_matrix(labels_true_np, final_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Final Ensemble Confusion Matrix")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred during final evaluation: {e}")
        traceback.print_exc()
```

```python
# file: improved_hybrid_classifier.py
# Minimal inline comments; comments explain WHY for important choices.

import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import cv2
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import classification_report, confusion_matrix
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.utils.data import WeightedRandomSampler
from torchvision import models
import albumentations as A
from albumentations.pytorch import ToTensorV2
import matplotlib.pyplot as plt

# ---------------- CONFIG ----------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 16
NUM_EPOCHS = 50
OVERSAMPLE_FACTOR = 3     # how many times to oversample Moderate
LR = 1e-4
WEIGHT_DECAY = 1e-5
PATIENCE = 10             # early stopping on Moderate recall
MODEL_SAVE_PATH = "best_effnet_hybrid.pth"

# ===== Assumption: these variables are already prepared by your preprocessing pipeline:
# train_df, val_df, test_df  -> pandas DataFrames with 'image_file_path' and 'class_label' columns
# train_pred_masks, val_pred_masks, test_pred_masks -> torch tensors shape (N,1,H,W) with binary masks
# train_features, val_features, test_features -> numpy arrays (N, num_shape_features)
# If you don't have them in your environment, compute them first as in your pipeline.
# --------------------------------------------------

# ---------- helper check ----------
def ensure_vars_exist(*names):
    missing = [n for n in names if n not in globals()]
    if missing:
        raise RuntimeError(f"Missing required precomputed variables in global scope: {missing}")

ensure_vars_exist(
    'train_df', 'val_df', 'test_df',
    'train_pred_masks', 'val_pred_masks', 'test_pred_masks',
    'train_features', 'val_features', 'test_features'
)

# --------------- SCALE FEATURES ---------------
scaler = RobustScaler()
train_features_scaled = scaler.fit_transform(train_features)
val_features_scaled   = scaler.transform(val_features)
test_features_scaled  = scaler.transform(test_features)

# --------------- AUGMENTATIONS ---------------
base_aug = A.Compose([
    A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
    A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2()
])

strong_aug = A.Compose([
    A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
    A.HorizontalFlip(p=0.9),
    A.VerticalFlip(p=0.6),
    A.ShiftScaleRotate(shift_limit=0.12, scale_limit=0.12, rotate_limit=25, p=0.9),
    A.RandomBrightnessContrast(p=0.6),
    A.CLAHE(p=0.4),
    A.GaussianBlur(p=0.4),
    A.CoarseDropout(max_holes=20, max_height=20, max_width=20, p=0.35),
    A.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225]),
    ToTensorV2()
])

# --------------- DATASET ---------------
class HybridDataset(Dataset):
    def __init__(self, df, pred_masks, shape_feats, transform, minority_classes=None, minority_transform=None, use_mask=True):
        self.df = df.reset_index(drop=True)
        self.masks = pred_masks
        self.shape_feats = shape_feats
        self.transform = transform
        self.minority_classes = minority_classes or []
        self.minority_transform = minority_transform or transform
        self.use_mask = use_mask
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = row['image_file_path']
        img = cv2.imread(img_path)
        if img is None:
            # fallback to zeros if file missing
            img = np.zeros((IMAGE_SIZE[0], IMAGE_SIZE[1], 3), dtype=np.uint8)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        label = int(row['class_label'])
        transform = self.minority_transform if label in self.minority_classes else self.transform
        augmented = transform(image=img)
        img_t = augmented['image']  # tensor
        if self.use_mask:
            mask = self.masks[idx]  # cpu tensor (1,H,W)
            if mask.shape[1:] != (IMAGE_SIZE[0], IMAGE_SIZE[1]):
                # resize mask if necessary
                mask_np = mask.squeeze(0).numpy().astype(np.uint8)
                mask_np = cv2.resize(mask_np, (IMAGE_SIZE[1], IMAGE_SIZE[0]), interpolation=cv2.INTER_NEAREST)
                mask = torch.from_numpy(mask_np).unsqueeze(0).float()
            # broadcast to 3 channels and combine (light emphasis on mask regions)
            mask3 = mask.repeat(3,1,1).to(img_t.device)
            img_t = 0.75 * img_t + 0.25 * (img_t * mask3)
        shape_feat = torch.tensor(self.shape_feats[idx], dtype=torch.float32)
        label_t = torch.tensor(label, dtype=torch.long)
        return img_t, shape_feat, label_t

# instantiate datasets
minority_classes = [1]  # Moderate label
train_ds = HybridDataset(train_df, train_pred_masks, train_features_scaled, base_aug, minority_classes, strong_aug, use_mask=True)
val_ds   = HybridDataset(val_df, val_pred_masks, val_features_scaled, base_aug, use_mask=True)
test_ds  = HybridDataset(test_df, test_pred_masks, test_features_scaled, base_aug, use_mask=True)

# --------------- SAMPLER (oversample Moderate) ---------------
labels_np = train_df['class_label'].to_numpy()
class_counts = np.bincount(labels_np)
class_counts = np.where(class_counts == 0, 1, class_counts)
print("Train class counts (0=Mild,1=Moderate,2=Severe):", class_counts)

# compute sample weights: baseline inverse freq, then multiply moderate by factor
inv_freq = (1.0 / class_counts)
sample_weights = np.array([inv_freq[l] * (OVERSAMPLE_FACTOR if l == 1 else 1.0) for l in labels_np], dtype=np.float32)
sample_weights = torch.from_numpy(sample_weights)
num_samples = int(len(sample_weights) * 1.5)  # train more samples per epoch
train_sampler = WeightedRandomSampler(sample_weights, num_samples=num_samples, replacement=True)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, sampler=train_sampler, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# --------------- MODEL (EfficientNet + SE + Attention Fusion) ---------------
class SEBlock(nn.Module):
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.gap = nn.AdaptiveAvgPool1d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, channels // reduction),
            nn.ReLU(inplace=True),
            nn.Linear(channels // reduction, channels),
            nn.Sigmoid()
        )
    def forward(self, x):
        # x: (B, C)
        # apply gating on feature vector
        if x.dim() == 2:
            s = self.fc(x)
            return x * s
        return x

class AttentionFusion(nn.Module):
    def __init__(self, dim_img, dim_tab, hidden=128):
        super().__init__()
        self.attn = nn.Sequential(
            nn.Linear(dim_img + dim_tab, hidden),
            nn.ReLU(),
            nn.Linear(hidden, dim_img + dim_tab),
            nn.Sigmoid()
        )
    def forward(self, img_feat, tab_feat):
        combined = torch.cat([img_feat, tab_feat], dim=1)
        alpha = self.attn(combined)
        return combined * alpha

class EffNetHybrid(nn.Module):
    def __init__(self, n_classes=3, n_tab=5, eff_version='b0'):
        super().__init__()
        # try to load EfficientNet pretrained if possible
        try:
            if eff_version == 'b0':
                self.backbone = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
            elif eff_version == 'b2':
                self.backbone = models.efficientnet_b2(weights=models.EfficientNet_B2_Weights.IMAGENET1K_V1)
            else:
                self.backbone = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
        except Exception:
            # fallback: default constructor without weights
            try:
                self.backbone = models.efficientnet_b0(weights=None)
            except Exception:
                # as last resort, use densenet121
                self.backbone = models.densenet121(weights=None)
        # get feature dim
        feat_dim = None
        if hasattr(self.backbone, 'classifier'):
            try:
                feat_dim = self.backbone.classifier.in_features
                self.backbone.classifier = nn.Identity()
            except Exception:
                pass
        if feat_dim is None:
            # fallback for models like densenet
            feat_dim = 1024
            self.backbone.classifier = nn.Identity()

        self.se = SEBlock(feat_dim, reduction=16)
        self.tabular = nn.Sequential(
            nn.Linear(n_tab, 64), nn.BatchNorm1d(64), nn.ReLU(),
            nn.Dropout(0.2), nn.Linear(64, 32), nn.ReLU()
        )
        self.fusion = AttentionFusion(feat_dim, 32, hidden=128)
        self.head = nn.Sequential(
            nn.Linear(feat_dim + 32, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.35),
            nn.Linear(256, n_classes)
        )

    def forward(self, img, tab):
        img_feat = self.backbone(img)           # (B, feat_dim)
        img_feat = self.se(img_feat)            # calibrate channels/features
        tab_feat = self.tabular(tab)
        fused = self.fusion(img_feat, tab_feat)
        out = self.head(fused)
        return out

# instantiate model
n_shape = train_features_scaled.shape[1]
model = EffNetHybrid(n_classes=3, n_tab=n_shape, eff_version='b0').to(DEVICE)

# --------------- CLASS-BALANCED FOCAL LOSS ---------------
class CBFocalLoss(nn.Module):
    def __init__(self, samples_per_cls, num_classes, beta=0.999, gamma=2.0, device=DEVICE):
        super().__init__()
        effective_num = 1.0 - np.power(beta, samples_per_cls)
        weights = (1.0 - beta) / (effective_num + 1e-12)
        weights = weights / np.sum(weights) * num_classes
        self.weights = torch.tensor(weights, dtype=torch.float32).to(device)
        self.gamma = gamma
    def forward(self, logits, targets):
        ce = F.cross_entropy(logits, targets, reduction='none', weight=self.weights)
        pt = torch.exp(-ce)
        loss = ((1 - pt) ** self.gamma) * ce
        return loss.mean()

cbf_loss = CBFocalLoss(class_counts, num_classes=3, beta=0.999, gamma=2.0, device=DEVICE)

# --------------- OPTIMIZER & SCHEDULER ---------------
optimizer = torch.optim.AdamW(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
from torch.optim.lr_scheduler import ReduceLROnPlateau
scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=4, verbose=True)

# --------------- TRAIN / VAL LOOP ---------------
best_moderate = 0.0
no_improve = 0

for epoch in range(NUM_EPOCHS):
    model.train()
    running_loss = 0.0
    for imgs, tabs, labs in tqdm(train_loader, desc=f"Train Epoch {epoch+1}/{NUM_EPOCHS}"):
        imgs = imgs.to(DEVICE); tabs = tabs.to(DEVICE); labs = labs.to(DEVICE)
        optimizer.zero_grad()
        logits = model(imgs, tabs)
        loss = cbf_loss(logits, labs)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    avg_loss = running_loss / max(1, len(train_loader))

    # validation
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for imgs, tabs, labs in val_loader:
            imgs = imgs.to(DEVICE); tabs = tabs.to(DEVICE); labs = labs.to(DEVICE)
            logits = model(imgs, tabs)
            preds = torch.argmax(logits, dim=1)
            all_preds.extend(preds.cpu().numpy().tolist())
            all_labels.extend(labs.cpu().numpy().tolist())

    # compute class-wise recall
    try:
        report = classification_report(all_labels, all_preds, output_dict=True, zero_division=0)
        moderate_recall = report['Moderate']['recall'] if 'Moderate' in report else report['1']['recall']
    except Exception:
        # sometimes keys are numeric
        report = classification_report(all_labels, all_preds, target_names=['Mild','Moderate','Severe'], output_dict=True, zero_division=0)
        moderate_recall = report['Moderate']['recall']

    print(f"Epoch {epoch+1} | Loss: {avg_loss:.4f} | Val Moderate recall: {moderate_recall:.4f}")

    scheduler.step(moderate_recall)

    if moderate_recall > best_moderate + 1e-6:
        best_moderate = moderate_recall
        torch.save(model.state_dict(), MODEL_SAVE_PATH)
        no_improve = 0
        print(f"  -> New best moderate recall: {best_moderate:.4f} (model saved)")
    else:
        no_improve += 1
        print(f"  -> No improvement ({no_improve}/{PATIENCE})")
        if no_improve >= PATIENCE:
            print("Early stopping triggered based on Moderate recall.")
            break

# --------------- Load best & evaluate on test ---------------
if os.path.exists(MODEL_SAVE_PATH):
    model.load_state_dict(torch.load(MODEL_SAVE_PATH))
model.eval()

test_preds = []; test_labels = []
with torch.no_grad():
    for imgs, tabs, labs in tqdm(test_loader, desc="Testing"):
        imgs = imgs.to(DEVICE); tabs = tabs.to(DEVICE); labs = labs.to(DEVICE)
        logits = model(imgs, tabs)
        preds = torch.argmax(logits, dim=1)
        test_preds.extend(preds.cpu().tolist())
        test_labels.extend(labs.cpu().tolist())

print("\n--- Test Classification Report ---")
print(classification_report(test_labels, test_preds, target_names=['Mild','Moderate','Severe'], zero_division=0))

# Confusion matrix visualization
cm = confusion_matrix(test_labels, test_preds, labels=[0,1,2])
fig, ax = plt.subplots(figsize=(6,6))
im = ax.imshow(cm, interpolation='nearest')
ax.set_xticks([0,1,2]); ax.set_yticks([0,1,2])
ax.set_xticklabels(['Mild','Moderate','Severe']); ax.set_yticklabels(['Mild','Moderate','Severe'])
ax.set_ylabel('True'); ax.set_xlabel('Pred')
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='white' if cm[i,j] > cm.max()/2 else 'black')
plt.title('Confusion Matrix')
plt.show()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦à¦Ÿà¦¿ à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦¬à§‡) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        # 'Moderate' à¦•à§à¦²à¦¾à¦¸à§‡à¦° à¦œà¦¨à§à¦¯ à¦¶à¦•à§à¦¤à¦¿à¦¶à¦¾à¦²à§€ à¦…à¦—à¦®à§‡à¦¨à§à¦Ÿà§‡à¦¶à¦¨
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦¶à§à¦§à§à¦®à¦¾à¦¤à§à¦° à¦®à¦¾à¦¸à§à¦• à¦•à¦°à¦¾ à¦…à¦‚à¦¶ à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡ (à¦¬à§à¦¯à¦¾à¦•à¦—à§à¦°à¦¾à¦‰à¦¨à§à¦¡ à¦¬à¦¾à¦¦)
                masked_image = img_tensor * pred_mask.clone().detach()
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # *** 'Moderate' (class 1) à¦•à§‡ à¦Ÿà¦¾à¦°à§à¦—à§‡à¦Ÿ à¦•à¦°à¦¾ ***
        minority_classes_list = [1] 
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° à¦à¦¬à¦‚ à¦“à¦¯à¦¼à§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # à§§. FocalLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à¦¯à¦¼à§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # à§¨. WeightedRandomSampler (à¦à¦–à¦¨ à¦†à¦° à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦¬à§‡ à¦¨à¦¾)
        # sample_weights = 1.0 / class_counts[labels_np]
        # sample_weights = torch.from_numpy(sample_weights).float()
        # num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        #
        # train_sampler = torch.utils.data.WeightedRandomSampler(
        #     weights=sample_weights,
        #     num_samples=num_samples_per_epoch,
        #     replacement=True
        # )

        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: FocalLoss-à¦à¦° à¦¸à¦¾à¦¥à§‡ à¦¸à§à¦¯à¦¾à¦®à§à¦ªà¦²à¦¾à¦° (sampler) à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à§Ÿ à¦¨à¦¾, à¦¤à¦¾à¦‡ sampler=None à¦à¦¬à¦‚ shuffle=True
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=None, shuffle=True, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (FocalLoss + No Sampler). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (DenseNet-121 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.densenet121(weights=None)
                cnn_feature_dim = self.cnn_base.classifier.in_features  # 1024
                self.cnn_base.classifier = nn.Identity()

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2), 
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.3),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        
        # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: WeightedCrossEntropyLoss-à¦à¦° à¦¬à¦¦à¦²à§‡ FocalLoss (alpha weights à¦¸à¦¹)
        criterion = FocalLoss(alpha=alpha_weights, gamma=3.0).to(device)
        
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=5e-5, weight_decay=1e-5) # 1e-5 -> 5e-5
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (FocalLoss + No Sampler)") # <-- à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
# ğŸ“¦ Install required libraries
!pip install -q albumentations scikit-image tqdm

# ğŸ§  Full SimCLR-style Classification Pipeline with Age & Gender
import os, cv2, torch, numpy as np, pandas as pd, matplotlib.pyplot as plt
from PIL import Image
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2

# --- CONFIG ---
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --- METADATA ---
metadata_df = pd.read_csv(METADATA_PATH)
metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
metadata_df.dropna(subset=['radiological_evaluation', 'Age', 'Gender'], inplace=True)
metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
metadata_df['ID'] = metadata_df['ID'].astype(str)
metadata_df['Gender'] = metadata_df['Gender'].map({'Male': 0, 'Female': 1})
class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
class_label_map = {1: 0, 2: 1, 3: 2}
metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
valid_ids = set(metadata_df['ID'].tolist())

def create_dataframe_from_ids(directory, allowed_ids):
    data = []
    if not os.path.exists(directory): return pd.DataFrame(data)
    for folder_name in os.listdir(directory):
        if folder_name in allowed_ids:
            folder_path = os.path.join(directory, folder_name)
            images_dir = os.path.join(folder_path, 'images')
            if os.path.exists(images_dir):
                for image_file in os.listdir(images_dir):
                    mask_path = os.path.join(folder_path, 'masks', image_file)
                    if os.path.exists(mask_path):
                        image_path = os.path.join(images_dir, image_file)
                        data.append((folder_name, image_path, mask_path))
    return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

# --- BALANCED SAMPLER ---
def get_sampler(df):
    class_counts = df['class_label'].value_counts().to_dict()
    weights = df['class_label'].map(lambda x: 1.0 / class_counts[x])
    return WeightedRandomSampler(weights.values, len(weights))

# --- DATASET ---
class CirrhosisMetaDataset(Dataset):
    def __init__(self, df, augment=True):
        self.df = df.reset_index(drop=True)
        self.transform = A.Compose([
            A.Resize(*IMAGE_SIZE),
            A.HorizontalFlip(p=0.5),
            A.RandomBrightnessContrast(p=0.5),
            A.Normalize(),
            ToTensorV2()
        ]) if augment else A.Compose([
            A.Resize(*IMAGE_SIZE),
            A.Normalize(),
            ToTensorV2()
        ])

    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image = cv2.imread(row['image_file_path'], cv2.IMREAD_GRAYSCALE)
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        image = self.transform(image=image)['image']
        metadata = torch.tensor([row['Age'], row['Gender']], dtype=torch.float32)
        label = row['class_label']
        return image, metadata, label

# --- MODEL ---
class SimCLRMetaClassifier(nn.Module):
    def __init__(self, num_classes=3):
        super().__init__()
        self.encoder = models.resnet50(pretrained=True)
        self.encoder.fc = nn.Identity()
        self.meta_fc = nn.Sequential(
            nn.Linear(2, 16),
            nn.ReLU(),
            nn.Dropout(0.2)
        )
        self.classifier = nn.Sequential(
            nn.Linear(2048 + 16, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )

    def forward(self, x, meta):
        img_feat = self.encoder(x)
        meta_feat = self.meta_fc(meta)
        combined = torch.cat([img_feat, meta_feat], dim=1)
        return self.classifier(combined)

# --- TRAINING ---
def train_model(model, train_loader, val_loader, epochs=20):
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='max', patience=3, factor=0.5)
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        train_loss, correct = 0, 0
        for x, meta, y in tqdm(train_loader):
            x, meta, y = x.to(device), meta.to(device), y.to(device)
            optimizer.zero_grad()
            out = model(x, meta)
            loss = criterion(out, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            correct += (out.argmax(1) == y).sum().item()
        acc = correct / len(train_loader.dataset)
        print(f"Epoch {epoch+1}: Train Loss={train_loss:.4f}, Accuracy={acc:.4f}")

        model.eval()
        val_correct = 0
        with torch.no_grad():
            for x, meta, y in val_loader:
                x, meta, y = x.to(device), meta.to(device), y.to(device)
                out = model(x, meta)
                val_correct += (out.argmax(1) == y).sum().item()
        val_acc = val_correct / len(val_loader.dataset)
        print(f"Validation Accuracy: {val_acc:.4f}")
        scheduler.step(val_acc)

# --- EVALUATION ---
def evaluate_model(model, test_loader):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for x, meta, y in test_loader:
            x, meta = x.to(device), meta.to(device)
            out = model(x, meta)
            preds = out.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(y.numpy())
    print(classification_report(all_labels, all_preds, target_names=['Mild', 'Moderate', 'Severe']))
    cm = confusion_matrix(all_labels, all_preds)
    ConfusionMatrixDisplay(cm, display_labels=['Mild', 'Moderate', 'Severe']).plot(cmap='Blues')
    plt.show()

# --- RUN ---
train_loader = DataLoader(CirrhosisMetaDataset(train_df), batch_size=32, sampler=get_sampler(train_df))
val_loader = DataLoader(CirrhosisMetaDataset(val_df, augment=False), batch_size=32)
test_loader = DataLoader(CirrhosisMetaDataset(test_df, augment=False), batch_size=32)

model = SimCLRMetaClassifier(num_classes=3)
train_model(model, train_loader, val_loader, epochs=20)
evaluate_model(model, test_loader)
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224) 

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, xx1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
# (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (à¦à¦–à¦¨ à¦†à¦° à¦¬à§à¦¯à¦¬à¦¹à§ƒà¦¤ à¦¹à¦šà§à¦›à§‡ à¦¨à¦¾, à¦¤à¦¬à§‡ à¦°à¦¾à¦–à¦¾ à¦¹à¦²à§‹) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Aggressive Augmentation à¦¬à¦¨à§à¦§ (à¦šà¦¾à¦²à§ à¦†à¦›à§‡)
        minority_classes_list = [] 
        
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 *** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: à¦¶à§à¦§à§ à¦“à§Ÿà§‡à¦Ÿà§‡à¦¡ à¦²à¦¸ (No Sampler) *** ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)
        print(f"Training set class counts (0=Mild, 1=Moderate, 2=Severe): {class_counts}")

        # CrossEntropyLoss-à¦à¦° à¦œà¦¨à§à¦¯ à¦•à§à¦²à¦¾à¦¸ à¦“à§Ÿà§‡à¦Ÿ
        weights = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        weights = weights / weights.sum() # Normalize
        alpha_weights = weights.to(device)
        print(f"Using alpha weights for loss: {alpha_weights}")

        # --- WeightedRandomSampler (à¦¬à¦¾à¦¦ à¦¦à§‡à¦“à§Ÿà¦¾ à¦¹à¦²à§‹) ---
        # train_sampler = torch.utils.data.WeightedRandomSampler(...)

        # *** à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: DataLoader-à¦ shuffle=True à¦¬à§à¦¯à¦¬à¦¹à¦¾à¦° à¦•à¦°à¦¾ à¦¹à¦šà§à¦›à§‡ (Sampler à¦¨à§‡à¦‡) ***
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (Weighted CE Loss + No Sampler). ğŸ“Š")

        # --- 7.4 EfficientNet-B4 à¦­à¦¿à¦¤à§à¦¤à¦¿à¦• Hybrid Classifier ---
        # (à¦à¦‡ à¦¸à§‡à¦•à¦¶à¦¨à§‡ à¦•à§‹à¦¨à§‹ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨ à¦¨à§‡à¦‡)
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.efficientnet_b4(weights=models.EfficientNet_B4_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.efficientnet_b4(weights=None)
                
                cnn_feature_dim = self.cnn_base.classifier[1].in_features # 1792
                self.cnn_base.classifier = nn.Identity()

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.2),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.2)
                )
                
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.3),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)             # (B, 1792)
                tabular_features = self.tabular_mlp(shape_features) # (B, 32)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        
        # --- *** à¦ªà§à¦°à¦§à¦¾à¦¨ à¦ªà¦°à¦¿à¦¬à¦°à§à¦¤à¦¨: Weighted CrossEntropyLoss (à¦ªà§à¦¨à¦°à¦¾à§Ÿ) *** ---
        criterion = nn.CrossEntropyLoss(weight=alpha_weights).to(device)
        
        # à¦…à¦ªà§à¦Ÿà¦¿à¦®à¦¾à¦‡à¦œà¦¾à¦° (lr=5e-5)
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=5e-5, weight_decay=1e-5)
        
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training (EfficientNet-B4 + Weighted CE Loss + No Sampler) ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (EfficientNet-B4 + Weighted CE Loss)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
# <--- MODIFIED: Added imports for texture features (GLCM)
from skimage.feature import graycomatrix, graycoprops
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape AND Texture features from masks ---
        # <--- MODIFIED: Replaced get_shape_features with get_all_features
        print("\nEngineering shape AND texture features...")
        def get_all_features(dataframe, mask_tensor):
            feature_list = []
            shape_properties = ('area', 'perimeter', 'solidity', 'eccentricity')
            texture_properties = ('contrast', 'energy', 'homogeneity', 'correlation')
            
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                row = dataframe.iloc[i]
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                
                # --- Shape Features ---
                props = regionprops_table(mask, properties=shape_properties)
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    shape_features = {prop: 0 for prop in shape_properties}
                    shape_features['circularity'] = 0
                else:
                    shape_features = props_df.head(1).to_dict('records')[0]
                    area = shape_features.get('area', 0)
                    perimeter = shape_features.get('perimeter', 0)
                    shape_features['circularity'] = (4 * np.pi * area) / (perimeter ** 2 + 1e-6)

                # --- Texture Features (GLCM) ---
                try:
                    # Load and resize original image to match mask size
                    image = cv2.imread(row['image_file_path'])
                    image = cv2.resize(image, IMAGE_SIZE)
                    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                    
                    # Apply mask to grayscale image
                    masked_gray = gray_img * mask
                    
                    # Calculate GLCM on the masked region
                    # We must offset by 1 to avoid computing on 0-0 pairs (background)
                    # We compute on non-zero elements
                    non_zero_pixels = masked_gray[masked_gray > 0]
                    if len(non_zero_pixels) > 20: # Need enough pixels to compute GLCM
                        glcm = graycomatrix(masked_gray, [1], [0], levels=256, symmetric=True, normed=True)
                        # We must remove the [0,0] entry which represents background-background pairs
                        glcm[0, 0] = 0
                        glcm = glcm / (glcm.sum() + 1e-6) # Re-normalize
                        
                        texture_features = {prop: graycoprops(glcm, prop)[0, 0] for prop in texture_properties}
                    else:
                        texture_features = {prop: 0 for prop in texture_properties}
                except Exception as e:
                    print(f"Warning: GLCM failed for index {i}, using zeros. Error: {e}")
                    texture_features = {prop: 0 for prop in texture_properties}
                
                # Combine all features
                all_features = {**shape_features, **texture_features}
                feature_list.append(all_features)
            
            feature_df = pd.DataFrame(feature_list)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_all_features(train_df, train_pred_masks)
        val_features_df   = get_all_features(val_df, val_pred_masks)
        test_features_df  = get_all_features(test_df, test_pred_masks)
        # <--- MODIFIED: Ensure column order is consistent
        train_cols = train_features_df.columns
        val_features_df = val_features_df.reindex(columns=train_cols, fill_value=0)
        test_features_df = test_features_df.reindex(columns=train_cols, fill_value=0)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        print(f"Features extracted. Total features: {train_features.shape[1]}")
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                
                # <--- MODIFIED: Changed masking strategy to "hard mask"
                # This zeroes out the background, forcing the model to look only at the liver
                masked_image = img_tensor * pred_mask.clone().detach()
                
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Stronger aug for Moderate + Severe
        minority_classes_list = [1, 2] # Labels 1 (Mod) and 2 (Severe)
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 <--- MODIFIED: Removed WeightedRandomSampler ---
        # We will use class weights in the loss function instead.

        # <--- MODIFIED: Create simple, shuffled loaders
        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready (using class-weighted loss). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (DenseNet-121 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.densenet121(weights=None)
                
                cnn_feature_dim = self.cnn_base.classifier.in_features  # 1024
                self.cnn_base.classifier = nn.Identity()  # output (B, 1024)

                # <--- MODIFIED: Increased MLP size to handle more features
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 128), nn.ReLU(), nn.Dropout(0.4), # Increased
                    nn.Linear(128, 64), nn.ReLU(), nn.Dropout(0.4) # Increased
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 64, 256), nn.ReLU(), nn.Dropout(0.5), # Changed input
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)              # (B, 1024)
                tabular_features = self.tabular_mlp(shape_features) # (B, 64)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        
        # <--- MODIFIED: Calculate class weights for CrossEntropyLoss
        # This replaces both FocalLoss and WeightedRandomSampler
        print("\nCalculating class weights for loss function...")
        labels_np = train_df['class_label'].to_numpy()
        # Ensure we count all 3 classes, even if one is missing in the array (minlength=3)
        class_counts = np.bincount(labels_np, minlength=3)
        print(f"Class counts (Mild, Mod, Severe): {class_counts}")
        
        # Avoid division by zero if a class has 0 samples
        class_counts = np.where(class_counts == 0, 1, class_counts)
        
        # Inverse frequency weighting
        class_weights_raw = 1.0 / torch.tensor(class_counts, dtype=torch.float)
        
        # Normalize weights (sum to 1)
        class_weights = class_weights_raw / class_weights_raw.sum()
        
        # Send to device
        class_weights = class_weights.to(device)
        
        print(f"Calculated class weights: {class_weights.cpu().numpy()}")
        
        # <--- MODIFIED: Use CrossEntropyLoss with class weights
        criterion = nn.CrossEntropyLoss(weight=class_weights)
        
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (DenseNet-121 + WeightedLoss + Texture)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x):
        return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3)
        d2 = self.decoder2(d3, e2)
        d1 = self.decoder1(d2, e1)
        d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images)
                    p2 = model2(images)
                    p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self):
                return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                # keep some context beyond mask
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [1, 2]  # stronger aug for Moderate + Severe
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler (balanced per epoch) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max()
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero

        # sample weight = inverse freq for each sample
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()

        # choose balanced epoch size
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))

        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(
            train_cls_dataset,
            batch_size=16,
            sampler=train_sampler,
            num_workers=2,
            pin_memory=True
        )
        val_cls_loader = DataLoader(val_cls_dataset, 16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, 16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (EfficientNet-B0 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.efficientnet_b0(weights=None)
                cnn_feature_dim = self.cnn_base.classifier[-1].in_features
                self.cnn_base.classifier = nn.Identity()
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        criterion = FocalLoss(alpha=None, gamma=3.0)  # single correction: oversampling
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (WeightedRandomSampler)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                # keep some context beyond mask
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Stronger aug for Moderate + Severe
        minority_classes_list = [1, 2]
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (ResNet-34 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.resnet34(weights=models.ResNet34_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.resnet34(weights=None)
                cnn_feature_dim = self.cnn_base.fc.in_features  # 512
                self.cnn_base.fc = nn.Identity()  # output (B, 512)

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)                 # (B, 512)
                tabular_features = self.tabular_mlp(shape_features) # (B, 32)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        criterion = FocalLoss(alpha=None, gamma=3.0)  # single correction: oversampling
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (ResNet-34 + WeightedRandomSampler)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                # keep some context beyond mask
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Stronger aug for Moderate + Severe
        minority_classes_list = [1, 2]
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (DenseNet-121 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.densenet121(weights=None)
                # DenseNet-121: classifier is nn.Linear(1024, 1000)
                cnn_feature_dim = self.cnn_base.classifier.in_features  # 1024
                self.cnn_base.classifier = nn.Identity()  # output (B, 1024)

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)                 # (B, 1024)
                tabular_features = self.tabular_mlp(shape_features) # (B, 32)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        criterion = FocalLoss(alpha=None, gamma=3.0)  # single correction: oversampling
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (DenseNet-121 + WeightedRandomSampler)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                # keep some context beyond mask
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Stronger aug for Moderate + Severe
        minority_classes_list = [1, 2]
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (ConvNeXt-Tiny + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.convnext_tiny(weights=models.ConvNeXt_Tiny_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.convnext_tiny(weights=None)
                # ConvNeXt classifier: [LayerNorm, Flatten, Linear]; use Identity on final Linear
                cnn_feature_dim = self.cnn_base.classifier[-1].in_features  # 768 for tiny
                self.cnn_base.classifier[-1] = nn.Identity()  # output (B, 768)

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)               # (B, 768)
                tabular_features = self.tabular_mlp(shape_features) # (B, 32)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined_features)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        criterion = FocalLoss(alpha=None, gamma=3.0)  # single correction: oversampling
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (ConvNeXt-Tiny + WeightedRandomSampler)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# Optional: SimCLR++ checkpoint for the encoder (set path if you have one)
SIMCLR_CKPT = None  # e.g. '/kaggle/input/simclrpp-unet/resnet50_unet_simclrpp.pth'

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODELS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                # keep some context beyond mask
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Stronger aug for Moderate + Severe
        minority_classes_list = [1, 2]
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (SimCLR++ U-Net encoder + tabular fusion) ---
        class SimCLRPPUNetBackbone(nn.Module):
            """
            ResNet50 encoder (as in your U-Nets) to extract pooled features.
            Optional SimCLR-style projection head (off by default).
            """
            def __init__(self, proj_dim=256, use_projection=False):
                super().__init__()
                base_model = models.resnet50(weights=None)
                layers = list(base_model.children())
                self.encoder0 = nn.Sequential(*layers[:3])   # conv1+bn1+relu
                self.encoder1 = nn.Sequential(*layers[3:5])  # maxpool + layer1
                self.encoder2 = layers[5]                    # layer2
                self.encoder3 = layers[6]                    # layer3
                self.encoder4 = layers[7]                    # layer4 (2048 ch)
                self.gap = nn.AdaptiveAvgPool2d(1)
                self.feat_dim = 2048
                self.use_projection = use_projection
                if use_projection:
                    self.proj = nn.Sequential(
                        nn.Linear(self.feat_dim, self.feat_dim, bias=False),
                        nn.BatchNorm1d(self.feat_dim),
                        nn.ReLU(inplace=True),
                        nn.Linear(self.feat_dim, proj_dim, bias=True)
                    )
                    self.out_dim = proj_dim
                else:
                    self.proj = None
                    self.out_dim = self.feat_dim

                if SIMCLR_CKPT is not None and os.path.exists(SIMCLR_CKPT):
                    ckpt = torch.load(SIMCLR_CKPT, map_location='cpu')
                    missing, unexpected = self.load_state_dict(ckpt, strict=False)
                    print(f"[SimCLR++] loaded. missing={len(missing)}, unexpected={len(unexpected)}")

            def forward(self, x, return_proj=False):
                e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
                h  = self.gap(e4).flatten(1)   # [B, 2048]
                if self.use_projection or return_proj:
                    z = self.proj(h)
                    if return_proj:
                        return h, z
                    return z
                return h

        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5, use_projection=False, proj_dim=256):
                super().__init__()
                self.backbone = SimCLRPPUNetBackbone(proj_dim=proj_dim, use_projection=use_projection)
                img_feat_dim = self.backbone.out_dim
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(img_feat_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                img_features = self.backbone(image)                 # (B, 2048) or proj_dim
                tabular_features = self.tabular_mlp(shape_features) # (B, 32)
                combined = torch.cat([img_features, tabular_features], dim=1)
                return self.fusion_classifier(combined)
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1], use_projection=False, proj_dim=256).to(device)
        criterion = FocalLoss(alpha=None, gamma=3.0)  # ONLY sampler balancing
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (SimCLR++ U-Net encoder + WeightedRandomSampler)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- CoAtNet backbone needs timm ---
try:
    import timm
except ImportError as e:
    raise ImportError("Please install timm first: pip install -U timm") from e

# ------------------------
# 1) CONFIG
# ------------------------
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR   = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

BATCH_SIZE = 16
LR = 1e-4
NUM_EPOCHS = 50
PATIENCE = 10
FOCAL_GAMMA = 3.0

# Set a specific CoAtNet name you want; otherwise we'll auto-pick
COATNET_NAME = None

# ------------------------
# 2) SETUP
# ------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")
torch.backends.cudnn.benchmark = True

# ------------------------
# 3) DATAFRAME CREATION AND MERGE
# ------------------------
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)

    class_name_map  = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0,     2: 1,         3: 2}
    metadata_df['class_name']  = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df   = pd.merge(create_dataframe_from_ids(VAL_DIR,   valid_ids), metadata_df, on='ID')
    test_df  = pd.merge(create_dataframe_from_ids(TEST_DIR,  valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")

    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# ------------------------
# 4) DATASET FOR INITIAL SEGMENTATION
# ------------------------
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self): return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# ------------------------
# 5) SEGMENTATION MODELS (same as before)
# ------------------------
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# ------------------------
# 6) LOAD SEG MODELS
# ------------------------
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# ------------------------
# FOCAL LOSS
# ------------------------
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=FOCAL_GAMMA, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# 7) CLASSIFICATION (HYBRID)
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Masks ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset   = SegmentationInputDataset(val_df)
        test_seg_dataset  = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)

        # --- 7.3 Datasets (Albumentations v2 API) ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.Affine(translate_percent=(0.0, 0.05), scale=(0.95, 1.05), rotate=(-10, 10), p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(num_holes=8, max_h_size=8, max_w_size=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.Affine(translate_percent=(0.0, 0.10), scale=(0.90, 1.10), rotate=(-20, 20), p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(num_holes=16, max_h_size=16, max_w_size=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path']); image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [1, 2]  # Moderate + Severe stronger aug
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,  val_pred_masks,  val_features,  eval_transform)
        test_cls_dataset  = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)

        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))
        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=BATCH_SIZE, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 Hybrid Classifier (CoAtNet + tabular) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5, coatnet_name=COATNET_NAME):
                super().__init__()
                available = timm.list_models('*coatnet*')
                if len(available) == 0:
                    raise RuntimeError("No CoAtNet variants found in timm. Try: pip install -U timm")

                preferred = [
                    'coatnet_pico_rw_224', 'coatnet_nano_rw_224',
                    'coatnet_rmlp_0_224', 'coatnet_rmlp_1_224', 'coatnet_rmlp_2_224',
                    'coatnet_0_rw_224', 'coatnet_1_rw_224', 'coatnet_2_rw_224',
                    'coatnet_3_rw_224', 'coatnet_4_rw_224'
                ]

                if coatnet_name is not None:
                    if coatnet_name not in available:
                        raise ValueError(f"Requested CoAtNet '{coatnet_name}' not in available list.\nAvailable: {available}")
                    picked = coatnet_name
                else:
                    picked = next((n for n in preferred if n in available), available[0])

                print(f"[CoAtNet] Using backbone: {picked}")

                # try pretrained; if not available, fallback to random init
                try:
                    self.cnn_base = timm.create_model(picked, pretrained=True, num_classes=0, global_pool='avg')
                except RuntimeError as e:
                    if "No pretrained weights exist" in str(e):
                        print(f"[CoAtNet] Pretrained weights not found for {picked}. Using random init.")
                        self.cnn_base = timm.create_model(picked, pretrained=False, num_classes=0, global_pool='avg')
                    else:
                        raise

                cnn_feature_dim = getattr(self.cnn_base, 'num_features', None)
                if cnn_feature_dim is None:
                    with torch.no_grad():
                        dummy = torch.zeros(1, 3, 224, 224)
                        feat = self.cnn_base(dummy)
                        cnn_feature_dim = feat.shape[1]

                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(shape_features)
                combined = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined)

        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        criterion = FocalLoss(alpha=None, gamma=FOCAL_GAMMA)  # ONLY oversampling used to balance
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=LR)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)

        best_val_accuracy = 0.0
        epochs_no_improve = 0
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(NUM_EPOCHS):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{NUM_EPOCHS} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{NUM_EPOCHS} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= PATIENCE:
                print(f"\nEarly stopping triggered after {PATIENCE} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth', map_location=device))
        cls_model.to(device).eval()

        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())

        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))

        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")

        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (CoAtNet + WeightedRandomSampler)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
# --- MODIFIED: Added WeightedRandomSampler import ---
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
# --- CORRECTED PATH ---
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    # Ensure 'Age' and 'Gender' are present. Handle potential NaNs later.
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            print(f"Warning: Directory not found, skipping: {directory}")
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    # Merge will automatically bring in 'Age', 'Gender' columns
    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")

    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None # Set to None so the rest of the script skips

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        try:
            image_pil = Image.open(row['image_file_path']).convert('RGB')
            return self.transform(image_pil)
        except Exception as e:
            print(f"Error loading image {row['image_file_path']}: {e}")
            return torch.zeros((3, IMAGE_SIZE[0], IMAGE_SIZE[1])) # Return dummy tensor on error

# --- 5. MODEL ARCHITECTURE DEFINITIONS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x):
        return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3)
        d2 = self.decoder2(d3, e2)
        d1 = self.decoder1(d2, e1)
        d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None with oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            # Ensure alpha is on the correct device
            alpha_tensor = self.alpha.to(target.device) if isinstance(self.alpha, torch.Tensor) else self.alpha
            alpha_t = alpha_tensor[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()


# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors in setup or model loading.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images)
                    p2 = model2(images)
                    p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        # --- MODIFIED: ADD CLINICAL FEATURES ---
        print("\nCombining shape and clinical features (Age, Gender)...")

        # Get Age and Gender from the dataframes
        train_clinical_df = train_df[['Age', 'Gender']]
        val_clinical_df = val_df[['Age', 'Gender']]
        test_clinical_df = test_df[['Age', 'Gender']]

        # Combine with shape features
        train_tabular_df = pd.concat([train_features_df, train_clinical_df.reset_index(drop=True)], axis=1)
        val_tabular_df = pd.concat([val_features_df, val_clinical_df.reset_index(drop=True)], axis=1)
        test_tabular_df = pd.concat([test_features_df, test_clinical_df.reset_index(drop=True)], axis=1)

        # Handle potential NaNs in Age/Gender if any
        # Fill with the mean *from the training set* to avoid data leakage
        train_mean = train_tabular_df.mean()
        train_tabular_df.fillna(train_mean, inplace=True)
        val_tabular_df.fillna(train_mean, inplace=True)
        test_tabular_df.fillna(train_mean, inplace=True)

        print(f"Total tabular features being used: {list(train_tabular_df.columns)}")

        # Re-scale ALL tabular features together
        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_tabular_df)
        val_features = scaler.transform(val_tabular_df)
        test_features = scaler.transform(test_tabular_df)

        print(f"Total number of scaled features: {train_features.shape[1]}") # Should be 7

        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                # self.features is now the COMBINED (shape + clinical) array
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self):
                return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                label = int(row['class_label'])
                try:
                    image = cv2.imread(row['image_file_path'])
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                except Exception as e:
                    print(f"Error reading image {row['image_file_path']}: {e}")
                    # Return dummy data
                    img_tensor = torch.zeros((3, IMAGE_SIZE[0], IMAGE_SIZE[1]))
                    shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                    label_tensor = torch.tensor(label, dtype=torch.long)
                    return img_tensor, shape_feats, label_tensor

                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)

                # Using the original soft mask strategy
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())

                # 'shape_feats' now contains shape + clinical features
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)

                label_tensor = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label_tensor


        minority_classes_list = [1, 2]  # stronger aug for Moderate + Severe
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)

        # --- 7.3.1 SINGLE OVERSAMPLING METHOD: WeightedRandomSampler (balanced per epoch) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels, counts = np.unique(labels_np, return_counts=True)

        # Ensure class_counts array has entries for all possible labels up to max
        max_label_val = labels_np.max()
        class_counts = np.zeros(max_label_val + 1, dtype=int)
        class_counts[unique_labels] = counts

        # Get counts only for the labels present in the dataset
        present_labels = unique_labels
        present_class_counts = class_counts[present_labels]

        # Avoid div-by-zero for labels that might exist but have 0 samples
        class_counts_safe = np.where(class_counts == 0, 1, class_counts)

        # Create weights based on inverse frequency using the safe counts
        sample_weights = 1.0 / class_counts_safe[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()

        # Choose balanced epoch size based on max count among *present* classes
        num_samples_per_epoch = int(present_class_counts.max() * len(present_labels))


        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(
            train_cls_dataset,
            batch_size=16,
            sampler=train_sampler,
            num_workers=2,
            pin_memory=True
        )
        val_cls_loader = DataLoader(val_cls_dataset, 16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, 16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 MODIFIED: Hybrid Classifier (EfficientNet-B0 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            # --- SYNTAX FIX APPLIED ---
            def __init__(self, n_shape_features, n_classes=3):
                super().__init__()
                try:
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.efficientnet_b0(weights=None)
                cnn_feature_dim = self.cnn_base.classifier[-1].in_features
                self.cnn_base.classifier = nn.Identity()
                self.tabular_mlp = nn.Sequential(
                    # Use the dynamic feature count (e.g., 7)
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )

            def forward(self, image, shape_features): # 'shape_features' now holds all tabular data
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output

        # --- 7.5 Train ---
        # MODIFIED: Pass the new total feature count (e.g., 7)
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)

        criterion = FocalLoss(alpha=None, gamma=3.0)  # single correction: oversampling
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0

        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            # The loader provides 'images', 'shapes', 'labels'
            # 'shapes' now corresponds to our 7-feature tensor
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes) # This maps to forward(self, image, shape_features)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"     ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"     ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        # Load the best model for evaluation
        if os.path.exists('best_classifier_model.pth'):
            cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        else:
            print("Warning: 'best_classifier_model.pth' not found. Evaluating the last model state.")
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())

        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))

        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")

        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (WeightedRandomSampler, Clinical Feats)")
        plt.savefig("confusion_matrix_clinical.png") # Save the plot
        print("Confusion matrix plot saved as 'confusion_matrix_clinical.png'")
        # plt.show() # Remove plt.show() for non-interactive environments

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from torch.utils.data.sampler import WeightedRandomSampler
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# Optional: SimCLR++ checkpoint for the encoder (set path if you have one)
SIMCLR_CKPT = None  # e.g. '/kaggle/input/simclrpp-unet/resnet50_unet_simclrpp.pth'

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODEL ARCHITECTURES ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0']*2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1']*2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0']*3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2']*2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1']*3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0']*4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS (alpha=None -> no class weighting) ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean':
            return loss.mean()
        else:
            return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
            def __len__(self): return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                transformed = self.transform(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                masked_image = img_tensor * pred_mask.clone().detach()
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # no minority-class special aug; single transform everywhere
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features, transform=train_transform)
        val_cls_dataset   = HybridDataset(val_df,   val_pred_masks,   val_features,   transform=eval_transform)
        test_cls_dataset  = HybridDataset(test_df,  test_pred_masks,  test_features,  transform=eval_transform)
        
        # --- WeightedRandomSampler (ONLY) ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        max_label = unique_labels.max() if len(unique_labels) else 0
        class_counts = np.bincount(labels_np, minlength=max_label+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))

        train_sampler = WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with WeightedRandomSampler (ONLY). ğŸ“Š")

        # --- 7.4 UPDATED: HYBRID MODEL ARCHITECTURE with SimCLR++ U-Net encoder ---
        class SimCLRPPUNetBackbone(nn.Module):
            def __init__(self, proj_dim=256, use_projection=False):
                super().__init__()
                base_model = models.resnet50(weights=None)
                layers = list(base_model.children())
                self.encoder0 = nn.Sequential(*layers[:3])
                self.encoder1 = nn.Sequential(*layers[3:5])
                self.encoder2 = layers[5]
                self.encoder3 = layers[6]
                self.encoder4 = layers[7]
                self.gap = nn.AdaptiveAvgPool2d(1)
                self.feat_dim = 2048
                self.use_projection = use_projection
                if use_projection:
                    self.proj = nn.Sequential(
                        nn.Linear(self.feat_dim, self.feat_dim, bias=False),
                        nn.BatchNorm1d(self.feat_dim),
                        nn.ReLU(inplace=True),
                        nn.Linear(self.feat_dim, proj_dim, bias=True)
                    )
                    self.out_dim = proj_dim
                else:
                    self.proj = None
                    self.out_dim = self.feat_dim
                if SIMCLR_CKPT is not None and os.path.exists(SIMCLR_CKPT):
                    ckpt = torch.load(SIMCLR_CKPT, map_location='cpu')
                    self.load_state_dict(ckpt, strict=False)

            def forward(self, x, return_proj=False):
                e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
                h  = self.gap(e4).flatten(1)
                if self.use_projection or return_proj:
                    z = self.proj(h)
                    if return_proj:
                        return h, z
                    return z
                return h

        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5, use_projection=False, proj_dim=256):
                super().__init__()
                self.backbone = SimCLRPPUNetBackbone(proj_dim=proj_dim, use_projection=use_projection)
                img_feat_dim = self.backbone.out_dim
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(img_feat_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                img_features = self.backbone(image)
                tab_feats   = self.tabular_mlp(shape_features)
                fused       = torch.cat([img_features, tab_feats], dim=1)
                logits      = self.fusion_classifier(fused)
                return logits
        
        # --- 7.5 TRAIN with HYBRID MODEL ---
        cls_model = HybridClassifier(
            n_classes=3,
            n_shape_features=train_features.shape[1],
            use_projection=False,
            proj_dim=256
        ).to(device)

        criterion = FocalLoss(alpha=None, gamma=3.0)   # ONLY WeightedRandomSampler (no class weights)
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / len(train_cls_loader):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 FINAL EVALUATION ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        mild_accuracy = report['Mild']['recall']
        moderate_accuracy = report['Moderate']['recall']
        severe_accuracy = report['Severe']['recall']
        print(f"Accuracy for 'Mild' class: {mild_accuracy:.4f}")
        print(f"Accuracy for 'Moderate' class: {moderate_accuracy:.4f}")
        print(f"Accuracy for 'Severe' class: {severe_accuracy:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (WeightedRandomSampler ONLY)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset, SubsetRandomSampler
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, f1_score
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# ================================
# 1) CONFIGURATION
# ================================
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR   = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# Oversampling emphasis (key change): multiply Moderate (class 1)
OVERSAMPLE_MULTIPLIERS = {0: 1.0, 1: 2.0, 2: 1.0}  # tune 1.5â€“2.5 for class 1

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# ================================
# 2) DATAFRAME CREATION AND MERGE
# ================================
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)

    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df   = pd.merge(create_dataframe_from_ids(VAL_DIR,   valid_ids), metadata_df, on='ID')
    test_df  = pd.merge(create_dataframe_from_ids(TEST_DIR,  valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")

    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")

    # You can still compute class weights for reference, but we won't use alpha with focal (to avoid double balancing)
    class_weights_np = compute_class_weight('balanced',
                                            classes=np.unique(train_df['class_label'].values),
                                            y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights_np, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights (reference): {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# ================================
# 3) DATASET FOR INITIAL SEGMENTATION
# ================================
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# ================================
# 4) SEGMENTATION MODELS (as you had)
# ================================
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False),
            nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False),
            nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True)
        )
    def forward(self, x):
        return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0 = nn.Sequential(*base_layers[:3])
        self.encoder1 = nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3)
        d2 = self.decoder2(d3, e2)
        d1 = self.decoder1(d2, e1)
        d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0 = nn.Sequential(*base_layers[:3])
        self.encoder1 = nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads,
                                                       dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim),
                                                        nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0 = nn.Sequential(*layers[:3])
        self.encoder1 = nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# ================================
# 5) LOAD SEGMENTATION MODELS
# ================================
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# ================================
# 6) FOCAL LOSS (less aggressive, alpha=None)
# ================================
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=1.5, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# ================================
# 7) CLASSIFICATION STAGE (HYBRID)
# ================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset   = SegmentationInputDataset(val_df)
        test_seg_dataset  = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images)
                    p2 = model2(images)
                    p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape Features ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)

        # --- 7.3 Augmentations (class-aware for Moderate+Severe) ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([   # used for classes 1 (Moderate) and 2 (Severe)
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.6),
            A.GaussNoise(p=0.3),
            A.GridDistortion(p=0.2),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform,
                         minority_classes=None, minority_transform=None):
                self.df = dataframe.reset_index(drop=True)
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self):
                return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])

                aug = self.minority_transform if label in self.minority_classes else self.transform
                transformed = aug(image=image)
                img_tensor = transformed['image']

                pred_mask = self.masks[idx].to(img_tensor.device)  # [1,H,W]
                masked_image = img_tensor * pred_mask.clone().detach()

                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Key change: include Moderate (1) and Severe (2) for stronger aug
        minority_classes_list = [1, 2]

        train_cls_dataset = HybridDataset(
            train_df, train_pred_masks, train_features,
            transform=train_transform,
            minority_classes=minority_classes_list,
            minority_transform=minority_transform
        )
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)

        # --- 7.4 OVERSAMPLING with Moderate emphasis (single balancing method) ---
        labels_np = train_df['class_label'].to_numpy()
        class_to_indices = {c: np.where(labels_np == c)[0] for c in np.unique(labels_np)}
        raw_counts = {int(c): int(len(ix)) for c, ix in class_to_indices.items()}
        max_raw = max(len(ix) for ix in class_to_indices.values())
        # target per class with multiplier (Moderate > others)
        targets = {c: int(np.ceil(max_raw * OVERSAMPLE_MULTIPLIERS.get(int(c), 1.0))) for c in class_to_indices.keys()}
        print(f"\nRaw counts: {raw_counts} | targets (with multipliers): {targets}")

        rng = np.random.default_rng(2024)
        oversampled_indices = []
        for c, ix in class_to_indices.items():
            if len(ix) == 0:
                continue
            need = max(0, targets[c] - len(ix))
            if need > 0:
                extra = rng.choice(ix, size=need, replace=True)
                merged = np.concatenate([ix, extra])
            else:
                merged = ix
            oversampled_indices.append(merged)
        oversampled_indices = np.concatenate(oversampled_indices)
        rng.shuffle(oversampled_indices)

        # report after
        counts_after = {}
        after_labels = train_df['class_label'].iloc[oversampled_indices].to_numpy()
        for c in class_to_indices.keys():
            counts_after[int(c)] = int(np.sum(after_labels == c))
        print(f"Counts after oversample: {counts_after}")

        train_sampler = SubsetRandomSampler(oversampled_indices.tolist())

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler,
                                      num_workers=2, pin_memory=True)
        val_cls_loader = DataLoader(val_cls_dataset, batch_size=16, shuffle=False,
                                    num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, batch_size=16, shuffle=False,
                                     num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with Oversampling (Moderate emphasized). ğŸ“Š")

        # --- 7.5 CLASSIFIER (EfficientNet-B0 + shape features) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.efficientnet_b0(weights=None)
                cnn_feature_dim = self.cnn_base.classifier[-1].in_features
                self.cnn_base.classifier = nn.Identity()
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(shape_features)
                combined = torch.cat([cnn_features, tabular_features], dim=1)
                return self.fusion_classifier(combined)

        # --- 7.6 TRAINING (early stop/schedule on macro-F1) ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)

        # IMPORTANT: alpha=None because we already oversample (single balancing mechanism)
        criterion = FocalLoss(alpha=None, gamma=1.5)
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)

        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_macro_f1 = -1.0

        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            # Train
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            # Validate
            cls_model.eval()
            val_corrects = 0
            all_val_labels, all_val_preds = [], []
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)
                    all_val_labels.extend(labels.cpu().numpy().tolist())
                    all_val_preds.extend(preds.cpu().numpy().tolist())

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            macro_f1 = f1_score(all_val_labels, all_val_preds, average='macro')
            # Pull Moderate recall:
            val_report = classification_report(all_val_labels, all_val_preds,
                                               target_names=['Mild','Moderate','Severe'],
                                               zero_division=0, output_dict=True)
            moderate_recall = val_report['Moderate']['recall']

            print(f"Epoch {epoch + 1} | Loss: {running_loss / len(train_cls_loader):.4f} "
                  f"| Val Acc: {val_accuracy:.4f} | Macro-F1: {macro_f1:.4f} | Moderate R: {moderate_recall:.4f}")

            # Drive scheduler & early-stopping by macro-F1
            scheduler.step(macro_f1)

            if macro_f1 > best_macro_f1:
                best_macro_f1 = macro_f1
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved. Macro-F1: {best_macro_f1:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")

            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.7 FINAL EVALUATION ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth', map_location=device))
        cls_model.to(device)
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())

        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))

        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print("\n--- Individual Class Accuracies ---")
        print(f"Accuracy for 'Mild' class:     {report['Mild']['recall']:.4f}")
        print(f"Accuracy for 'Moderate' class: {report['Moderate']['recall']:.4f}")
        print(f"Accuracy for 'Severe' class:   {report['Severe']['recall']:.4f}")

        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (Oversampled, Moderate Emphasis)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset, SubsetRandomSampler
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# Optional: SimCLR++ checkpoint for the encoder (set path if you have one)
SIMCLR_CKPT = None  # e.g. '/kaggle/input/simclrpp-unet/resnet50_unet_simclrpp.pth'

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    class_weights = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights for Focal Loss: {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. SEGMENTATION MODEL ARCHITECTURES ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3); d2 = self.decoder2(d3, e2); d1 = self.decoder1(d2, e1); d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0']); self.X_1_0 = ConvBlock(256, d_ch['d1']); self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2']); self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1']); self.X_0_2 = ConvBlock(d_ch['d0']*2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3']); self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2']); self.X_1_2 = ConvBlock(d_ch['d1']*2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0']*3 + d_ch['d1'], d_ch['d0']); self.X_4_0 = ConvBlock(2048, d_ch['d4']); self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2']*2 + d_ch['d3'], d_ch['d2']); self.X_1_3 = ConvBlock(d_ch['d1']*3 + d_ch['d2'], d_ch['d1']); self.X_0_4 = ConvBlock(d_ch['d0']*4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0); x1_0 = self.X_1_0(e1); x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2); x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1)); x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3); x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1)); x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1)); x4_0 = self.X_4_0(trans_out); x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1)); x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOADING SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device)); model1.eval()
        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device)); model2.eval()
        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device)); model3.eval()
        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS IMPLEMENTATION ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2., reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma * ce_loss)
        if self.alpha is not None:
            alpha_t = self.alpha[target.data.view(-1)]
            loss = alpha_t * loss
        if self.reduction == 'mean':
            return loss.mean()
        else:
            return loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================

print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Generate Segmentation Predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images); p2 = model2(images); p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks   = get_segmentation_predictions(val_seg_dataset,   "Generating Val Masks")
        test_pred_masks  = get_segmentation_predictions(test_seg_dataset,  "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Feature Engineering from Masks ---
        print("\nEngineering shape features from masks...")

        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df   = get_shape_features(val_pred_masks)
        test_features_df  = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features   = scaler.transform(val_features_df)
        test_features  = scaler.transform(test_features_df)
        
        # --- 7.3 Prepare Data and Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform

            def __len__(self): return len(self.df)

            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = row['class_label']
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                masked_image = img_tensor * pred_mask.clone().detach()
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        minority_classes_list = [2]
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset   = HybridDataset(val_df,   val_pred_masks,   val_features,   eval_transform)
        test_cls_dataset  = HybridDataset(test_df,  test_pred_masks,  test_features,  eval_transform)
        
        # --- OVERSAMPLE (image-level) so each class has same number of samples ---
        labels_np = train_df['class_label'].to_numpy()
        class_to_indices = {c: np.where(labels_np == c)[0] for c in np.unique(labels_np)}
        counts_before = {int(c): int(len(ix)) for c, ix in class_to_indices.items()}
        target = max(len(ix) for ix in class_to_indices.values())
        print(f"\nClass counts before oversample: {counts_before} | target per class: {target}")

        oversampled_indices = []
        rng = np.random.default_rng(2024)
        for c, ix in class_to_indices.items():
            if len(ix) == 0: continue
            need = target - len(ix)
            merged = np.concatenate([ix, rng.choice(ix, size=need, replace=True)]) if need > 0 else ix
            oversampled_indices.append(merged)
        oversampled_indices = np.concatenate(oversampled_indices); rng.shuffle(oversampled_indices)

        counts_after = {}
        for c in class_to_indices.keys():
            counts_after[int(c)] = int(np.sum(train_df['class_label'].iloc[oversampled_indices].to_numpy() == c))
        print(f"Class counts after  oversample: {counts_after}")

        train_sampler = SubsetRandomSampler(oversampled_indices.tolist())

        train_cls_loader = DataLoader(train_cls_dataset, batch_size=16, sampler=train_sampler, num_workers=2, pin_memory=True)
        val_cls_loader   = DataLoader(val_cls_dataset,   16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader  = DataLoader(test_cls_dataset,  16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with Oversampling. ğŸ“Š")

        # --- 7.4 UPDATED: HYBRID MODEL ARCHITECTURE with SimCLR++ U-Net encoder ---
        class SimCLRPPUNetBackbone(nn.Module):
            """
            Reuse a ResNet50 encoder (as in your U-Nets) to extract pooled features.
            Optional SimCLR-style projection head is included but off by default.
            """
            def __init__(self, proj_dim=256, use_projection=False):
                super().__init__()
                base_model = models.resnet50(weights=None)
                layers = list(base_model.children())
                self.encoder0 = nn.Sequential(*layers[:3])   # conv1+bn1+relu
                self.encoder1 = nn.Sequential(*layers[3:5])  # maxpool + layer1
                self.encoder2 = layers[5]                    # layer2
                self.encoder3 = layers[6]                    # layer3
                self.encoder4 = layers[7]                    # layer4 (2048 ch)

                self.gap = nn.AdaptiveAvgPool2d(1)
                self.feat_dim = 2048
                self.use_projection = use_projection
                if use_projection:
                    self.proj = nn.Sequential(
                        nn.Linear(self.feat_dim, self.feat_dim, bias=False),
                        nn.BatchNorm1d(self.feat_dim),
                        nn.ReLU(inplace=True),
                        nn.Linear(self.feat_dim, proj_dim, bias=True)
                    )
                    self.out_dim = proj_dim
                else:
                    self.proj = None
                    self.out_dim = self.feat_dim

                if SIMCLR_CKPT is not None and os.path.exists(SIMCLR_CKPT):
                    ckpt = torch.load(SIMCLR_CKPT, map_location='cpu')
                    missing, unexpected = self.load_state_dict(ckpt, strict=False)
                    print(f"[SimCLR++] loaded. missing={len(missing)}, unexpected={len(unexpected)}")

            def forward(self, x, return_proj=False):
                e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
                h  = self.gap(e4).flatten(1)   # [B, 2048]
                if self.use_projection or return_proj:
                    z = self.proj(h)
                    if return_proj:
                        return h, z
                    return z
                return h

        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5, use_projection=False, proj_dim=256):
                super().__init__()
                self.backbone = SimCLRPPUNetBackbone(proj_dim=proj_dim, use_projection=use_projection)
                img_feat_dim = self.backbone.out_dim
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(img_feat_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                img_features = self.backbone(image)     # [B, 2048] (or proj_dim if projection used)
                tab_feats   = self.tabular_mlp(shape_features)
                fused       = torch.cat([img_features, tab_feats], dim=1)
                logits      = self.fusion_classifier(fused)
                return logits
        
        # --- 7.5 TRAIN with HYBRID MODEL ---
        cls_model = HybridClassifier(
            n_classes=3,
            n_shape_features=train_features.shape[1],
            use_projection=False,  # set True only if you specifically want projection features
            proj_dim=256
        ).to(device)

        # You are already oversampling; you may try alpha=None to rely purely on oversampling.
        criterion = FocalLoss(alpha=class_weights, gamma=3.0)
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / len(train_cls_loader):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 FINAL EVALUATION ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        mild_accuracy = report['Mild']['recall']
        moderate_accuracy = report['Moderate']['recall']
        severe_accuracy = report['Severe']['recall']
        print(f"Accuracy for 'Mild' class: {mild_accuracy:.4f}")
        print(f"Accuracy for 'Moderate' class: {moderate_accuracy:.4f}")
        print(f"Accuracy for 'Severe' class: {severe_accuracy:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (Oversampled, SimCLR++ U-Net encoder)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```

```python
import os
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import torch.nn.functional as F
import torch.nn as nn
from torchvision import models, transforms
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight
from torch.optim.lr_scheduler import ReduceLROnPlateau
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
from skimage.measure import regionprops_table
from sklearn.preprocessing import StandardScaler

# --- 1. CONFIGURATION ---
MODEL1_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model1.pth'
MODEL2_PATH = '/kaggle/input/ourensemble/3URESNET/best_transunetpp_model.pth'
MODEL3_PATH = '/kaggle/input/ourensemble/3URESNET/best_segmentation_model3.pth'
METADATA_PATH = '/kaggle/input/t2metadataaa2/T2_age_gender_evaluation.csv'
TEST_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/test'
TRAIN_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/train'
VAL_DIR = '/kaggle/input/dataa2/Cirrhosis_T2_2D/valid'
IMAGE_SIZE = (224, 224)

# --- 2. SETUP ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} âš™ï¸")

# --- 3. DATAFRAME CREATION AND MERGE ---
print("\n--- Creating DataFrames ---")
try:
    metadata_df = pd.read_csv(METADATA_PATH)
    metadata_df.rename(columns={'Patient ID': 'ID', 'Radiological Evaluation': 'radiological_evaluation'}, inplace=True)
    metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
    metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)
    metadata_df['ID'] = metadata_df['ID'].astype(str)
    class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
    class_label_map = {1: 0, 2: 1, 3: 2}
    metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
    metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
    metadata_df.dropna(subset=['class_name'], inplace=True)
    valid_ids = set(metadata_df['ID'].tolist())

    def create_dataframe_from_ids(directory, allowed_ids):
        data = []
        if not os.path.exists(directory):
            return pd.DataFrame(data)
        for folder_name in os.listdir(directory):
            if folder_name in allowed_ids:
                folder_path = os.path.join(directory, folder_name)
                images_dir = os.path.join(folder_path, 'images')
                if os.path.exists(images_dir):
                    for image_file in os.listdir(images_dir):
                        mask_path = os.path.join(folder_path, 'masks', image_file)
                        if os.path.exists(mask_path):
                            image_path = os.path.join(images_dir, image_file)
                            data.append((folder_name, image_path, mask_path))
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

    train_df = pd.merge(create_dataframe_from_ids(TRAIN_DIR, valid_ids), metadata_df, on='ID')
    val_df = pd.merge(create_dataframe_from_ids(VAL_DIR, valid_ids), metadata_df, on='ID')
    test_df = pd.merge(create_dataframe_from_ids(TEST_DIR, valid_ids), metadata_df, on='ID')

    if train_df.empty or val_df.empty or test_df.empty:
        raise ValueError("One or more dataframes are empty. Check your file paths and metadata.")
    
    print(f"Successfully created dataframes: {len(train_df)} train, {len(val_df)} val, {len(test_df)} test.")
    
    # (Optionally still compute class weights just to log; not used in loss when oversampling)
    class_weights = compute_class_weight('balanced', classes=np.unique(train_df['class_label'].values), y=train_df['class_label'].values)
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)
    print(f"\nCalculated Class Weights (not used with oversampling): {class_weights}")

except Exception as e:
    print(f"Error during data setup: {e}")
    test_df = None

# --- 4. DATASET FOR INITIAL SEGMENTATION ---
class SegmentationInputDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.transform = transforms.Compose([
            transforms.Resize(IMAGE_SIZE, transforms.InterpolationMode.BILINEAR),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image_pil = Image.open(row['image_file_path']).convert('RGB')
        return self.transform(image_pil)

# --- 5. MODEL ARCHITECTURE DEFINITIONS ---
class ConvBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x):
        return self.conv(x)

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels + skip_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU(inplace=True)
               self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(out_channels)
    def forward(self, x, skip_connection):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = torch.cat([x, skip_connection], dim=1)
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        return x

class UNetWithResNet50Encoder(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        self.decoder3 = DecoderBlock(2048, 1024, 512)
        self.decoder2 = DecoderBlock(512, 512, 256)
        self.decoder1 = DecoderBlock(256, 256, 128)
        self.decoder0 = DecoderBlock(128, 64, 64)
        self.final_conv = nn.Conv2d(64, n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x)
        e1 = self.encoder1(e0)
        e2 = self.encoder2(e1)
        e3 = self.encoder3(e2)
        e4 = self.encoder4(e3)
        d3 = self.decoder3(e4, e3)
        d2 = self.decoder2(d3, e2)
        d1 = self.decoder1(d2, e1)
        d0 = self.decoder0(d1, e0)
        out = F.interpolate(d0, scale_factor=2, mode='bilinear', align_corners=True)
        return self.final_conv(out)

class TransUNetPP(nn.Module):
    def __init__(self, n_classes=1, img_dim=224, vit_dim=768, vit_depth=12, vit_heads=12):
        super().__init__()
        base_model = models.resnet50(weights=None)
        base_layers = list(base_model.children())
        self.encoder0, self.encoder1 = nn.Sequential(*base_layers[:3]), nn.Sequential(*base_layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = base_layers[5], base_layers[6], base_layers[7]
        num_patches, self.patch_dim = (img_dim // 32) ** 2, 2048
        self.pos_embedding = nn.Parameter(torch.randn(1, num_patches, vit_dim))
        self.patch_to_embedding = nn.Linear(self.patch_dim, vit_dim)
        transformer_layer = nn.TransformerEncoderLayer(d_model=vit_dim, nhead=vit_heads, dim_feedforward=vit_dim * 4, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=vit_depth)
        self.transformer_output_to_conv = nn.Sequential(nn.Linear(vit_dim, self.patch_dim), nn.LayerNorm(self.patch_dim))
        d_ch = {'d0': 64, 'd1': 128, 'd2': 256, 'd3': 512, 'd4': 1024}
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
        self.X_0_0 = ConvBlock(64, d_ch['d0'])
        self.X_1_0 = ConvBlock(256, d_ch['d1'])
        self.X_0_1 = ConvBlock(d_ch['d0'] + d_ch['d1'], d_ch['d0'])
        self.X_2_0 = ConvBlock(512, d_ch['d2'])
        self.X_1_1 = ConvBlock(d_ch['d1'] + d_ch['d2'], d_ch['d1'])
        self.X_0_2 = ConvBlock(d_ch['d0'] * 2 + d_ch['d1'], d_ch['d0'])
        self.X_3_0 = ConvBlock(1024, d_ch['d3'])
        self.X_2_1 = ConvBlock(d_ch['d2'] + d_ch['d3'], d_ch['d2'])
        self.X_1_2 = ConvBlock(d_ch['d1'] * 2 + d_ch['d2'], d_ch['d1'])
        self.X_0_3 = ConvBlock(d_ch['d0'] * 3 + d_ch['d1'], d_ch['d0'])
        self.X_4_0 = ConvBlock(2048, d_ch['d4'])
        self.X_3_1 = ConvBlock(d_ch['d3'] + d_ch['d4'], d_ch['d3'])
        self.X_2_2 = ConvBlock(d_ch['d2'] * 2 + d_ch['d3'], d_ch['d2'])
        self.X_1_3 = ConvBlock(d_ch['d1'] * 3 + d_ch['d2'], d_ch['d1'])
        self.X_0_4 = ConvBlock(d_ch['d0'] * 4 + d_ch['d1'], d_ch['d0'])
        self.final_conv = nn.Conv2d(d_ch['d0'], n_classes, kernel_size=1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        bs, _, h, w = e4.shape
        trans_in = self.patch_to_embedding(e4.flatten(2).transpose(1, 2)) + self.pos_embedding
        trans_out = self.transformer_output_to_conv(self.transformer_encoder(trans_in)).transpose(1, 2).view(bs, self.patch_dim, h, w)
        x0_0 = self.X_0_0(e0)
        x1_0 = self.X_1_0(e1)
        x0_1 = self.X_0_1(torch.cat([x0_0, self.upsample(x1_0)], 1))
        x2_0 = self.X_2_0(e2)
        x1_1 = self.X_1_1(torch.cat([x1_0, self.upsample(x2_0)], 1))
        x0_2 = self.X_0_2(torch.cat([x0_0, x0_1, self.upsample(x1_1)], 1))
        x3_0 = self.X_3_0(e3)
        x2_1 = self.X_2_1(torch.cat([x2_0, self.upsample(x3_0)], 1))
        x1_2 = self.X_1_2(torch.cat([x1_0, x1_1, self.upsample(x2_1)], 1))
        x0_3 = self.X_0_3(torch.cat([x0_0, x0_1, x0_2, self.upsample(x1_2)], 1))
        x4_0 = self.X_4_0(trans_out)
        x3_1 = self.X_3_1(torch.cat([x3_0, self.upsample(x4_0)], 1))
        x2_2 = self.X_2_2(torch.cat([x2_0, x2_1, self.upsample(x3_1)], 1))
        x1_3 = self.X_1_3(torch.cat([x1_0, x1_1, x1_2, self.upsample(x2_2)], 1))
        x0_4 = self.X_0_4(torch.cat([x0_0, x0_1, x0_2, x0_3, self.upsample(x1_3)], 1))
        return F.interpolate(self.final_conv(x0_4), scale_factor=2, mode='bilinear', align_corners=True)

class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super().__init__()
        self.W_g = nn.Sequential(nn.Conv2d(F_g, F_int, 1), nn.BatchNorm2d(F_int))
        self.W_x = nn.Sequential(nn.Conv2d(F_l, F_int, 1), nn.BatchNorm2d(F_int))
        self.psi = nn.Sequential(nn.Conv2d(F_int, 1, 1), nn.BatchNorm2d(1), nn.Sigmoid())
        self.relu = nn.ReLU(inplace=True)
    def forward(self, g, x):
        psi = self.relu(self.W_g(g) + self.W_x(x))
        return x * self.psi(psi)

class AttentionUNet(nn.Module):
    def __init__(self, n_classes=1):
        super().__init__()
        base = models.resnet50(weights=None)
        layers = list(base.children())
        self.encoder0, self.encoder1 = nn.Sequential(*layers[:3]), nn.Sequential(*layers[3:5])
        self.encoder2, self.encoder3, self.encoder4 = layers[5], layers[6], layers[7]
        self.upconv3 = nn.ConvTranspose2d(2048, 1024, 2, 2)
        self.attn3 = AttentionGate(1024, 1024, 512)
        self.dec_conv3 = ConvBlock(2048, 1024)
        self.upconv2 = nn.ConvTranspose2d(1024, 512, 2, 2)
        self.attn2 = AttentionGate(512, 512, 256)
        self.dec_conv2 = ConvBlock(1024, 512)
        self.upconv1 = nn.ConvTranspose2d(512, 256, 2, 2)
        self.attn1 = AttentionGate(256, 256, 128)
        self.dec_conv1 = ConvBlock(512, 256)
        self.upconv0 = nn.ConvTranspose2d(256, 64, 2, 2)
        self.attn0 = AttentionGate(64, 64, 32)
        self.dec_conv0 = ConvBlock(128, 64)
        self.final_up = nn.ConvTranspose2d(64, 32, 2, 2)
        self.final_conv = nn.Conv2d(32, n_classes, 1)
    def forward(self, x):
        e0 = self.encoder0(x); e1 = self.encoder1(e0); e2 = self.encoder2(e1); e3 = self.encoder3(e2); e4 = self.encoder4(e3)
        d3 = self.upconv3(e4); x3 = self.attn3(d3, e3); d3 = self.dec_conv3(torch.cat((x3, d3), 1))
        d2 = self.upconv2(d3); x2 = self.attn2(d2, e2); d2 = self.dec_conv2(torch.cat((x2, d2), 1))
        d1 = self.upconv1(d2); x1 = self.attn1(d1, e1); d1 = self.dec_conv1(torch.cat((x1, d1), 1))
        d0 = self.upconv0(d1); x0 = self.attn0(d0, e0); d0 = self.dec_conv0(torch.cat((x0, d0), 1))
        return self.final_conv(self.final_up(d0))

# --- 6. LOAD SEGMENTATION MODELS ---
models_loaded = False
if test_df is not None:
    try:
        print("\n--- Loading segmentation models... ---")
        model1 = UNetWithResNet50Encoder(n_classes=1).to(device)
        model1.load_state_dict(torch.load(MODEL1_PATH, map_location=device))
        model1.eval()

        model2 = TransUNetPP(n_classes=1).to(device)
        model2.load_state_dict(torch.load(MODEL2_PATH, map_location=device))
        model2.eval()

        model3 = AttentionUNet(n_classes=1).to(device)
        model3.load_state_dict(torch.load(MODEL3_PATH, map_location=device))
        model3.eval()

        print("Segmentation models loaded successfully! âœ…")
        models_loaded = True
    except Exception as e:
        print(f"Error loading files: {e}")
        models_loaded = False

# --- FOCAL LOSS ---
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=3.0, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha  # keep None when oversampling
        self.reduction = reduction
    def forward(self, input, target):
        ce_loss = F.cross_entropy(input, target, reduction='none')
        pt = torch.exp(-ce_loss)
        loss = ((1 - pt) ** self.gamma) * ce_loss
        if self.alpha is not None:
            alpha_t = self.alpha[target.view(-1)]
            loss = alpha_t * loss
        return loss.mean() if self.reduction == 'mean' else loss.sum()

# =================================================================================
# --- 7. CLASSIFICATION STAGE (HYBRID MODEL) ---
# =================================================================================
print("\n\n--- Starting Classification Stage with Hybrid Model ---")
if not models_loaded or test_df is None:
    print("Skipping classification stage due to errors.")
else:
    try:
        # --- 7.1 Segmentation predictions ---
        print("\nGenerating segmentation predictions...")
        train_seg_dataset = SegmentationInputDataset(train_df)
        val_seg_dataset = SegmentationInputDataset(val_df)
        test_seg_dataset = SegmentationInputDataset(test_df)

        def get_segmentation_predictions(dataset, desc):
            loader = DataLoader(dataset, batch_size=16, shuffle=False, num_workers=2)
            all_preds = []
            with torch.no_grad():
                for images in tqdm(loader, desc=desc):
                    images = images.to(device)
                    p1 = model1(images)
                    p2 = model2(images)
                    p3 = model3(images)
                    ensembled = torch.sigmoid(torch.mean(torch.stack([p1, p2, p3]), dim=0))
                    all_preds.append((ensembled > 0.5).float().cpu())
            return torch.cat(all_preds)

        train_pred_masks = get_segmentation_predictions(train_seg_dataset, "Generating Train Masks")
        val_pred_masks = get_segmentation_predictions(val_seg_dataset, "Generating Val Masks")
        test_pred_masks = get_segmentation_predictions(test_seg_dataset, "Generating Test Masks")
        print("Segmentation predictions generated successfully.")

        # --- 7.2 Shape features from masks ---
        print("\nEngineering shape features from masks...")
        def get_shape_features(mask_tensor):
            feature_df = pd.DataFrame()
            for i in tqdm(range(len(mask_tensor)), desc="Calculating Features"):
                mask = mask_tensor[i, 0].numpy().astype(np.uint8)
                props = regionprops_table(mask, properties=('area', 'perimeter', 'solidity', 'eccentricity'))
                props_df = pd.DataFrame(props)
                if props_df.empty:
                    props_df = pd.DataFrame({'area': [0], 'perimeter': [0], 'solidity': [0], 'eccentricity': [0]})
                feature_df = pd.concat([feature_df, props_df.head(1)], ignore_index=True)
            feature_df['circularity'] = (4 * np.pi * feature_df['area']) / (feature_df['perimeter'] ** 2 + 1e-6)
            feature_df.fillna(0, inplace=True)
            return feature_df

        train_features_df = get_shape_features(train_pred_masks)
        val_features_df = get_shape_features(val_pred_masks)
        test_features_df = get_shape_features(test_pred_masks)

        scaler = StandardScaler()
        train_features = scaler.fit_transform(train_features_df)
        val_features = scaler.transform(val_features_df)
        test_features = scaler.transform(test_features_df)
        
        # --- 7.3 Datasets ---
        train_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.05, rotate_limit=10, p=0.5),
            A.RandomBrightnessContrast(p=0.3),
            A.GaussNoise(p=0.2),
            A.OpticalDistortion(p=0.2),
            A.CoarseDropout(max_holes=8, max_height=8, max_width=8, p=0.2),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        eval_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])
        minority_transform = A.Compose([
            A.Resize(height=IMAGE_SIZE[0], width=IMAGE_SIZE[1]),
            A.HorizontalFlip(p=0.8),
            A.VerticalFlip(p=0.5),
            A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.8),
            A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.5),
            A.GaussNoise(p=0.3),
            A.OpticalDistortion(p=0.3),
            A.CoarseDropout(max_holes=16, max_height=16, max_width=16, p=0.3),
            A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ToTensorV2()
        ])

        class HybridDataset(Dataset):
            def __init__(self, dataframe, pred_masks, shape_features, transform, minority_classes=None, minority_transform=None):
                self.df = dataframe
                self.masks = pred_masks
                self.features = shape_features
                self.transform = transform
                self.minority_classes = minority_classes if minority_classes else []
                self.minority_transform = minority_transform if minority_transform else transform
            def __len__(self):
                return len(self.df)
            def __getitem__(self, idx):
                row = self.df.iloc[idx]
                image = cv2.imread(row['image_file_path'])
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                label = int(row['class_label'])
                transformed = (self.minority_transform if label in self.minority_classes else self.transform)(image=image)
                img_tensor = transformed['image']
                pred_mask = self.masks[idx].to(img_tensor.device)
                # keep some context beyond the mask to help moderate class
                masked_image = 0.7 * img_tensor + 0.3 * (img_tensor * pred_mask.clone().detach())
                shape_feats = torch.tensor(self.features[idx], dtype=torch.float)
                label = torch.tensor(label, dtype=torch.long)
                return masked_image, shape_feats, label

        # Apply stronger aug to Moderate(1) + Severe(2)
        minority_classes_list = [1, 2]
        train_cls_dataset = HybridDataset(train_df, train_pred_masks, train_features,
                                          transform=train_transform,
                                          minority_classes=minority_classes_list,
                                          minority_transform=minority_transform)
        val_cls_dataset = HybridDataset(val_df, val_pred_masks, val_features, eval_transform)
        test_cls_dataset = HybridDataset(test_df, test_pred_masks, test_features, eval_transform)
        
        # --- 7.3.1 Balanced oversampling with WeightedRandomSampler ---
        labels_np = train_df['class_label'].to_numpy()
        unique_labels = np.unique(labels_np)
        class_counts = np.bincount(labels_np, minlength=unique_labels.max()+1)
        class_counts = np.where(class_counts == 0, 1, class_counts)  # avoid div-by-zero

        # sample weight = inverse of its class freq
        sample_weights = 1.0 / class_counts[labels_np]
        sample_weights = torch.from_numpy(sample_weights).float()

        # choose number of samples per epoch (balanced)
        num_samples_per_epoch = int(class_counts.max() * len(unique_labels))

        train_sampler = torch.utils.data.WeightedRandomSampler(
            weights=sample_weights,
            num_samples=num_samples_per_epoch,
            replacement=True
        )

        train_cls_loader = DataLoader(
            train_cls_dataset,
            batch_size=16,
            sampler=train_sampler,
            num_workers=2,
            pin_memory=True
        )
        val_cls_loader = DataLoader(val_cls_dataset, 16, shuffle=False, num_workers=2, pin_memory=True)
        test_cls_loader = DataLoader(test_cls_dataset, 16, shuffle=False, num_workers=2, pin_memory=True)
        print("\nHybrid datasets and loaders are ready with Balanced Oversampling. ğŸ“Š")

        # --- 7.4 Hybrid Classifier (EfficientNet-B0 + tabular fusion) ---
        class HybridClassifier(nn.Module):
            def __init__(self, n_classes=3, n_shape_features=5):
                super().__init__()
                try:
                    self.cnn_base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.DEFAULT)
                except Exception:
                    self.cnn_base = models.efficientnet_b0(weights=None)  # offline fallback
                cnn_feature_dim = self.cnn_base.classifier[-1].in_features
                self.cnn_base.classifier = nn.Identity()
                self.tabular_mlp = nn.Sequential(
                    nn.Linear(n_shape_features, 64), nn.ReLU(), nn.Dropout(0.4),
                    nn.Linear(64, 32), nn.ReLU(), nn.Dropout(0.4)
                )
                self.fusion_classifier = nn.Sequential(
                    nn.Linear(cnn_feature_dim + 32, 256), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(256, n_classes)
                )
            def forward(self, image, shape_features):
                cnn_features = self.cnn_base(image)
                tabular_features = self.tabular_mlp(shape_features)
                combined_features = torch.cat([cnn_features, tabular_features], dim=1)
                output = self.fusion_classifier(combined_features)
                return output
        
        # --- 7.5 Train ---
        cls_model = HybridClassifier(n_classes=3, n_shape_features=train_features.shape[1]).to(device)
        criterion = FocalLoss(alpha=None, gamma=3.0)  # <-- no alpha when oversampling
        optimizer = torch.optim.AdamW(cls_model.parameters(), lr=1e-4)
        scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.2, patience=5, verbose=True)
        num_epochs = 50
        patience = 10
        epochs_no_improve = 0
        best_val_accuracy = 0.0
        
        print("\n--- Starting Hybrid Model Training ---")
        for epoch in range(num_epochs):
            cls_model.train()
            running_loss = 0.0
            for images, shapes, labels in tqdm(train_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Train)"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                optimizer.zero_grad()
                outputs = cls_model(images, shapes)
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item()

            cls_model.eval()
            val_corrects = 0
            with torch.no_grad():
                for images, shapes, labels in tqdm(val_cls_loader, desc=f"Epoch {epoch + 1}/{num_epochs} (Validation)"):
                    images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                    outputs = cls_model(images, shapes)
                    _, preds = torch.max(outputs, 1)
                    val_corrects += torch.sum(preds == labels.data)

            val_accuracy = val_corrects.double() / len(val_cls_dataset)
            print(f"Epoch {epoch + 1} | Loss: {running_loss / max(1,len(train_cls_loader)):.4f} | Val Acc: {val_accuracy:.4f}")
            scheduler.step(val_accuracy)

            if val_accuracy > best_val_accuracy:
                best_val_accuracy = val_accuracy
                torch.save(cls_model.state_dict(), 'best_classifier_model.pth')
                epochs_no_improve = 0
                print(f"    ->âœ… New best model saved: {best_val_accuracy:.4f}")
            else:
                epochs_no_improve += 1
                print(f"    ->âŒ No improvement for {epochs_no_improve} epochs.")
            if epochs_no_improve >= patience:
                print(f"\nEarly stopping triggered after {patience} epochs.")
                break

        print("\n--- Training Finished ---")

        # --- 7.6 Evaluation ---
        print("\n--- Evaluating Hybrid Model on Test Set ---")
        cls_model.load_state_dict(torch.load('best_classifier_model.pth'))
        cls_model.eval()
        all_labels, all_preds = [], []
        with torch.no_grad():
            for images, shapes, labels in tqdm(test_cls_loader, desc="Testing"):
                images, shapes, labels = images.to(device), shapes.to(device), labels.to(device)
                outputs = cls_model(images, shapes)
                _, preds = torch.max(outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())
        
        class_names = ['Mild', 'Moderate', 'Severe']
        print("\n--- Classification Report ---")
        report = classification_report(all_labels, all_preds, target_names=class_names, zero_division=0, output_dict=True)
        print(classification_report(all_labels, all_preds, target_names=class_names, zero_division=0))
        
        print("\n--- Individual Class Accuracies ---")
        mild_accuracy = report['Mild']['recall']
        moderate_accuracy = report['Moderate']['recall']
        severe_accuracy = report['Severe']['recall']
        print(f"Accuracy for 'Mild' class: {mild_accuracy:.4f}")
        print(f"Accuracy for 'Moderate' class: {moderate_accuracy:.4f}")
        print(f"Accuracy for 'Severe' class: {severe_accuracy:.4f}")
        
        print("\n--- Confusion Matrix ---")
        cm = confusion_matrix(all_labels, all_preds, labels=range(len(class_names)))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
        fig, ax = plt.subplots(figsize=(8, 8))
        disp.plot(ax=ax, cmap=plt.cm.Blues)
        plt.title("Confusion Matrix for Hybrid Model (Balanced Oversampling)")
        plt.show()

    except Exception as e:
        print(f"\nAn error occurred: {e}")
        import traceback
        traceback.print_exc()
```
