```python
import os
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import models, transforms
import torchvision.transforms.functional as F_T
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
from tqdm import tqdm
from PIL import Image

# --- Imports for LIME ---
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries, slic
from skimage.transform import resize # Import resize

# --- Set Seeds for Reproducibility ---
def set_seed(seed):
    """Sets the seed for reproducibility."""
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# Set CUDA_LAUNCH_BLOCKING for easier debugging
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} ‚öôÔ∏è")

# --- Data Loading and Preparation ---
try:
    metadata_df = pd.read_csv('/kaggle/input/t2metadata/T2_age_gender_evaluation.csv')
    print("Successfully loaded metadata CSV.")
except FileNotFoundError:
    print("Error: 'T2_age_gender_evaluation.csv' not found. Please ensure the file is in the correct path.")
    metadata_df = pd.DataFrame(columns=['Patient ID', 'Age', 'Gender', 'Radiological Evaluation'])

# Rename and process columns
metadata_df.rename(columns={
    'Patient ID': 'ID',
    'Radiological Evaluation': 'radiological_evaluation'
}, inplace=True)
metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)

# Map numerical classes to names
class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
class_label_map = {1: 0, 2: 1, 3: 2}

metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
class_names = ['Mild', 'Moderate', 'Severe']
class_map = {0: 'Mild', 1: 'Moderate', 2: 'Severe'}

print("\n--- Classification Task Setup ---")
print("Class Mapping for Evaluation:", class_map)

metadata_df['ID'] = metadata_df['ID'].astype(str)
valid_ids = set(metadata_df['ID'].tolist())

def create_dataframe_from_ids(directory, allowed_ids):
    data = []
    if not os.path.exists(directory):
        print(f"Warning: Directory '{directory}' not found. Skipping.")
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])
    for folder_name in os.listdir(directory):
        if folder_name in allowed_ids:
            folder_path = os.path.join(directory, folder_name)
            images_dir = os.path.join(folder_path, 'images')
            masks_dir = os.path.join(folder_path, 'masks')
            if os.path.exists(images_dir) and os.path.exists(masks_dir):
                for image_file in os.listdir(images_dir):
                    image_path = os.path.join(images_dir, image_file)
                    mask_path = os.path.join(masks_dir, image_file)
                    if os.path.exists(mask_path):
                        data.append((folder_name, image_path, mask_path))
    return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

train_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/train'
valid_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/valid'
test_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/test'

train_df = pd.merge(create_dataframe_from_ids(train_dir, valid_ids), metadata_df, on='ID')
val_df = pd.merge(create_dataframe_from_ids(valid_dir, valid_ids), metadata_df, on='ID')
test_df = pd.merge(create_dataframe_from_ids(test_dir, valid_ids), metadata_df, on='ID')

print(f"\nTraining samples: {len(train_df)}, Validation samples: {len(val_df)}, Test samples: {len(test_df)}")

# --- Dataset and Transforms ---
class SegmentationTransform:
    def __init__(self, image_size=(224, 224), is_train=True):
        self.image_size = image_size
        self.is_train = is_train
    def __call__(self, image, mask):
        image = F_T.resize(image, self.image_size, interpolation=F_T.InterpolationMode.BILINEAR)
        mask = F_T.resize(mask, self.image_size, interpolation=F_T.InterpolationMode.NEAREST)
        if self.is_train:
            if random.random() > 0.5: image, mask = F_T.hflip(image), F_T.hflip(mask)
            if random.random() > 0.5: image, mask = F_T.vflip(image), F_T.vflip(mask)
            angle = random.uniform(-15, 15)
            image, mask = F_T.rotate(image, angle), F_T.rotate(mask, angle)
        image = F_T.to_tensor(image)
        mask = F_T.to_tensor(mask)
        image = F_T.normalize(image, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        mask = (mask > 0).float()
        return image, mask

class LiverDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image = Image.open(row['image_file_path']).convert('RGB')
        mask = Image.open(row['mask_file_path']).convert('L')
        class_label = torch.tensor(row['class_label'], dtype=torch.long)
        if self.transform:
            image, mask = self.transform(image, mask)
        return image, mask, class_label

# --- MODEL: CCT-Unet with Classification Head ---
class Tokenizer(nn.Module):
    """Convolutional Tokenizer for CCT."""
    def __init__(self, in_channels, embedding_dim, kernel_size=3, stride=1, padding=1):
        super().__init__()
        self.tokenizer = nn.Sequential(
            nn.Conv2d(in_channels, embedding_dim, kernel_size=kernel_size, stride=stride, padding=padding, bias=False),
            nn.ReLU(inplace=True)
        )
    def forward(self, x):
        return self.tokenizer(x)

class CCT(nn.Module):
    """Compact Convolutional Transformer Module."""
    def __init__(self, in_channels, embedding_dim=256, num_layers=4, num_heads=4, mlp_ratio=2.):
        super().__init__()
        self.tokenizer = Tokenizer(in_channels=in_channels, embedding_dim=embedding_dim)
        self.pos_embedding = nn.Parameter(torch.zeros(1, 1, embedding_dim))
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embedding_dim, nhead=num_heads, dim_feedforward=int(embedding_dim * mlp_ratio), batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

    def forward(self, x):
        B, C, H, W = x.shape
        x = self.tokenizer(x)
        x = x.flatten(2).transpose(1, 2)
        if self.pos_embedding.shape[1] != x.shape[1]:
            self.pos_embedding = nn.Parameter(F.interpolate(
                self.pos_embedding.permute(0, 2, 1), size=x.shape[1], mode='linear'
            ).permute(0, 2, 1))
        x = x + self.pos_embedding
        x = self.transformer_encoder(x)
        x = x.transpose(1, 2).view(B, -1, H, W)
        return x

class DecoderBlock(nn.Module):
    """U-Net decoder block."""
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Sequential(
            nn.Conv2d(out_channels + skip_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
    def forward(self, x, skip=None):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = self.relu(self.conv1(x))
        if skip is not None:
            x = torch.cat([x, skip], dim=1)
        return self.conv2(x)

class CCTUnet(nn.Module):
    """U-Net with a ResNet Encoder, CCT Bottleneck, and a Classification Head."""
    def __init__(self, n_seg_classes=1, n_cirrhosis_classes=3, pretrained=True):
        super().__init__()
        base_model = models.resnet34(weights=models.ResNet34_Weights.DEFAULT if pretrained else None)
        base_layers = list(base_model.children())
        self.e1 = nn.Sequential(*base_layers[:3])
        self.e2 = nn.Sequential(*base_layers[3:5])
        self.e3 = base_layers[5]
        self.e4 = base_layers[6]
        self.e5 = base_layers[7]
        self.cct_bottleneck = CCT(in_channels=512, embedding_dim=512, num_layers=3, num_heads=8)
        self.d1 = DecoderBlock(512, 256, 256)
        self.d2 = DecoderBlock(256, 128, 128)
        self.d3 = DecoderBlock(128, 64, 64)
        self.d4 = DecoderBlock(64, 64, 64)
        self.final_conv = nn.Conv2d(64, n_seg_classes, kernel_size=1)
        
        # New Classification Head
        self.classification_head = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(512, n_cirrhosis_classes)
        )

    def forward(self, x):
        # Encoder path
        s1 = self.e1(x)
        s2 = self.e2(s1)
        s3 = self.e3(s2)
        s4 = self.e4(s3)
        e_out = self.e5(s4)
        
        # Segmentation path
        b = self.cct_bottleneck(e_out)
        d = self.d1(b, s4)
        d = self.d2(d, s3)
        d = self.d3(d, s2)
        d = self.d4(d, s1)
        segmentation_output = self.final_conv(d)
        segmentation_output = F.interpolate(segmentation_output, scale_factor=2, mode='bilinear', align_corners=True)
        
        # Classification path
        classification_output = self.classification_head(e_out)
        
        return segmentation_output, classification_output

# --- Loss, Metrics, and Training Functions ---
class DiceBCELoss(nn.Module):
    def __init__(self, smooth=1e-6):
        super().__init__()
        self.smooth = smooth
    def forward(self, inputs, targets):
        inputs_sig = torch.sigmoid(inputs)
        inputs_flat = inputs_sig.view(-1)
        targets_flat = targets.view(-1)
        bce = F.binary_cross_entropy_with_logits(inputs, targets, reduction='mean')
        intersection = (inputs_flat * targets_flat).sum()
        dice_loss = 1 - (2. * intersection + self.smooth) / (inputs_flat.sum() + targets_flat.sum() + self.smooth)
        return bce + dice_loss

def calculate_metrics(preds, targets, smooth=1e-6):
    preds_sig = torch.sigmoid(preds)
    preds_bin = (preds_sig > 0.5).float()
    preds_flat, targets_flat = preds_bin.view(-1), targets.view(-1)
    intersection = (preds_flat * targets_flat).sum()
    total = preds_flat.sum() + targets_flat.sum()
    union = total - intersection
    dice = (2. * intersection + smooth) / (total + smooth)
    iou = (intersection + smooth) / (union + smooth)
    return dice.item(), iou.item()

def train_one_epoch(loader, model, optimizer, seg_loss_fn, cls_loss_fn, device):
    model.train()
    running_loss, total_seg_dice, total_cls_acc = 0.0, 0.0, 0.0
    for images, masks, labels in tqdm(loader, desc="Training"):
        images, masks, labels = images.to(device), masks.to(device), labels.to(device)
        optimizer.zero_grad()
        seg_outputs, cls_outputs = model(images)
        seg_loss = seg_loss_fn(seg_outputs, masks)
        cls_loss = cls_loss_fn(cls_outputs, labels)
        total_loss = seg_loss + cls_loss
        total_loss.backward()
        optimizer.step()
        running_loss += total_loss.item() * images.size(0)
        
        # Metrics for segmentation
        seg_dice, _ = calculate_metrics(seg_outputs, masks)
        total_seg_dice += seg_dice * images.size(0)
        
        # Metrics for classification
        _, predicted_cls = torch.max(cls_outputs, 1)
        total_cls_acc += (predicted_cls == labels).sum().item()
        
    avg_loss = running_loss / len(loader.dataset)
    avg_dice = total_seg_dice / len(loader.dataset)
    avg_acc = total_cls_acc / len(loader.dataset)
    return avg_loss, avg_dice, avg_acc

def evaluate_model(loader, model, seg_loss_fn, cls_loss_fn, device, class_mapping, dataset_name="Validation"):
    model.eval()
    all_preds_seg, all_masks, all_preds_cls, all_labels = [], [], [], []
    running_loss = 0.0
    with torch.no_grad():
        for images, masks, labels in tqdm(loader, desc=f"Evaluating on {dataset_name}"):
            images, masks = images.to(device), masks.to(device)
            seg_outputs, cls_outputs = model(images)
            running_loss += (seg_loss_fn(seg_outputs, masks).item() + cls_loss_fn(cls_outputs, labels.to(device)).item()) * images.size(0)
            all_preds_seg.append(seg_outputs.cpu()); all_masks.append(masks.cpu())
            all_preds_cls.append(cls_outputs.cpu()); all_labels.append(labels.cpu())
            
    all_preds_seg, all_masks = torch.cat(all_preds_seg), torch.cat(all_masks)
    all_preds_cls, all_labels = torch.cat(all_preds_cls), torch.cat(all_labels)

    avg_loss = running_loss / len(loader.dataset)
    overall_dice, overall_iou = calculate_metrics(all_preds_seg, all_masks)
    
    _, predicted_cls = torch.max(all_preds_cls, 1)
    overall_acc = (predicted_cls == all_labels).sum().item() / len(all_labels)

    print(f"\n--- {dataset_name} Report ---")
    print(f"Overall -> Loss: {avg_loss:.4f}, Segmentation Dice: {overall_dice:.4f}, Classification Accuracy: {overall_acc:.4f}")
    
    print("--- Per-Class Metrics ---")
    for class_idx, class_name in class_mapping.items():
        indices = (all_labels == class_idx).nonzero(as_tuple=True)[0]
        if len(indices) > 0:
            class_dice, class_iou = calculate_metrics(all_preds_seg[indices], all_masks[indices])
            print(f"Class '{class_name}' ({len(indices)} samples) -> Seg Dice: {class_dice:.4f}, mIoU: {class_iou:.4f}")
    return avg_loss, overall_dice

# --- FULL TRAINING LOOP ---
train_dataset = LiverDataset(train_df, transform=SegmentationTransform(is_train=True))
val_dataset = LiverDataset(val_df, transform=SegmentationTransform(is_train=False))
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)

model = CCTUnet(n_seg_classes=1, n_cirrhosis_classes=3, pretrained=True).to(device)
optimizer = AdamW(model.parameters(), lr=1e-4)
seg_criterion = DiceBCELoss()
cls_criterion = nn.CrossEntropyLoss()
num_epochs = 20
best_val_dice = 0.0

print("\n--- Starting Model Training with CCT-Unet with Classification Head ---")
if len(train_dataset) > 0:
    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")
        train_loss, train_dice, train_acc = train_one_epoch(train_loader, model, optimizer, seg_criterion, cls_criterion, device)
        print(f"Training -> Loss: {train_loss:.4f}, Seg Dice: {train_dice:.4f}, Cls Acc: {train_acc:.4f}")
        val_loss, val_dice = evaluate_model(val_loader, model, seg_criterion, cls_criterion, device, class_map)
        if val_dice > best_val_dice:
            best_val_dice = val_dice
            state_dict_to_save = {k: v for k, v in model.state_dict().items() if 'pos_embedding' not in k}
            torch.save(state_dict_to_save, 'best_segmentation_classification_model.pth')
            print(f"‚úÖ Model Saved! New Best Dice: {best_val_dice:.4f}")
else:
    print("üõë Training dataset is empty. Cannot start training.")

# --- FINAL TEST SET EVALUATION ---
print("\n--- Evaluating on the Test Set with the Best Model ---")
test_dataset_eval = LiverDataset(test_df, transform=SegmentationTransform(is_train=False))
test_loader = DataLoader(test_dataset_eval, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
try:
    model = CCTUnet(n_seg_classes=1, n_cirrhosis_classes=3, pretrained=False).to(device)
    model.load_state_dict(torch.load('best_segmentation_classification_model.pth', map_location=device), strict=False)
    print("Best model weights loaded for final test evaluation.")
    evaluate_model(test_loader, model, seg_criterion, cls_criterion, device, class_map, dataset_name="Test")
except FileNotFoundError:
    print("Model file 'best_segmentation_classification_model.pth' not found. Cannot perform final test evaluation.")
    model = None

# --- LIME EXPLANATION SECTION ---
print("\n--- Loading Best Model for LIME Explanation ---")
if model is None:
    print("Model not available because loading failed. Cannot perform LIME explanation.")
else:
    print("Using best model for LIME.")

def get_pil_transform():
    return transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def get_preprocess_transform():
    return transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def batch_predict_cirrhosis(images):
    model.eval()
    batch = torch.stack([get_preprocess_transform()(image) for image in images], dim=0).to(device).float()
    with torch.no_grad():
        _, logits_classification = model(batch)
    probs_classification = F.softmax(logits_classification, dim=1)
    return probs_classification.cpu().numpy()


def generate_lime_explanation_liver_focused(dataset, model, class_mapping):
    if model is None or len(dataset) == 0:
        print("Skipping LIME: model or dataset not available.")
        return

    print("\n--- Generating LIME Explanations (Liver-Focused) for Each Class ---")
    explainer = lime_image.LimeImageExplainer()
    test_df_viz = dataset.dataframe.drop_duplicates(subset=['class_label'])

    for _, row in test_df_viz.iterrows():
        class_label = row['class_label']
        class_name = row['class_name']
        image_path = row['image_file_path']
        mask_path = row['mask_file_path'] # Ground truth mask path

        original_image = Image.open(image_path).convert('RGB').resize((224, 224))
        ground_truth_mask_pil = Image.open(mask_path).convert('L').resize((224, 224))
        ground_truth_mask_np = np.array(ground_truth_mask_pil) > 0 # Binary numpy mask

        print(f"\nExplaining a sample for class: '{class_name}'...")
        
        # 1. Get the model's predicted liver mask for this image
        img_tensor_for_seg = get_preprocess_transform()(original_image).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_seg_mask_logits, _ = model(img_tensor_for_seg)
            pred_seg_mask_np = (torch.sigmoid(pred_seg_mask_logits) > 0.5).float().cpu().squeeze(0).permute(1, 2, 0).numpy()
            # Ensure it's a binary mask of the same size as the image (H, W) or (H, W, 1)
            pred_seg_mask_np = resize(pred_seg_mask_np, (224, 224), anti_aliasing=True).squeeze() > 0.5

        # Define a custom segmentation function for LIME based on the predicted liver mask
        # This function will generate superpixels ONLY within the predicted liver region.
        def custom_segmentation_fn(image):
            # LIME provides the image as (H, W, 3) numpy array
            # We use SLIC to get superpixels, but then filter them by the liver mask
            segments_slic = slic(image, n_segments=100, compactness=10, sigma=1, start_label=1)
            
            # Create a new segmentation map where non-liver pixels are assigned to a "background" superpixel (e.g., 0)
            # Or, even better, ensure superpixels outside the liver are grouped to a single ID.
            # For simplicity for LIME, we just want to ensure that if a superpixel is mostly outside the liver,
            # LIME can treat it as "background" for the explanation.
            
            # A more robust approach might be to zero out superpixels that don't overlap much with the liver.
            # For now, let's keep it simpler for direct usage with LIME's hide_color.
            
            # This is tricky because LIME's segmenter should ideally produce segments that respect the boundary.
            # A common trick is to mask out non-liver regions from the LIME explanation.
            # LIME's hide_and_replace_every_feature can use the segments directly.
            
            # For a truly "liver-focused" explanation, the segments themselves should be within the liver.
            # Let's try to adapt the segments to the predicted liver mask.
            
            # Option 1: Generate segments on the whole image, then mask the explanation.
            # This is simpler and often sufficient. LIME will still try to perturb outside,
            # but if those perturbations don't affect the classification, LIME won't highlight them.
            # The key is what `hide_color` represents.

            # Option 2 (more advanced): Create segments that respect the liver boundary.
            # This is harder with off-the-shelf SLIC. We'll stick to masking the explanation later.
            return segments_slic

        # When creating explanations, LIME perturbs superpixels by replacing them with hide_color.
        # We want the 'hide_color' to be applied *outside* the liver, but when a superpixel *inside* the liver
        # is hidden, it should be replaced with a neutral color (e.g., mean pixel value)
        
        # A simpler way to achieve liver-focused:
        # 1. Compute LIME explanation on the whole image.
        # 2. Apply the predicted liver mask to the resulting explanation.
        # This means only showing LIME regions that overlap with the liver.

        explanation = explainer.explain_instance(
            np.array(original_image), 
            batch_predict_cirrhosis,
            top_labels=1, 
            hide_color=0, # This hides by blacking out perturbed regions
            num_samples=1000
        )
        
        # Get LIME explanation mask
        temp_lime, lime_mask = explanation.get_image_and_mask(
            explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=True
        )

        # Apply the predicted liver mask to the LIME explanation
        # The lime_mask is (H, W) boolean. We want to only show explanation within pred_seg_mask_np
        
        # Ensure lime_mask has the same dimensions as pred_seg_mask_np
        # It usually does if temp_lime is resized to image dimensions
        
        # Create a combined mask: LIME-highlighted regions *AND* within the predicted liver.
        # The `temp_lime` already has non-highlighted regions hidden by `hide_rest=True` and `hide_color=0`.
        # We need to explicitly hide the parts of `temp_lime` that are *outside* the liver.
        
        # Convert original_image to float for blending
        original_image_float = np.array(original_image).astype(np.float32) / 255.0
        
        # Create an overlay for the LIME explanation, masked by the predicted liver
        lime_overlay = np.zeros_like(original_image_float)
        
        # Get the important superpixels identified by LIME
        segments = slic(np.array(original_image), n_segments=100, compactness=10, sigma=1, start_label=1)
        
        # The `explanation.local_exp` gives (segment_id, weight) pairs
        # We want to highlight segments with positive weight (contributing positively to the predicted class)
        
        # Create an importance map from LIME
        num_segments = np.max(segments) + 1 # +1 if segments start from 0
        
        # If segments start from 1 (due to start_label=1), then max(segments) is the number of segments
        # LIME's segment_id refers to the integer values in `segments`.
        
        # Create a mask for segments that are important and positive
        important_segments_mask = np.zeros(segments.shape, dtype=bool)
        for segment_id, weight in explanation.local_exp[explanation.top_labels[0]]:
            if weight > 0: # Highlight positive contributions
                important_segments_mask[segments == segment_id] = True

        # Now, combine this with the predicted liver mask
        final_highlight_mask = important_segments_mask & pred_seg_mask_np
        
        # Visualization: Use mark_boundaries on the original image with the final_highlight_mask
        # For mark_boundaries, the mask should be boolean.

        fig, axes = plt.subplots(1, 4, figsize=(24, 6))
        fig.suptitle(f"LIME Explanation for Class: {class_name} (Liver-Focused)", fontsize=16)

        axes[0].imshow(original_image); axes[0].set_title("Original Image"); axes[0].axis('off')
        axes[1].imshow(ground_truth_mask_np, cmap='gray'); axes[1].set_title("Ground Truth Mask"); axes[1].axis('off')
        axes[2].imshow(pred_seg_mask_np, cmap='gray'); axes[2].set_title("Model's Predicted Liver Mask"); axes[2].axis('off')
        
        # Plot the LIME explanation, highlighting only regions within the predicted liver
        # We use the original image and mark boundaries with our `final_highlight_mask`
        axes[3].imshow(mark_boundaries(original_image_float, final_highlight_mask, color=(1, 1, 0))); 
        axes[3].set_title("LIME: Cirrhosis Regions in Liver"); axes[3].axis('off')
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

# Run LIME on the test dataset (using untransformed images)
test_dataset_lime = LiverDataset(test_df, transform=None)
generate_lime_explanation_liver_focused(test_dataset_lime, model, class_map)
```

```python
import os
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import models, transforms
import torchvision.transforms.functional as F_T
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F
from torch.optim import AdamW
from tqdm import tqdm
from PIL import Image

# --- Imports for LIME ---
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries, slic
from skimage.transform import resize 

# --- Import for SWIN Transformer ---
import timm

# --- Set Seeds for Reproducibility ---
def set_seed(seed):
    """Sets the seed for reproducibility."""
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# Set CUDA_LAUNCH_BLOCKING for easier debugging
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} ‚öôÔ∏è")

# --- Data Loading and Preparation ---
try:
    metadata_df = pd.read_csv('/kaggle/input/t2metadata/T2_age_gender_evaluation.csv')
    print("Successfully loaded metadata CSV.")
except FileNotFoundError:
    print("Error: 'T2_age_gender_evaluation.csv' not found. Please ensure the file is in the correct path.")
    metadata_df = pd.DataFrame(columns=['Patient ID', 'Age', 'Gender', 'Radiological Evaluation'])

# Rename and process columns
metadata_df.rename(columns={
    'Patient ID': 'ID',
    'Radiological Evaluation': 'radiological_evaluation'
}, inplace=True)
metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)

# Map numerical classes to names
class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
class_label_map = {1: 0, 2: 1, 3: 2}

metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map)
class_names = ['Mild', 'Moderate', 'Severe']
class_map = {0: 'Mild', 1: 'Moderate', 2: 'Severe'}

print("\n--- Classification Task Setup ---")
print("Class Mapping for Evaluation:", class_map)

metadata_df['ID'] = metadata_df['ID'].astype(str)
valid_ids = set(metadata_df['ID'].tolist())

def create_dataframe_from_ids(directory, allowed_ids):
    data = []
    if not os.path.exists(directory):
        print(f"Warning: Directory '{directory}' not found. Skipping.")
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])
    for folder_name in os.listdir(directory):
        if folder_name in allowed_ids:
            folder_path = os.path.join(directory, folder_name)
            images_dir = os.path.join(folder_path, 'images')
            masks_dir = os.path.join(folder_path, 'masks')
            if os.path.exists(images_dir) and os.path.exists(masks_dir):
                for image_file in os.listdir(images_dir):
                    image_path = os.path.join(images_dir, image_file)
                    mask_path = os.path.join(masks_dir, image_file)
                    if os.path.exists(mask_path):
                        data.append((folder_name, image_path, mask_path))
    return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

train_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/train'
valid_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/valid'
test_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/test'

train_df = pd.merge(create_dataframe_from_ids(train_dir, valid_ids), metadata_df, on='ID')
val_df = pd.merge(create_dataframe_from_ids(valid_dir, valid_ids), metadata_df, on='ID')
test_df = pd.merge(create_dataframe_from_ids(test_dir, valid_ids), metadata_df, on='ID')

print(f"\nTraining samples: {len(train_df)}, Validation samples: {len(val_df)}, Test samples: {len(test_df)}")

# --- Dataset and Transforms ---
class SegmentationTransform:
    def __init__(self, image_size=(224, 224), is_train=True):
        self.image_size = image_size
        self.is_train = is_train
    def __call__(self, image, mask):
        image = F_T.resize(image, self.image_size, interpolation=F_T.InterpolationMode.BILINEAR)
        mask = F_T.resize(mask, self.image_size, interpolation=F_T.InterpolationMode.NEAREST)
        if self.is_train:
            if random.random() > 0.5: image, mask = F_T.hflip(image), F_T.hflip(mask)
            if random.random() > 0.5: image, mask = F_T.vflip(image), F_T.vflip(mask)
            angle = random.uniform(-15, 15)
            image, mask = F_T.rotate(image, angle), F_T.rotate(mask, angle)
        image = F_T.to_tensor(image)
        mask = F_T.to_tensor(mask)
        image = F_T.normalize(image, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        mask = (mask > 0).float()
        return image, mask

class LiverDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image = Image.open(row['image_file_path']).convert('RGB')
        mask = Image.open(row['mask_file_path']).convert('L')
        class_label = torch.tensor(row['class_label'], dtype=torch.long)
        if self.transform:
            image, mask = self.transform(image, mask)
        return image, mask, class_label

# --- MODEL: Swin-Unet with Classification Head ---
class DecoderBlock(nn.Module):
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Sequential(
            nn.Conv2d(out_channels + skip_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
    def forward(self, x, skip=None):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = self.relu(self.conv1(x))
        if skip is not None:
            x = torch.cat([x, skip], dim=1)
        return self.conv2(x)

class SwinUnet(nn.Module):
    def __init__(self, n_seg_classes=1, n_cirrhosis_classes=3, pretrained=True):
        super().__init__()
        self.n_seg_classes = n_seg_classes
        self.encoder = timm.create_model(
            'swin_tiny_patch4_window7_224',
            pretrained=pretrained,
            features_only=True,
            output_stride=32
        )
        encoder_channels = self.encoder.feature_info.channels()
        
        # Decoder Blocks
        self.d4 = DecoderBlock(encoder_channels[3], encoder_channels[2], 256)
        self.d3 = DecoderBlock(256, encoder_channels[1], 128)
        self.d2 = DecoderBlock(128, encoder_channels[0], 64)
        self.d1 = DecoderBlock(64, 0, 32)
        
        self.final_conv = nn.Conv2d(32, n_seg_classes, kernel_size=1)

        # Classification Head
        self.classification_head = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(encoder_channels[3], n_cirrhosis_classes)
        )

    def forward(self, x):
        skips_raw = self.encoder(x)
        skips = [s.permute(0, 3, 1, 2) for s in skips_raw]
        s0, s1, s2, s3 = skips[0], skips[1], skips[2], skips[3]
        
        # Segmentation path
        d = self.d4(s3, s2)
        d = self.d3(d, s1)
        d = self.d2(d, s0)
        d = self.d1(d)
        segmentation_output = self.final_conv(d)
        
        # Upsample to original image size
        segmentation_output = F.interpolate(segmentation_output, size=(224, 224), mode='bilinear', align_corners=False)
        
        # Classification path
        classification_output = self.classification_head(s3)
        
        return segmentation_output, classification_output

# --- Loss, Metrics, and Training Functions ---
class DiceBCELoss(nn.Module):
    def __init__(self, smooth=1e-6):
        super().__init__()
        self.smooth = smooth
    def forward(self, inputs, targets):
        inputs_sig = torch.sigmoid(inputs)
        inputs_flat = inputs_sig.view(-1)
        targets_flat = targets.view(-1)
        bce = F.binary_cross_entropy_with_logits(inputs, targets, reduction='mean')
        intersection = (inputs_flat * targets_flat).sum()
        dice_loss = 1 - (2. * intersection + self.smooth) / (inputs_flat.sum() + targets_flat.sum() + self.smooth)
        return bce + dice_loss

def calculate_metrics(preds, targets, smooth=1e-6):
    preds_sig = torch.sigmoid(preds)
    preds_bin = (preds_sig > 0.5).float()
    preds_flat, targets_flat = preds_bin.view(-1), targets.view(-1)
    intersection = (preds_flat * targets_flat).sum()
    total = preds_flat.sum() + targets_flat.sum()
    union = total - intersection
    dice = (2. * intersection + smooth) / (total + smooth)
    iou = (intersection + smooth) / (union + smooth)
    return dice.item(), iou.item()

def train_one_epoch(loader, model, optimizer, seg_loss_fn, cls_loss_fn, device):
    model.train()
    running_loss, total_seg_dice, total_cls_acc = 0.0, 0.0, 0.0
    for images, masks, labels in tqdm(loader, desc="Training"):
        images, masks, labels = images.to(device), masks.to(device), labels.to(device)
        optimizer.zero_grad()
        seg_outputs, cls_outputs = model(images)
        seg_loss = seg_loss_fn(seg_outputs, masks)
        cls_loss = cls_loss_fn(cls_outputs, labels)
        total_loss = seg_loss + cls_loss
        total_loss.backward()
        optimizer.step()
        running_loss += total_loss.item() * images.size(0)
        
        seg_dice, _ = calculate_metrics(seg_outputs, masks)
        total_seg_dice += seg_dice * images.size(0)
        
        _, predicted_cls = torch.max(cls_outputs, 1)
        total_cls_acc += (predicted_cls == labels).sum().item()
        
    avg_loss = running_loss / len(loader.dataset)
    avg_dice = total_seg_dice / len(loader.dataset)
    avg_acc = total_cls_acc / len(loader.dataset)
    return avg_loss, avg_dice, avg_acc

def evaluate_model(loader, model, seg_loss_fn, cls_loss_fn, device, class_mapping, dataset_name="Validation"):
    model.eval()
    all_preds_seg, all_masks, all_preds_cls, all_labels = [], [], [], []
    running_loss = 0.0
    with torch.no_grad():
        for images, masks, labels in tqdm(loader, desc=f"Evaluating on {dataset_name}"):
            images, masks = images.to(device), masks.to(device)
            seg_outputs, cls_outputs = model(images)
            running_loss += (seg_loss_fn(seg_outputs, masks).item() + cls_loss_fn(cls_outputs, labels.to(device)).item()) * images.size(0)
            all_preds_seg.append(seg_outputs.cpu()); all_masks.append(masks.cpu())
            all_preds_cls.append(cls_outputs.cpu()); all_labels.append(labels.cpu())
            
    all_preds_seg, all_masks = torch.cat(all_preds_seg), torch.cat(all_masks)
    all_preds_cls, all_labels = torch.cat(all_preds_cls), torch.cat(all_labels)

    avg_loss = running_loss / len(loader.dataset)
    overall_dice, overall_iou = calculate_metrics(all_preds_seg, all_masks)
    
    _, predicted_cls = torch.max(all_preds_cls, 1)
    overall_acc = (predicted_cls == all_labels).sum().item() / len(all_labels)

    print(f"\n--- {dataset_name} Report ---")
    print(f"Overall -> Loss: {avg_loss:.4f}, Segmentation Dice: {overall_dice:.4f}, Classification Accuracy: {overall_acc:.4f}")
    
    print("--- Per-Class Metrics ---")
    for class_idx, class_name in class_mapping.items():
        indices = (all_labels == class_idx).nonzero(as_tuple=True)[0]
        if len(indices) > 0:
            class_dice, class_iou = calculate_metrics(all_preds_seg[indices], all_masks[indices])
            print(f"Class '{class_name}' ({len(indices)} samples) -> Seg Dice: {class_dice:.4f}, mIoU: {class_iou:.4f}")
    return avg_loss, overall_dice

# --- FULL TRAINING LOOP ---
train_dataset = LiverDataset(train_df, transform=SegmentationTransform(is_train=True))
val_dataset = LiverDataset(val_df, transform=SegmentationTransform(is_train=False))
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)

model = SwinUnet(n_seg_classes=1, n_cirrhosis_classes=3, pretrained=True).to(device)
optimizer = AdamW(model.parameters(), lr=1e-4)
seg_criterion = DiceBCELoss()
cls_criterion = nn.CrossEntropyLoss()
num_epochs = 20
best_val_dice = 0.0

print("\n--- Starting Model Training with Swin-Unet with Classification Head ---")
if len(train_dataset) > 0:
    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")
        train_loss, train_dice, train_acc = train_one_epoch(train_loader, model, optimizer, seg_criterion, cls_criterion, device)
        print(f"Training -> Loss: {train_loss:.4f}, Seg Dice: {train_dice:.4f}, Cls Acc: {train_acc:.4f}")
        val_loss, val_dice = evaluate_model(val_loader, model, seg_criterion, cls_criterion, device, class_map)
        if val_dice > best_val_dice:
            best_val_dice = val_dice
            torch.save(model.state_dict(), 'best_segmentation_classification_model_swin.pth')
            print(f"‚úÖ Model Saved! New Best Dice: {best_val_dice:.4f}")
else:
    print("üõë Training dataset is empty. Cannot start training.")

# --- FINAL TEST SET EVALUATION ---
print("\n--- Evaluating on the Test Set with the Best Model ---")
test_dataset_eval = LiverDataset(test_df, transform=SegmentationTransform(is_train=False))
test_loader = DataLoader(test_dataset_eval, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
try:
    model = SwinUnet(n_seg_classes=1, n_cirrhosis_classes=3, pretrained=False).to(device)
    model.load_state_dict(torch.load('best_segmentation_classification_model_swin.pth', map_location=device), strict=False)
    print("Best model weights loaded for final test evaluation.")
    evaluate_model(test_loader, model, seg_criterion, cls_criterion, device, class_map, dataset_name="Test")
except FileNotFoundError:
    print("Model file 'best_segmentation_classification_model_swin.pth' not found. Cannot perform final test evaluation.")
    model = None

# --- LIME EXPLANATION SECTION ---
print("\n--- Loading Best Model for LIME Explanation ---")
if model is None:
    print("Model not available because loading failed. Cannot perform LIME explanation.")
else:
    print("Using best model for LIME.")

def get_pil_transform():
    return transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def get_preprocess_transform():
    return transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def batch_predict_cirrhosis(images):
    model.eval()
    batch = torch.stack([get_preprocess_transform()(image) for image in images], dim=0).to(device).float()
    with torch.no_grad():
        _, logits_classification = model(batch)
    probs_classification = F.softmax(logits_classification, dim=1)
    return probs_classification.cpu().numpy()

def generate_lime_explanation_liver_focused(dataset, model, class_mapping):
    if model is None or len(dataset) == 0:
        print("Skipping LIME: model or dataset not available.")
        return

    print("\n--- Generating LIME Explanations (Liver-Focused) for Each Class ---")
    explainer = lime_image.LimeImageExplainer()
    test_df_viz = dataset.dataframe.drop_duplicates(subset=['class_label'])

    for _, row in test_df_viz.iterrows():
        class_label = row['class_label']
        class_name = row['class_name']
        image_path = row['image_file_path']
        mask_path = row['mask_file_path']

        original_image = Image.open(image_path).convert('RGB').resize((224, 224))
        ground_truth_mask_pil = Image.open(mask_path).convert('L').resize((224, 224))
        ground_truth_mask_np = np.array(ground_truth_mask_pil) > 0

        print(f"\nExplaining a sample for class: '{class_name}'...")
        
        # 1. Get the model's predicted liver mask for this image
        img_tensor_for_seg = get_preprocess_transform()(original_image).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_seg_mask_logits, _ = model(img_tensor_for_seg)
            pred_seg_mask_np = (torch.sigmoid(pred_seg_mask_logits) > 0.5).float().cpu().squeeze(0).permute(1, 2, 0).numpy()
            pred_seg_mask_np = resize(pred_seg_mask_np, (224, 224), anti_aliasing=True).squeeze() > 0.5

        explanation = explainer.explain_instance(
            np.array(original_image), 
            batch_predict_cirrhosis,
            top_labels=1, 
            hide_color=0, 
            num_samples=1000
        )
        
        segments = slic(np.array(original_image), n_segments=100, compactness=10, sigma=1, start_label=1)
        
        important_segments_mask = np.zeros(segments.shape, dtype=bool)
        for segment_id, weight in explanation.local_exp[explanation.top_labels[0]]:
            if weight > 0:
                important_segments_mask[segments == segment_id] = True

        final_highlight_mask = important_segments_mask & pred_seg_mask_np
        
        fig, axes = plt.subplots(1, 4, figsize=(24, 6))
        fig.suptitle(f"LIME Explanation for Class: {class_name} (Liver-Focused)", fontsize=16)

        axes[0].imshow(original_image); axes[0].set_title("Original Image"); axes[0].axis('off')
        axes[1].imshow(ground_truth_mask_np, cmap='gray'); axes[1].set_title("Ground Truth Mask"); axes[1].axis('off')
        axes[2].imshow(pred_seg_mask_np, cmap='gray'); axes[2].set_title("Model's Predicted Liver Mask"); axes[2].axis('off')
        
        axes[3].imshow(mark_boundaries(np.array(original_image) / 255.0, final_highlight_mask, color=(1, 1, 0))); 
        axes[3].set_title("LIME: Cirrhosis Regions in Liver"); axes[3].axis('off')
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

# Run LIME on the test dataset (using untransformed images)
test_dataset_lime = LiverDataset(test_df, transform=None)
generate_lime_explanation_liver_focused(test_dataset_lime, model, class_map)
```

```python
import os
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchvision import models, transforms
import torchvision.transforms.functional as F_T
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
from tqdm import tqdm
from PIL import Image
from sklearn.metrics import accuracy_score, f1_score

# --- Imports for LIME ---
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries

# --- Import for Vision Transformer models ---
import timm

# --- Set Seeds for Reproducibility ---
def set_seed(seed):
    """Sets the seed for reproducibility."""
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# Set CUDA_LAUNCH_BLOCKING for easier debugging
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device} ‚öôÔ∏è")

# --- Data Loading and Preparation ---
try:
    metadata_df = pd.read_csv('/kaggle/input/t2metadata/T2_age_gender_evaluation.csv')
    print("Successfully loaded metadata CSV.")
except FileNotFoundError:
    print("Error: 'T2_age_gender_evaluation.csv' not found. Please ensure the file is in the correct path.")
    metadata_df = pd.DataFrame(columns=['Patient ID', 'Age', 'Gender', 'Radiological Evaluation'])

# Rename and process columns
metadata_df.rename(columns={
    'Patient ID': 'ID',
    'Radiological Evaluation': 'radiological_evaluation'
}, inplace=True)
metadata_df.dropna(subset=['radiological_evaluation'], inplace=True)
metadata_df['radiological_evaluation'] = metadata_df['radiological_evaluation'].astype(int)

# Map numerical classes to names
class_name_map = {1: 'Mild', 2: 'Moderate', 3: 'Severe'}
class_label_map = {1: 0, 2: 1, 3: 2} # Explicitly define the mapping

metadata_df['class_name'] = metadata_df['radiological_evaluation'].map(class_name_map)
metadata_df['class_label'] = metadata_df['radiological_evaluation'].map(class_label_map) # Use the explicit map
class_names = ['Mild', 'Moderate', 'Severe']
class_map = {0: 'Mild', 1: 'Moderate', 2: 'Severe'} # Ensure this is also correct

print("\n--- Classification Task Setup ---")
print("Class Mapping for Evaluation:", class_map)

metadata_df['ID'] = metadata_df['ID'].astype(str)
valid_ids = set(metadata_df['ID'].tolist())

def create_dataframe_from_ids(directory, allowed_ids):
    data = []
    if not os.path.exists(directory):
        print(f"Warning: Directory '{directory}' not found. Skipping.")
        return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])
    for folder_name in os.listdir(directory):
        if folder_name in allowed_ids:
            folder_path = os.path.join(directory, folder_name)
            images_dir = os.path.join(folder_path, 'images')
            masks_dir = os.path.join(folder_path, 'masks')
            if os.path.exists(images_dir) and os.path.exists(masks_dir):
                for image_file in os.listdir(images_dir):
                    image_path = os.path.join(images_dir, image_file)
                    mask_path = os.path.join(masks_dir, image_file)
                    if os.path.exists(mask_path):
                        data.append((folder_name, image_path, mask_path))
    return pd.DataFrame(data, columns=['ID', 'image_file_path', 'mask_file_path'])

train_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/train'
valid_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/valid'
test_dir = '/kaggle/input/datat2/Cirrhosis_T2_2D/test'

train_df = pd.merge(create_dataframe_from_ids(train_dir, valid_ids), metadata_df, on='ID')
val_df = pd.merge(create_dataframe_from_ids(valid_dir, valid_ids), metadata_df, on='ID')
test_df = pd.merge(create_dataframe_from_ids(test_dir, valid_ids), metadata_df, on='ID')

print(f"\nTraining samples: {len(train_df)}, Validation samples: {len(val_df)}, Test samples: {len(test_df)}")

# --- Dataset and Transforms ---
class SegmentationTransform:
    def __init__(self, image_size=(224, 224), is_train=True):
        self.image_size = image_size
        self.is_train = is_train
    def __call__(self, image, mask):
        image = F_T.resize(image, self.image_size, interpolation=F_T.InterpolationMode.BILINEAR)
        mask = F_T.resize(mask, self.image_size, interpolation=F_T.InterpolationMode.NEAREST)
        if self.is_train:
            if random.random() > 0.5: image, mask = F_T.hflip(image), F_T.hflip(mask)
            if random.random() > 0.5: image, mask = F_T.vflip(image), F_T.vflip(mask)
            angle = random.uniform(-15, 15)
            image, mask = F_T.rotate(image, angle), F_T.rotate(mask, angle)
        image = F_T.to_tensor(image)
        mask = F_T.to_tensor(mask)
        image = F_T.normalize(image, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        mask = (mask > 0).float()
        return image, mask

class LiverDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform
    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        image = Image.open(row['image_file_path']).convert('RGB')
        mask = Image.open(row['mask_file_path']).convert('L')
        class_label = torch.tensor(row['class_label'], dtype=torch.long)
        if self.transform:
            image, mask = self.transform(image, mask)
        return image, mask, class_label

# --- MODEL: EaNet-Unet ---
class ExternalAttention(nn.Module):
    """Implementation of External Attention."""
    def __init__(self, d_model, S=64):
        super().__init__()
        self.mk = nn.Linear(d_model, S, bias=False)
        self.mv = nn.Linear(S, d_model, bias=False)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        attn = self.mk(x)
        attn = self.softmax(attn)
        attn = attn / torch.sum(attn, dim=2, keepdim=True)
        out = self.mv(attn)
        return out

class EaNet(nn.Module):
    """A single block of EaNet, combining ExternalAttention with a FeedForward network."""
    def __init__(self, in_channels, d_model, S=64, mlp_ratio=2.):
        super().__init__()
        self.d_model = d_model
        self.proj_in = nn.Conv2d(in_channels, d_model, kernel_size=1)
        self.attn = ExternalAttention(d_model, S)
        self.norm1 = nn.LayerNorm(d_model)
        self.mlp = nn.Sequential(
            nn.Linear(d_model, int(d_model * mlp_ratio)),
            nn.GELU(),
            nn.Linear(int(d_model * mlp_ratio), d_model)
        )
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):
        B, C, H, W = x.shape
        x = self.proj_in(x)
        x_flat = x.flatten(2).transpose(1, 2)
        x_attn = self.attn(x_flat)
        x_flat = x_flat + self.norm1(x_attn)
        x_mlp = self.mlp(x_flat)
        x_out = x_flat + self.norm2(x_mlp)
        return x_out.transpose(1, 2).view(B, self.d_model, H, W)

class DecoderBlock(nn.Module):
    """U-Net decoder block."""
    def __init__(self, in_channels, skip_channels, out_channels):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU(inplace=True)
        self.conv2 = nn.Sequential(
            nn.Conv2d(out_channels + skip_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )
    def forward(self, x, skip=None):
        x = F.interpolate(x, scale_factor=2, mode='bilinear', align_corners=True)
        x = self.relu(self.conv1(x))
        if skip is not None:
            x = torch.cat([x, skip], dim=1)
        return self.conv2(x)

class EaNetUnet(nn.Module):
    """Multi-Task EaNet model with a ResNet encoder, EaNet bottleneck, and two heads."""
    def __init__(self, num_seg_classes=1, num_cls_classes=3, pretrained=True):
        super().__init__()
        base_model = models.resnet34(weights=models.ResNet34_Weights.DEFAULT if pretrained else None)
        base_layers = list(base_model.children())
        self.e1 = nn.Sequential(*base_layers[:3])
        self.e2 = nn.Sequential(*base_layers[3:5])
        self.e3 = base_layers[5]
        self.e4 = base_layers[6]
        self.e5 = base_layers[7]
        self.eanet_bottleneck = EaNet(in_channels=512, d_model=512, S=64)
        
        # Segmentation Decoder
        self.d1 = DecoderBlock(512, 256, 256)
        self.d2 = DecoderBlock(256, 128, 128)
        self.d3 = DecoderBlock(128, 64, 64)
        self.d4 = DecoderBlock(64, 64, 64)
        self.final_conv_seg = nn.Conv2d(64, num_seg_classes, kernel_size=1)
        
        # Classification Head
        self.classification_head = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(512, num_cls_classes) # Correct linear layer input
        )

    def forward(self, x):
        # Encoder
        s1 = self.e1(x) # 64 channels, /2
        s2 = self.e2(s1) # 64 channels, /4
        s3 = self.e3(s2) # 128 channels, /8
        s4 = self.e4(s3) # 256 channels, /16
        e_out = self.e5(s4) # 512 channels, /32
        
        # EANet Bottleneck
        b = self.eanet_bottleneck(e_out)
        
        # Classification
        classification_output = self.classification_head(b)

        # Segmentation Decoder
        d = self.d1(b, s4)
        d = self.d2(d, s3)
        d = self.d3(d, s2)
        d = self.d4(d, s1)
        segmentation_output = self.final_conv_seg(d)
        segmentation_output = F.interpolate(segmentation_output, scale_factor=2, mode='bilinear', align_corners=True)
        
        return segmentation_output, classification_output

# --- Multi-Task Loss ---
class MultiTaskLoss(nn.Module):
    def __init__(self, alpha=0.5, beta=0.5):
        super().__init__()
        self.seg_loss_fn = DiceBCELoss()
        self.cls_loss_fn = nn.CrossEntropyLoss()
        self.alpha = alpha # Weight for segmentation loss
        self.beta = beta # Weight for classification loss

    def forward(self, seg_preds, seg_targets, cls_preds, cls_targets):
        seg_loss = self.seg_loss_fn(seg_preds, seg_targets)
        cls_loss = self.cls_loss_fn(cls_preds, cls_targets)
        return (self.alpha * seg_loss) + (self.beta * cls_loss), seg_loss, cls_loss

# --- Loss, Metrics, and Training Functions ---
class DiceBCELoss(nn.Module):
    def __init__(self, smooth=1e-6):
        super().__init__()
        self.smooth = smooth
    def forward(self, inputs, targets):
        inputs_sig = torch.sigmoid(inputs)
        inputs_flat = inputs_sig.view(-1)
        targets_flat = targets.view(-1)
        bce = F.binary_cross_entropy_with_logits(inputs, targets, reduction='mean')
        intersection = (inputs_flat * targets_flat).sum()
        dice_loss = 1 - (2. * intersection + self.smooth) / (inputs_flat.sum() + targets_flat.sum() + self.smooth)
        return bce + dice_loss

def calculate_metrics(preds, targets, smooth=1e-6):
    preds_sig = torch.sigmoid(preds)
    preds_bin = (preds_sig > 0.5).float()
    preds_flat, targets_flat = preds_bin.view(-1), targets.view(-1)
    intersection = (preds_flat * targets_flat).sum()
    total = preds_flat.sum() + targets_flat.sum()
    union = total - intersection
    dice = (2. * intersection + smooth) / (total + smooth)
    iou = (intersection + smooth) / (union + smooth)
    return dice.item(), iou.item()

def train_one_epoch(loader, model, optimizer, loss_fn, device):
    model.train()
    running_loss, running_seg_loss, running_cls_loss = 0.0, 0.0, 0.0
    total_dice, total_iou = 0.0, 0.0
    for images, masks, labels in tqdm(loader, desc="Training"):
        images, masks, labels = images.to(device), masks.to(device), labels.to(device)
        optimizer.zero_grad()
        seg_preds, cls_preds = model(images)
        total_loss, seg_loss, cls_loss = loss_fn(seg_preds, masks, cls_preds, labels)
        total_loss.backward()
        optimizer.step()
        running_loss += total_loss.item() * images.size(0)
        running_seg_loss += seg_loss.item() * images.size(0)
        running_cls_loss += cls_loss.item() * images.size(0)
        dice, iou = calculate_metrics(seg_preds, masks)
        total_dice += dice * images.size(0)
        total_iou += iou * images.size(0)
    return (running_loss / len(loader.dataset),
            running_seg_loss / len(loader.dataset),
            running_cls_loss / len(loader.dataset),
            total_dice / len(loader.dataset),
            total_iou / len(loader.dataset))

def evaluate_model(loader, model, loss_fn, device, class_mapping, dataset_name="Validation"):
    model.eval()
    all_seg_preds, all_masks, all_cls_preds, all_labels = [], [], [], []
    running_loss = 0.0
    with torch.no_grad():
        for images, masks, labels in tqdm(loader, desc=f"Evaluating on {dataset_name}"):
            images, masks, labels = images.to(device), masks.to(device), labels.to(device)
            seg_preds, cls_preds = model(images)
            total_loss, _, _ = loss_fn(seg_preds, masks, cls_preds, labels)
            running_loss += total_loss.item() * images.size(0)
            all_seg_preds.append(seg_preds.cpu()); all_masks.append(masks.cpu()); all_cls_preds.append(cls_preds.cpu()); all_labels.append(labels.cpu())
    all_seg_preds, all_masks = torch.cat(all_seg_preds), torch.cat(all_masks)
    all_cls_preds, all_labels = torch.cat(all_cls_preds), torch.cat(all_labels)

    # Segmentation Metrics
    avg_loss, (overall_dice, overall_iou) = running_loss / len(loader.dataset), calculate_metrics(all_seg_preds, all_masks)
    
    # Classification Metrics
    cls_preds_labels = torch.argmax(all_cls_preds, dim=1)
    cls_accuracy = accuracy_score(all_labels.numpy(), cls_preds_labels.numpy())
    cls_f1 = f1_score(all_labels.numpy(), cls_preds_labels.numpy(), average='weighted')

    print(f"\n--- {dataset_name} Report ---")
    print(f"Overall Multi-Task Loss: {avg_loss:.4f}")
    print(f"Segmentation -> Dice: {overall_dice:.4f}, mIoU: {overall_iou:.4f}")
    print(f"Classification -> Accuracy: {cls_accuracy:.4f}, F1-Score: {cls_f1:.4f}")
    print("--- Per-Class Segmentation Metrics ---")
    for class_idx, class_name in class_mapping.items():
        indices = (all_labels == class_idx).nonzero(as_tuple=True)[0]
        if len(indices) > 0:
            class_dice, class_iou = calculate_metrics(all_seg_preds[indices], all_masks[indices])
            print(f"Class '{class_name}' ({len(indices)} samples) -> Dice: {class_dice:.4f}, mIoU: {class_iou:.4f}")
    return avg_loss, overall_dice

# --- FULL TRAINING LOOP ---
train_dataset = LiverDataset(train_df, transform=SegmentationTransform(is_train=True))
val_dataset = LiverDataset(val_df, transform=SegmentationTransform(is_train=False))
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)

# Instantiate the new EaNetUnet model
model = EaNetUnet(num_seg_classes=1, num_cls_classes=len(class_names), pretrained=True).to(device)
optimizer = AdamW(model.parameters(), lr=1e-4)
criterion = MultiTaskLoss()
num_epochs = 20
best_val_dice = 0.0

print("\n--- Starting Model Training with EaNet-Unet ---")
if len(train_dataset) > 0:
    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")
        train_loss, train_seg_loss, train_cls_loss, train_dice, train_iou = train_one_epoch(train_loader, model, optimizer, criterion, device)
        print(f"Training -> Total Loss: {train_loss:.4f} (Seg: {train_seg_loss:.4f}, Cls: {train_cls_loss:.4f}), Dice: {train_dice:.4f}")
        val_loss, val_dice = evaluate_model(val_loader, model, criterion, device, class_map)
        if val_dice > best_val_dice:
            best_val_dice = val_dice
            torch.save(model.state_dict(), 'best_segmentation_model.pth')
            print(f"‚úÖ Model Saved! New Best Dice: {best_val_dice:.4f}")
else:
    print("üõë Training dataset is empty. Cannot start training.")

# --- FINAL TEST SET EVALUATION ---
print("\n--- Evaluating on the Test Set with the Best Model ---")
test_dataset_eval = LiverDataset(test_df, transform=SegmentationTransform(is_train=False))
test_loader = DataLoader(test_dataset_eval, batch_size=16, shuffle=False, num_workers=2, pin_memory=True)
try:
    model = EaNetUnet(num_seg_classes=1, num_cls_classes=len(class_names), pretrained=False).to(device)
    model.load_state_dict(torch.load('best_segmentation_model.pth', map_location=device))
    print("Best model weights loaded for final test evaluation.")
    evaluate_model(test_loader, model, criterion, device, class_map, dataset_name="Test")

except FileNotFoundError:
    print("Model file 'best_segmentation_model.pth' not found. Cannot perform final test evaluation.")
    model = None # Ensure model is None if loading fails


# --- LIME EXPLANATION SECTION ---
print("\n--- Loading Best Model for LIME Explanation ---")
if model is None:
    print("Model not available because loading failed. Cannot perform LIME explanation.")
else:
    print("Using best model for LIME.")

def get_pil_transform():
    return transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def get_preprocess_transform():
    return transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def batch_predict(images):
    """Corrected batch_predict for LIME classification explanation."""
    model.eval()
    batch = torch.stack([get_preprocess_transform()(image) for image in images], dim=0).to(device).float()
    with torch.no_grad():
        _, logits = model(batch) # Use the classification output
    probs = F.softmax(logits, dim=1) # Apply softmax for multi-class classification
    return probs.cpu().numpy()

def generate_lime_explanation(dataset, model, class_mapping):
    if model is None or len(dataset) == 0:
        print("Skipping LIME: model or dataset not available.")
        return

    print("\n--- Generating LIME Explanations for Each Class ---")
    explainer = lime_image.LimeImageExplainer()
    test_df_viz = dataset.dataframe.drop_duplicates(subset=['class_label'])

    for _, row in test_df_viz.iterrows():
        class_label = row['class_label']
        class_name = row['class_name']
        image_path = row['image_file_path']
        mask_path = row['mask_file_path']

        original_image = Image.open(image_path).convert('RGB').resize((224, 224))
        ground_truth_mask = Image.open(mask_path).convert('L').resize((224, 224))

        print(f"\nExplaining a sample for class: '{class_name}'...")
        explanation = explainer.explain_instance(np.array(original_image), batch_predict, top_labels=1, hide_color=0, num_samples=1000)
        
        # Get LIME mask for the predicted class
        temp_img, lime_mask = explanation.get_image_and_mask(
            explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=True)
        
        # Get the predicted segmentation mask
        img_tensor = get_pil_transform()(original_image).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_seg, _ = model(img_tensor)
            pred_mask = (torch.sigmoid(pred_seg) > 0.5).float().cpu().squeeze(0).permute(1, 2, 0)
        
        fig, axes = plt.subplots(1, 4, figsize=(24, 6))
        fig.suptitle(f"LIME Explanation for Class: {class_name}", fontsize=16)

        axes[0].imshow(original_image); axes[0].set_title("Original Image"); axes[0].axis('off')
        axes[1].imshow(ground_truth_mask, cmap='gray'); axes[1].set_title("Ground Truth Mask"); axes[1].axis('off')
        axes[2].imshow(pred_mask, cmap='gray'); axes[2].set_title("Model's Predicted Seg. Mask"); axes[2].axis('off')
        axes[3].imshow(mark_boundaries(temp_img / 2 + 0.5, lime_mask)); axes[3].set_title("LIME: Important Regions for Classification"); axes[3].axis('off')
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

# Run LIME on the test dataset (using untransformed images)
test_dataset_lime = LiverDataset(test_df, transform=None)
generate_lime_explanation(test_dataset_lime, model, class_map)
```

```python
# --- LIME EXPLANATION SECTION ---
print("\n--- Loading Best Model for LIME Explanation ---")
# Instantiate and load the multi-task model
try:
    model = EaNetUnet(num_seg_classes=1, num_cls_classes=len(class_names), pretrained=False).to(device)
    model.load_state_dict(torch.load('best_segmentation_model.pth', map_location=device))
    print("Best model weights loaded for LIME explanation.")
except FileNotFoundError:
    print("Model file 'best_segmentation_model.pth' not found. Cannot perform LIME explanation.")
    model = None

if model is None:
    print("Model not available because loading failed. Cannot perform LIME explanation.")
else:
    print("Using best model for LIME.")

def get_pil_transform():
    return transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def get_preprocess_transform():
    return transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

def batch_predict_cirrhosis(images):
    """
    Corrected batch_predict for LIME. It uses the classification head of the
    multi-task model to return class probabilities.
    """
    model.eval()
    batch = torch.stack([get_preprocess_transform()(image) for image in images], dim=0).to(device).float()
    with torch.no_grad():
        # The model returns two outputs: segmentation and classification.
        _, logits_classification = model(batch)
    
    # Apply softmax to get probabilities for the multi-class classification.
    probs_classification = F.softmax(logits_classification, dim=1)
    return probs_classification.cpu().numpy()

def generate_lime_explanation_liver_focused(dataset, model, class_mapping):
    """
    Generates LIME explanations and combines them with the model's segmentation mask
    for a more targeted visualization of important regions within the liver.
    """
    if model is None or len(dataset) == 0:
        print("Skipping LIME: model or dataset not available.")
        return

    print("\n--- Generating LIME Explanations (Liver-Focused) for Each Class ---")
    explainer = lime_image.LimeImageExplainer()
    
    # Get one unique sample from each class (Mild, Moderate, Severe)
    test_df_viz = dataset.dataframe.drop_duplicates(subset=['class_label'])

    for _, row in test_df_viz.iterrows():
        class_name = row['class_name']
        image_path = row['image_file_path']
        mask_path = row['mask_file_path']

        original_image = Image.open(image_path).convert('RGB').resize((224, 224))
        ground_truth_mask_np = np.array(Image.open(mask_path).convert('L').resize((224, 224))) > 0
        
        print(f"\nExplaining a sample for class: '{class_name}'...")
        
        # 1. Get the model's predicted liver mask for this image.
        img_tensor_for_seg = get_preprocess_transform()(original_image).unsqueeze(0).to(device)
        with torch.no_grad():
            pred_seg_mask_logits, _ = model(img_tensor_for_seg)
            pred_seg_mask_np = (torch.sigmoid(pred_seg_mask_logits) > 0.5).float().cpu().squeeze(0).permute(1, 2, 0).numpy()
            pred_seg_mask_np = resize(pred_seg_mask_np, (224, 224), anti_aliasing=True).squeeze() > 0.5
        
        # 2. Generate LIME explanation for the classification task.
        explanation = explainer.explain_instance(
            np.array(original_image), 
            batch_predict_cirrhosis,
            top_labels=1, 
            hide_color=0, 
            num_samples=1000,
            segmentation_fn=slic
        )
        
        # 3. Use LIME's internal segmentation to get the important segments.
        temp_img, lime_mask = explanation.get_image_and_mask(
            explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=False
        )

        # 4. Combine LIME's highlighted regions with the predicted liver mask.
        final_highlight_mask = (lime_mask > 0) & pred_seg_mask_np

        # 5. Visualize the results.
        fig, axes = plt.subplots(1, 4, figsize=(24, 6))
        fig.suptitle(f"LIME Explanation for Class: {class_name} (Liver-Focused)", fontsize=16)

        axes[0].imshow(original_image); axes[0].set_title("Original Image"); axes[0].axis('off')
        axes[1].imshow(ground_truth_mask_np, cmap='gray'); axes[1].set_title("Ground Truth Mask"); axes[1].axis('off')
        axes[2].imshow(pred_seg_mask_np, cmap='gray'); axes[2].set_title("Model's Predicted Liver Mask"); axes[2].axis('off')
        
        # Mark boundaries on the original image, highlighting only the intersection.
        axes[3].imshow(mark_boundaries(np.array(original_image) / 255.0, final_highlight_mask, color=(1, 1, 0))); 
        axes[3].set_title("LIME: Cirrhosis Regions in Liver"); axes[3].axis('off')
        
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

# Run LIME on the test dataset (using untransformed images)
test_dataset_lime = LiverDataset(test_df, transform=None)
generate_lime_explanation_liver_focused(test_dataset_lime, model, class_map)
```
